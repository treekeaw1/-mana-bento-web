```python
import pandas as pd
import numpy as np
from collections import Counter
from datetime import datetime, timedelta
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler # เพิ่ม StandardScaler
from sklearn.metrics import accuracy_score, classification_report
import warnings

# Suppress all warnings for cleaner output
warnings.filterwarnings('ignore')

# --- 0. ข้อมูล CSV ล่าสุด ---
csv_content = """วันที่,รางวัลที่ 1 (6 หลัก),เลข 2 ตัวล่าง
2025/06/16,507392,06
2025/06/01,559352,20
2025/05/16,251309,87
2025/05/02,213388,06
2025/04/16,266227,85
2025/04/01,669687,36
2025/03/16,757563,32
2025/03/01,818894,54
2025/02/16,847377,50
2025/02/01,558700,51
2025/01/17,807779,23
2025/01/02,730209,51
2024/12/16,097863,21
2024/12/01,669843,61
2024/11/16,187221,38
2024/11/01,536044,32
2024/10/16,482962,00
2024/10/01,718665,59
2024/09/16,608662,37
2024/09/01,199606,94
2024/08/16,095867,28
2024/08/01,407041,46
2024/07/16,367336,21
2024/07/01,434503,89
2024/06/16,518504,31
2024/06/01,530593,42
2024/05/16,205690,60
2024/05/02,980116,17
2024/04/16,943598,79
2024/04/01,803481,90
2024/03/16,997626,78
2024/03/01,253603,79
2024/02/16,941395,43
2024/02/01,607063,09
2024/01/17,105979,61
2023/12/30,625544,89
2023/12/16,356757,85
2023/12/01,251097,91
2023/11/16,557990,14
2023/11/01,743951,63
2023/10/16,931446,44
2023/10/01,727202,66
2023/09/16,320812,46
2023/09/01,915478,91
2023/08/16,471782,67
2023/07/31,260453,11
2023/07/16,169530,62
2023/07/01,922605,16
2023/06/16,264872,30
2023/06/01,125272,09
2023/05/16,132903,99
2023/05/02,843019,65
2023/04/16,984906,71
2023/04/01,087907,99
2023/03/16,025873,73
2023/03/01,417652,55
2023/02/16,590417,80
2023/02/01,297411,92
2023/01/17,812519,47
2022/12/30,157196,58
2022/12/16,845093,14
2022/12/01,375805,08
2022/11/16,121789,64
2022/11/01,913106,70
2022/10/16,613106,15
2022/10/01,484669,50
2022/09/16,943703,75
2022/09/01,929332,83
2022/08/16,331583,42
2022/08/01,436594,14
2022/07/16,620405,53
2022/07/01,981417,61
2022/06/16,361807,92
2022/06/01,319196,02
2022/05/16,155012,06
2022/05/02,658642,09
2022/04/16,395919,58
2022/04/01,970618,10
2022/03/16,737867,03
2022/03/01,061905,07
2022/02/17,098597,57
2022/02/01,944308,30
2022/01/17,880159,92
2021/12/30,819068,36
2021/12/16,639235,83
2021/12/01,077258,82
2021/11/16,032761,57
2021/11/01,045037,95
2021/10/16,386372,38
2021/10/01,578171,83
2021/09/16,070935,90
2021/09/01,114475,79
2021/08/16,046750,23
2021/08/01,910261,69
2021/07/16,556725,70
2021/07/01,713517,29
2021/06/16,691861,17
2021/06/01,292972,45
2021/05/16,684579,14
2021/05/02,501272,18
2021/04/16,100787,56
2021/04/01,472270,05
2021/03/16,890422,19
2021/03/01,835538,73
2021/02/16,424603,39
2021/02/01,912307,97
2021/01/17,384395,15
2020/12/30,803628,19
2020/12/16,201303,70
2020/12/01,100994,84
2020/11/16,972661,46
2020/11/01,506404,40
2020/10/16,286051,38
2020/10/01,837893,59
2020/09/16,244083,57
2020/09/01,999997,98
2020/08/16,945811,88
2020/08/01,569391,92
2020/07/16,873286,53
2020/07/01,347258,83
2020/06/16,516967,64
2020/06/01,831567,24
2020/04/01,051095,22
2020/03/16,503446,77
2020/03/01,875938,98
2020/02/16,781403,94
2020/02/01,589227,06
2020/01/17,491774,68
2019/12/30,510541,81
2019/12/16,529924,97
2019/12/01,453522,81
2019/11/16,017223,32
2019/11/01,967375,79
2019/10/16,812564,15
2019/10/01,691197,59
2019/09/16,340388,85
2019/09/01,798787,20
2019/08/16,775476,89
2019/08/01,387006,58
2019/07/15,369765,88
2019/07/01,943647,86
2019/06/16,174055,29
2019/06/01,516461,46
2019/05/16,962526,71
2019/05/02,061324,25
2019/04/16,570331,23
2019/04/01,109767,52
2019/03/16,724628,64
2019/03/01,345650,65
2019/02/16,074824,56
2019/02/01,967134,04
2019/01/17,197079,65
2018/12/30,735867,02
2018/12/16,356564,62
2018/12/01,021840,67
2018/11/16,989903,16
2018/11/01,149840,58
2018/10/16,200515,93
2018/10/01,452643,99
2018/09/16,149760,79
2018/09/01,734510,26
2018/08/16,586117,10
2018/08/01,386602,78
2018/07/16,596324,27
2018/07/01,963623,83
2018/06/16,223131,46
2018/06/01,988117,95
2018/05/16,075629,20
2018/05/02,248038,85
2018/04/16,739229,60
2018/04/01,412073,85
2018/03/16,218559,82
2018/03/02,759415,29
2018/02/16,309915,39
2018/02/01,026853,31
2018/01/17,203823,50
2017/12/30,911234,98
2017/12/16,955596,89
2017/12/01,451005,33
2017/11/16,292391,98
2017/11/01,533726,85
2017/10/16,413494,86
2017/10/01,880714,52
2017/09/16,170143,71
2017/09/01,143224,65
2017/08/16,715431,37
2017/08/01,756519,36
2017/07/16,820327,87
2017/07/01,112360,26
2017/06/16,943142,47
2017/06/01,053630,61
2017/05/16,454891,53
2017/05/02,008656,35
2017/04/16,816729,40
2017/04/01,392785,80
2017/03/16,273863,92
2017/03/01,978453,78
2017/02/16,229116,14
2017/02/01,054672,42
2017/01/17,145157,25
2016/12/30,377712,46
2016/12/16,435286,35
2016/12/01,086069,77
2016/11/16,858383,44
2016/11/01,785438,86
2016/10/16,571947,98
2016/10/01,887102,33
2016/09/16,240650,42
2016/09/01,638684,62
2016/08/16,254004,33
2016/08/01,272932,57
2016/07/16,449764,55
2016/07/01,082460,53
2016/06/16,073816,79
2016/06/01,511825,14
2016/05/16,141737,98
2016/05/02,399459,02
2016/04/16,221609,87
2016/04/01,066720,92
2016/03/16,134918,32
2016/03/01,439686,06
2016/02/16,356364,98
2016/02/01,927800,09
2016/01/17,304371,50
2015/12/30,008217,02
2015/12/17,930255,08
2015/12/01,915350,78
2015/11/16,795283,03
2015/11/01,361211,45
2015/10/16,968630,62
2015/10/01,594825,07
2015/09/16,743148,06
2015/09/01,021094,89
2015/08/16,033363,40
2015/08/01,518677,53
2015/07/16,121507,49
2015/07/01,759049,26
2015/06/16,644742,05
2015/06/02,388881,65
2015/05/16,011421,38
2015/05/02,543466,30
2015/04/16,506260,38
2015/04/01,605704,70
2015/03/16,048151,92
2015/03/01,240237,34
2015/02/16,001864,90
2015/02/01,155537,79
2015/01/16,244351,74
2014/12/30,461704,57
2014/12/16,948354,90
2014/12/01,480449,11
2014/11/16,479804,25
2014/11/01,206608,44
2014/10/16,656409,94
2014/10/01,375615,44
2014/09/16,772269,35
2014/09/01,856763,22
2014/08/16,662842,91
2014/08/01,766391,82
2014/07/16,468728,45
2014/07/01,378477,39
2014/06/16,673920,95
2014/06/01,781198,18
2014/05/16,087523,20
2014/05/02,103297,52
2014/04/16,153406,26
2014/04/01,028866,95
2014/03/16,531404,79
2014/03/01,906318,35
2014/02/16,384245,01
2014/02/01,180149,95
2014/01/16,306902,52
2013/12/30,561072,48
2013/12/16,341767,79
2013/12/01,168795,27
2013/11/16,806925,28
2013/11/01,739804,47
2013/10/16,963289,60
2013/10/01,647882,14
2013/09/16,562684,63
2013/09/01,548123,05
2013/08/16,321327,20
2013/08/01,356435,82
2013/07/16,566996,86
2013/07/01,646905,51
2013/06/16,289673,69
2013/06/01,935489,90
2013/05/16,687125,56
2013/05/02,603458,07
2013/04/16,843846,86
2013/04/01,571688,53
2013/03/16,968433,52
2013/03/01,976241,37
2013/02/16,368257,09
2013/02/01,565566,66
2013/01/16,820981,08
2012/12/30,302358,00
2012/12/16,529524,72
2012/12/01,110443,43
2012/11/16,639500,15
2012/11/01,524694,63
2012/10/16,281343,28
2012/10/01,124025,58
2012/09/16,540143,79
2012/09/01,329997,07
2012/08/16,683877,28
2012/08/01,895590,50
2012/07/16,904050,11
2012/07/01,915900,60
2012/06/16,159373,51
2012/06/01,882727,38
2012/05/16,814418,31
2012/05/02,889501,29
2012/04/16,583470,62
2012/04/01,257562,69
2012/03/16,607064,08
2012/03/01,222518,79
2012/02/16,648684,18
2012/02/01,320605,32
2012/01/16,451445,81
2011/12/30,526402,65
2011/12/16,884178,21
2011/12/01,408147,02
2011/11/16,997777,57
2011/11/01,805540,54
2011/10/16,955756,83
2011/10/01,511052,15
2011/09/16,731198,28
2011/09/01,724533,85
2011/08/16,536960,62
2011/08/01,218756,12
2011/07/16,116556,12
2011/07/01,622953,51
2011/06/16,351276,88
2011/06/01,562370,46
2011/05/16,406417,05
2011/05/02,054136,85
2011/04/16,825988,44
2011/04/01,814931,01
2011/03/16,593331,96
2011/03/01,656037,97
2011/02/16,481746,27
2011/02/01,610089,55
2011/01/16,281062,23
2010/12/30,884112,49
2010/12/16,334380,24
2010/12/01,181752,09
2010/11/16,813993,43
2010/11/01,191100,59
2010/10/16,621377,42
2010/10/01,488372,02
2010/09/16,017422,66
2010/09/01,354656,11
2010/08/16,911097,64
2010/08/01,210008,10
2010/07/16,180387,34
2010/07/01,480239,68
2010/06/16,500104,73
2010/06/01,444874,81
2010/05/16,480012,12
2010/05/02,360371,06
2010/04/16,211743,96
2010/04/01,959517,22
2010/03/16,364222,97
2010/03/01,215227,97
2010/02/16,133707,03
2010/02/01,186312,14
2010/01/16,073577,67
2009/12/30,994304,87
2009/12/16,685141,05
2009/12/01,776980,59
2009/11/16,055986,58
2009/11/01,689140,85
2009/10/16,258487,00
2009/10/01,169387,06
2009/09/16,202912,48
2009/09/01,015865,32
2009/08/16,462933,96
2009/08/01,154986,92
2009/07/16,000816,94
2009/07/01,207542,66
2009/06/16,930456,15
2009/06/01,777661,26
2009/05/16,111411,54
2009/05/02,294452,11
2009/04/16,368415,33
2009/04/01,816578,50
2009/03/16,268812,36
2009/03/01,553091,67
2009/02/16,038730,93
2009/02/01,534533,69
2009/01/16,743212,25
2008/12/30,218596,22
2008/12/16,074114,25
2008/12/01,205434,05
2008/11/16,002612,20
2008/11/01,272028,76
2008/10/16,431277,98
2008/10/01,882911,67
2008/09/16,012377,56
2008/09/01,695993,09
2008/08/16,380377,36
2008/08/01,850348,11
2008/07/16,257374,41
2008/07/01,943671,50
2008/06/16,729111,75
2008/06/01,414875,35
2008/05/16,329231,69
2008/05/02,453011,62
2008/04/16,982800,64
2008/04/01,012653,71
2008/03/16,074946,33
2008/03/01,936685,05
2008/02/16,137054,80
2008/02/01,212684,26
2008/01/16,556010,81
2007/12/30,595411,81
2007/12/16,513501,96
2007/12/01,113410,18
2007/11/16,562481,73
2007/11/01,927907,88
2007/10/16,032988,48
2007/10/01,430667,76
2007/09/16,499336,45
2007/09/01,331810,69
2007/08/16,476207,93
2007/08/01,429924,29
2007/07/16,527384,77
2007/07/01,565151,76
2007/06/16,393194,41
2007/06/01,836393,05
2007/05/16,232897,25
2007/05/02,430374,81
2007/04/16,405105,63
2007/04/01,622780,93
2007/03/16,876763,85
2007/03/01,742425,61
2007/02/16,277859,95
2007/02/01,769925,56
2007/01/16,838739,55
2006/12/30,778584,07
2006/12/16,147977,45
2006/12/01,270052,12
2006/11/16,562856,94
2006/11/01,910957,29
2006/10/16,264825,58
2006/10/01,952335,92
2006/09/16,217948,41
2006/09/01,381761,44
2006/08/16,977486,91
2006/08/01,238654,88
2006/07/16,512434,48
2006/07/01,952890,65
2006/06/16,100935,17
2006/06/01,810850,99
2006/05/16,100344,71
2006/05/02,024554,27
2006/04/16,038564,49
2006/04/01,738365,95
2006/03/16,936177,30
2006/03/01,582473,43
2006/02/16,317009,66
2006/02/01,412729,87
2006/01/16,432747,79
2005/12/30,492955,94
2005/12/16,449565,86
2005/12/01,388551,17
2005/11/16,742518,80
2005/11/01,970577,98
2005/10/16,041072,71
2005/10/01,766482,24
2005/09/16,214768,10
2005/09/01,316933,17
2005/08/16,475560,68
2005/08/01,961633,26
2005/07/16,477452,13
2005/07/01,009554,87
2005/06/16,793070,44
2005/06/01,176893,35
2005/05/16,867134,97
2005/05/02,772467,43
2005/04/16,119327,10
2005/04/01,815753,69
2005/03/16,196345,03
2005/03/01,800751,93
2005/02/16,816422,53
2005/02/01,540054,34
2005/01/16,335022,08
2004/12/30,168858,28
2004/12/16,479372,17
2004/12/01,504658,69
2004/11/16,754622,64
2004/11/01,185966,23
2004/10/16,355858,62
2004/10/01,110866,66
2004/09/16,923373,59
2004/09/01,096597,70
2004/08/16,335921,59
2004/08/02,868990,45
2004/07/16,205588,25
2004/07/01,312471,66
2004/06/16,208713,87
2004/06/01,614144,72
2004/05/16,589207,13
2004/05/02,653403,91
2004/04/16,705832,12
2004/04/01,196391,62
2004/03/16,615366,69
2004/03/01,697483,50
2004/02/16,698002,00
2004/02/01,216822,77
2004/01/16,731342,96
2003/12/30,739447,64
2003/12/16,177947,87
2003/12/01,991307,78
2003/11/16,238511,68
2003/11/01,941438,47
2003/10/16,305500,03
2003/10/01,912040,43
2003/09/16,600589,53
2003/09/01,187813,92
2003/08/16,354771,00
2003/08/01,766098,91
2003/07/16,679545,75
"""

# --- 1. การเตรียมและทำความสะอาดข้อมูล (รวมถึง Feature Engineering) ---
print("--- เริ่มต้นการเตรียมและประมวลผลข้อมูล ---")

# อ่านข้อมูล CSV
df = pd.read_csv(pd.io.common.StringIO(csv_content))

# แปลงคอลัมน์ 'วันที่' เป็น datetime object
df['วันที่'] = pd.to_datetime(df['วันที่'], format='%Y/%m/%d')
df = df.rename(columns={'วันที่': 'Date', 'รางวัลที่ 1 (6 หลัก)': 'เลขหกหลักรางวัลที่1', 'เลข 2 ตัวล่าง': 'เลขท้าย2ตัว'})

# เรียงลำดับข้อมูลตามวันที่
df = df.sort_values(by='Date').reset_index(drop=True)

# เติมเลขศูนย์หน้าให้ครบ 6 หลักสำหรับรางวัลที่ 1 และ 2 หลักสำหรับเลขท้าย 2 ตัวล่าง
df['เลขหกหลักรางวัลที่1'] = df['เลขหกหลักรางวัลที่1'].astype(str).str.zfill(6)
df['เลขท้าย2ตัว'] = df['เลขท้าย2ตัว'].astype(str).str.zfill(2)

# แยกเลขท้าย 3 ตัวรางวัลที่ 1 (จาก 3 หลักสุดท้ายของเลขรางวัลที่ 1)
df['เลขท้าย3ตัวรางวัลที่1'] = df['เลขหกหลักรางวัลที่1'].str[-3:]

# สร้าง Features
for i in range(6):
    df[f'R1_Digit_{i+1}'] = df['เลขหกหลักรางวัลที่1'].str[i].astype(int)
for i in range(2):
    df[f'L2_Digit_{i+1}'] = df['เลขท้าย2ตัว'].str[i].astype(int)

df['R1_Last3_Hundreds'] = df['เลขท้าย3ตัวรางวัลที่1'].str[0].astype(int)
df['R1_Last3_Tens'] = df['เลขท้าย3ตัวรางวัลที่1'].str[1].astype(int)
df['R1_Last3_Units'] = df['เลขท้าย3ตัวรางวัลที่1'].str[2].astype(int)
df['L2_Tens'] = df['เลขท้าย2ตัว'].str[0].astype(int)
df['L2_Units'] = df['เลขท้าย2ตัว'].str[1].astype(int)

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek
df['DayOfMonth'] = df['Date'].dt.day

df['R1_Sum'] = df['เลขหกหลักรางวัลที่1'].apply(lambda x: sum(int(d) for d in str(x)))
df['R1_Even_Count'] = df['เลขหกหลักรางวัลที่1'].apply(lambda x: sum(1 for d in str(x) if int(d) % 2 == 0))
df['L2_Sum'] = df['เลขท้าย2ตัว'].apply(lambda x: sum(int(d) for d in str(x))) # เพิ่ม L2_Sum
df['L2_Even_Count'] = df['เลขท้าย2ตัว'].apply(lambda x: sum(1 for d in str(x) if int(d) % 2 == 0)) # เพิ่ม L2_Even_Count


# Features จากงวดก่อนหน้า (Prev_L2, Prev_R1_Last3)
# ใช้ .copy() เพื่อหลีกเลี่ยง SettingWithCopyWarning
df['Prev_L2'] = df['เลขท้าย2ตัว'].shift(1).fillna('00').astype(str).copy()
df['Prev_R1_Last3'] = df['เลขท้าย3ตัวรางวัลที่1'].shift(1).fillna('000').astype(str).copy()

df['Prev_L2_Tens'] = df['Prev_L2'].str[0].astype(int)
df['Prev_L2_Units'] = df['Prev_L2'].str[1].astype(int)
df['Prev_R1_Last3_Hundreds'] = df['Prev_R1_Last3'].str[0].astype(int)
df['Prev_R1_Last3_Tens'] = df['Prev_R1_Last3'].str[1].astype(int)
df['Prev_R1_Last3_Units'] = df['Prev_R1_Last3'].str[2].astype(int)

# เพิ่ม Lag Features เพิ่มเติม (งวดก่อนหน้า 2-3 งวด)
for lag in range(2, 4): # Lag 2 และ Lag 3
    df[f'Prev_L2_Lag{lag}'] = df['เลขท้าย2ตัว'].shift(lag).fillna('00').astype(str)
    df[f'Prev_R1_Last3_Lag{lag}'] = df['เลขท้าย3ตัวรางวัลที่1'].shift(lag).fillna('000').astype(str)

    df[f'Prev_L2_Lag{lag}_Tens'] = df[f'Prev_L2_Lag{lag}'].str[0].astype(int)
    df[f'Prev_L2_Lag{lag}_Units'] = df[f'Prev_L2_Lag{lag}'].str[1].astype(int)
    df[f'Prev_R1_Last3_Lag{lag}_Hundreds'] = df[f'Prev_R1_Last3_Lag{lag}'].str[0].astype(int)
    df[f'Prev_R1_Last3_Lag{lag}_Tens'] = df[f'Prev_R1_Last3_Lag{lag}'].str[1].astype(int)
    df[f'Prev_R1_Last3_Lag{lag}_Units'] = df[f'Prev_R1_Last3_Lag{lag}'].str[2].astype(int)

# กำหนด Features และ Target Variables
features = [
    'Year', 'Month', 'Day', 'DayOfWeek', 'DayOfMonth',
    'R1_Sum', 'R1_Even_Count', 'L2_Sum', 'L2_Even_Count', # เพิ่ม L2_Sum, L2_Even_Count
    'Prev_L2_Tens', 'Prev_L2_Units',
    'Prev_R1_Last3_Hundreds', 'Prev_R1_Last3_Tens', 'Prev_R1_Last3_Units',
    'Prev_L2_Lag2_Tens', 'Prev_L2_Lag2_Units', # เพิ่ม Lag 2
    'Prev_R1_Last3_Lag2_Hundreds', 'Prev_R1_Last3_Lag2_Tens', 'Prev_R1_Last3_Lag2_Units',
    'Prev_L2_Lag3_Tens', 'Prev_L2_Lag3_Units', # เพิ่ม Lag 3
    'Prev_R1_Last3_Lag3_Hundreds', 'Prev_R1_Last3_Lag3_Tens', 'Prev_R1_Last3_Lag3_Units'
]

target_cols = [
    'R1_Last3_Hundreds', 'R1_Last3_Tens', 'R1_Last3_Units',
    'L2_Tens', 'L2_Units'
]

# ดรอปแถวที่มีค่า NaN จากการ shift (ตามจำนวน Lag สูงสุด)
df_ml = df.dropna(subset=features + target_cols).copy()

# ตรวจสอบว่า df_ml มีข้อมูลเพียงพอหรือไม่
if df_ml.empty:
    print("! หลังจากเตรียมข้อมูลแล้ว DataFrame ว่างเปล่า ไม่สามารถฝึกโมเดลได้.")
    exit()

print("✅ ข้อมูลถูกเตรียมและสร้าง Features เรียบร้อยแล้ว")
print(f"จำนวนข้อมูลสำหรับการฝึกโมเดล: {len(df_ml)} แถว")

# --- 2. การสร้างและฝึกโมเดล (Train models) พร้อม Hyperparameter Tuning และ TimeSeriesSplit ---
print("\n--- เริ่มต้นการสร้างและฝึกโมเดลด้วย Hyperparameter Tuning และ TimeSeriesSplit ---")

trained_models = {}
label_encoders = {}
scalers = {} # สำหรับ StandardScaler
best_params_found = {} # เก็บ best params ที่หาได้

model_params = {
    'RandomForest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators': [50, 100, 150],
            'max_depth': [None, 5, 10],
            'min_samples_split': [2, 5],
        }
    },
    'XGBoost': { # เปลี่ยนชื่อเป็น GradientBoostingClassifier
        'model': GradientBoostingClassifier(random_state=42),
        'params': {
            'n_estimators': [50, 100, 150],
            'learning_rate': [0.05, 0.1, 0.2],
            'max_depth': [3, 5],
        }
    },
    'MLPClassifier': {
        'model': MLPClassifier(random_state=42, max_iter=500), # เพิ่ม max_iter
        'params': {
            'hidden_layer_sizes': [(50,), (50, 20), (100,)],
            'activation': ['relu', 'tanh'],
            'alpha': [0.0001, 0.001], # L2 penalty (regularization)
        }
    }
}

# ใช้ TimeSeriesSplit เพื่อแบ่งข้อมูล
tscv = TimeSeriesSplit(n_splits=3) # จำนวน splits น้อยๆ ก่อน เนื่องจากข้อมูลน้อย

for target in target_cols:
    print(f"\nกำลังฝึกและปรับจูนโมเดลสำหรับ: {target}")
    trained_models[target] = {}

    Y = df_ml[target]
    X = df_ml[features]

    # Scale Features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    scalers[target] = scaler # เก็บ scaler ไว้ใช้ตอน predict

    le = LabelEncoder()
    le.fit(np.arange(10)) # Fit บนค่าทั้งหมดที่เป็นไปได้ (0-9)
    Y_encoded = le.transform(Y)
    label_encoders[target] = le

    for model_name, config in model_params.items():
        print(f"  - กำลังปรับจูน {model_name}...")
        try:
            grid_search = GridSearchCV(
                config['model'],
                config['params'],
                cv=tscv, # ใช้ TimeSeriesSplit
                scoring='accuracy',
                n_jobs=-1, # ใช้ทุก core CPU
                verbose=0
            )
            grid_search.fit(X_scaled, Y_encoded)

            best_model = grid_search.best_estimator_
            best_params = grid_search.best_params_
            best_score = grid_search.best_score_

            trained_models[target][model_name] = best_model
            best_params_found[f"{target}_{model_name}"] = best_params
            print(f"    ✅ ฝึก {model_name} เสร็จสิ้น. Best Score: {best_score:.4f}, Best Params: {best_params}")

        except Exception as e:
            print(f"    ❌ เกิดข้อผิดพลาดในการฝึก {model_name} สำหรับ {target}: {e}")
            trained_models[target][model_name] = None

print("✅ โมเดลทั้งหมดถูกฝึกและปรับจูนเรียบร้อยแล้ว")

# --- 3. การทำ Backtesting (ใช้ข้อมูล Test Set ที่เหลือ) ---
print("\n--- เริ่มต้นการทำ Backtesting ---")

# แบ่งข้อมูลสำหรับ Train และ Test (ตามลำดับเวลา) สำหรับ Backtesting
train_size_backtest = int(len(df_ml) * 0.8)
train_df_backtest = df_ml.iloc[:train_size_backtest]
test_df_backtest = df_ml.iloc[train_size_backtest:]

backtest_results = []
if not test_df_backtest.empty:
    for index, row in test_df_backtest.iterrows():
        # ต้องใช้ scaler ที่ถูก fit มาจากข้อมูลทั้งหมด หรือ Train set เท่านั้น
        # ในที่นี้ เราใช้ scaler ที่ fit จาก df_ml ทั้งหมด เพื่อให้ predict ได้เลย
        test_features_scaled = scalers[target_cols[0]].transform(pd.DataFrame([row[features]])) # ใช้ scaler ของ target แรก (สมมติว่าเหมือนกัน)

        predicted_r1_last3_digits = []
        predicted_l2_digits = []

        # ทำนายแต่ละหลักสำหรับ Backtesting
        for target_name in target_cols:
            preds_from_all_models = []

            if target_name in trained_models and trained_models[target_name]:
                for model_name, model in trained_models[target_name].items():
                    if model: # ตรวจสอบว่าโมเดลไม่ได้เป็น None
                        try:
                            # ใช้ scaler ที่ถูกต้องสำหรับ target นั้นๆ
                            current_scaler = scalers.get(target_name, scalers[target_cols[0]])
                            scaled_features_for_pred = current_scaler.transform(pd.DataFrame([row[features]]))

                            pred_encoded = model.predict(scaled_features_for_pred)[0]
                            pred_digit = label_encoders[target_name].inverse_transform([pred_encoded])[0]
                            preds_from_all_models.append(pred_digit)
                        except Exception as e:
                            # print(f"คำเตือน: ข้อผิดพลาดในการทำนาย Backtest ด้วย {model_name} สำหรับ {target_name}: {e}")
                            preds_from_all_models.append(0) # ค่า default หากทำนายไม่ได้

            if preds_from_all_models:
                most_common_digit = Counter(preds_from_all_models).most_common(1)[0][0]
            else:
                most_common_digit = 0

            if 'R1_Last3' in target_name:
                predicted_r1_last3_digits.append(str(most_common_digit))
            elif 'L2' in target_name:
                predicted_l2_digits.append(str(most_common_digit))

        predicted_r1_last3 = "".join(predicted_r1_last3_digits)
        predicted_l2 = "".join(predicted_l2_digits)

        actual_r1_last3 = row['เลขท้าย3ตัวรางวัลที่1']
        actual_l2 = row['เลขท้าย2ตัว']

        backtest_results.append({
            'Date': row['Date'],
            'เลขท้าย3ตัวรางวัลที่1_Actual': actual_r1_last3,
            'Predicted_R1_Last3': predicted_r1_last3,
            'R1_Last3_Match': (predicted_r1_last3 == actual_r1_last3),
            'เลขท้าย2ตัว_Actual': actual_l2,
            'Predicted_L2': predicted_l2,
            'L2_Match': (predicted_l2 == actual_l2)
        })

backtest_df = pd.DataFrame(backtest_results)

if backtest_df.empty:
    print("! ไม่มีข้อมูล Test สำหรับ Backtesting หรือข้อมูล Test ว่างเปล่า.")
else:
    print("✅ การทำ Backtesting เสร็จสิ้น")

print("--- เริ่มต้นการประมวลผลคำทำนายและแสดงผลลัพธ์ ---")

# --- 4. กำหนดงวดที่จะทำนาย ---
last_draw_date = df['Date'].max()
print(f"\nงวดล่าสุดในข้อมูลคือ: {last_draw_date.strftime('%d %b %Y (%A)')}")

# คำนวณวันที่ของงวดถัดไป
if last_draw_date.day == 1 or last_draw_date.day == 2:
    next_draw_date = last_draw_date.replace(day=16)
elif last_draw_date.day == 16 or last_draw_date.day == 17:
    if last_draw_date.month == 12:
        next_draw_date = last_draw_date.replace(year=last_draw_date.year + 1, month=1, day=1)
    else:
        next_draw_date = last_draw_date.replace(month=last_draw_date.month + 1, day=1)
else: # If the last draw was on neither the 1st/2nd or 16th/17th, find the next legal draw date
    temp_date = last_draw_date + timedelta(days=1)
    while temp_date.day not in [1, 16]:
        temp_date += timedelta(days=1)
    next_draw_date = temp_date

print(f"งวดที่จะทำนายคือ: {next_draw_date.strftime('%d %b %Y (%A)')}")

# --- 5. สร้างข้อมูล Feature สำหรับงวดที่จะทำนาย (ใช้ข้อมูลล่าสุดและข้อมูลย้อนหลังที่สร้างไว้) ---
last_row_data = df.iloc[-1]
# ใช้ข้อมูลของงวดล่าสุดใน df เพื่อสร้าง features สำหรับงวดถัดไป
next_data_features_input = {col: 0 for col in features} # Initialize with zeros

next_data_features_input['Year'] = next_draw_date.year
next_data_features_input['Month'] = next_draw_date.month
next_data_features_input['Day'] = next_draw_date.day
next_data_features_input['DayOfWeek'] = next_draw_date.dayofweek
next_data_features_input['DayOfMonth'] = next_draw_date.day

# ใช้ค่าเฉลี่ยของ R1_Sum และ R1_Even_Count จากข้อมูลทั้งหมด
next_data_features_input['R1_Sum'] = df['R1_Sum'].mean() if 'R1_Sum' in df and not df['R1_Sum'].empty else 0
next_data_features_input['R1_Even_Count'] = df['R1_Even_Count'].mean() if 'R1_Even_Count' in df and not df['R1_Even_Count'].empty else 0
next_data_features_input['L2_Sum'] = df['L2_Sum'].mean() if 'L2_Sum' in df and not df['L2_Sum'].empty else 0
next_data_features_input['L2_Even_Count'] = df['L2_Even_Count'].mean() if 'L2_Even_Count' in df and not df['L2_Even_Count'].empty else 0


# ดึงค่า Prev_L2 และ Prev_R1_Last3 จากแถวสุดท้ายของ df
# ซึ่งคือเลขที่ออกในงวดล่าสุด (ซึ่งจะเป็น Prev สำหรับงวดถัดไป)
next_data_features_input['Prev_L2_Tens'] = int(last_row_data['เลขท้าย2ตัว'][0])
next_data_features_input['Prev_L2_Units'] = int(last_row_data['เลขท้าย2ตัว'][1])
next_data_features_input['Prev_R1_Last3_Hundreds'] = int(last_row_data['เลขท้าย3ตัวรางวัลที่1'][0])
next_data_features_input['Prev_R1_Last3_Tens'] = int(last_row_data['เลขท้าย3ตัวรางวัลที่1'][1])
next_data_features_input['Prev_R1_Last3_Units'] = int(last_row_data['เลขท้าย3ตัวรางวัลที่1'][2])

# ดึงค่า Lag 2 และ Lag 3 จาก df
# ตรวจสอบให้แน่ใจว่า df มีจำนวนแถวเพียงพอสำหรับดึงค่า lag
if len(df) >= 2:
    prev_2_l2 = df.iloc[-2]['เลขท้าย2ตัว']
    prev_2_r1_last3 = df.iloc[-2]['เลขท้าย3ตัวรางวัลที่1']
    next_data_features_input['Prev_L2_Lag2_Tens'] = int(prev_2_l2[0])
    next_data_features_input['Prev_L2_Lag2_Units'] = int(prev_2_l2[1])
    next_data_features_input['Prev_R1_Last3_Lag2_Hundreds'] = int(prev_2_r1_last3[0])
    next_data_features_input['Prev_R1_Last3_Lag2_Tens'] = int(prev_2_r1_last3[1])
    next_data_features_input['Prev_R1_Last3_Lag2_Units'] = int(prev_2_r1_last3[2])
else: # ถ้าข้อมูลมีน้อยกว่า 2 แถว
    for col in ['Prev_L2_Lag2_Tens', 'Prev_L2_Lag2_Units', 'Prev_R1_Last3_Lag2_Hundreds', 'Prev_R1_Last3_Lag2_Tens', 'Prev_R1_Last3_Lag2_Units']:
        next_data_features_input[col] = 0

if len(df) >= 3:
    prev_3_l2 = df.iloc[-3]['เลขท้าย2ตัว']
    prev_3_r1_last3 = df.iloc[-3]['เลขท้าย3ตัวรางวัลที่1']
    next_data_features_input['Prev_L2_Lag3_Tens'] = int(prev_3_l2[0])
    next_data_features_input['Prev_L2_Lag3_Units'] = int(prev_3_l2[1])
    next_data_features_input['Prev_R1_Last3_Lag3_Hundreds'] = int(prev_3_r1_last3[0])
    next_data_features_input['Prev_R1_Last3_Lag3_Tens'] = int(prev_3_r1_last3[1])
    next_data_features_input['Prev_R1_Last3_Lag3_Units'] = int(prev_3_r1_last3[2])
else: # ถ้าข้อมูลมีน้อยกว่า 3 แถว
    for col in ['Prev_L2_Lag3_Tens', 'Prev_L2_Lag3_Units', 'Prev_R1_Last3_Lag3_Hundreds', 'Prev_R1_Last3_Lag3_Tens', 'Prev_R1_Last3_Lag3_Units']:
        next_data_features_input[col] = 0


next_draw_X_df = pd.DataFrame([next_data_features_input])
# Ensure columns order matches features list
next_draw_X = next_draw_X_df[features]


print("\n--- ข้อมูล Features สำหรับงวดที่จะทำนาย ---")
print(next_draw_X)

# --- 6. ทำนายผลโดยใช้ Ensemble Model (โหวตจากโมเดลทั้งหมด) ---
print("\n--- กำลังทำนายตัวเลขสำหรับงวดถัดไป... ---")

predicted_digits = {}
confidence_scores = {}

for target_name in target_cols:
    preds_from_all_models = []
    prob_distributions = []

    if not (target_name in trained_models and trained_models[target_name]):
        print(f"คำเตือน: ไม่มีโมเดลที่ถูกฝึกสำหรับ {target_name}. จะใช้ค่าเริ่มต้น 0 และความเชื่อมั่น 0.0")
        predicted_digits[target_name] = 0
        confidence_scores[target_name] = 0.0
        continue

    # Scale the input features for prediction
    scaler = scalers.get(target_name, scalers[target_cols[0]]) # ใช้ scaler ที่ถูกต้อง
    next_draw_X_scaled = scaler.transform(next_draw_X)

    for model_name, model in trained_models[target_name].items():
        if model is None:
            continue
        try:
            pred_encoded = model.predict(next_draw_X_scaled)[0]
            pred_digit = label_encoders[target_name].inverse_transform([pred_encoded])[0]
            preds_from_all_models.append(pred_digit)

            if hasattr(model, 'predict_proba'):
                prob_dist_encoded = model.predict_proba(next_draw_X_scaled)[0]
                full_prob_dist = np.zeros(10)
                if hasattr(label_encoders[target_name], 'classes_') and label_encoders[target_name].classes_ is not None:
                    # Map the probabilities back to the original classes (0-9)
                    for i, class_label_encoded in enumerate(label_encoders[target_name].classes_):
                        if i < len(prob_dist_encoded): # Ensure index is within bounds
                            full_prob_dist[class_label_encoded] = prob_dist_encoded[i]
                prob_distributions.append(full_prob_dist)
            else:
                prob_distributions.append(np.full(10, 1/10)) # Default if no proba method

        except Exception as e:
            print(f"คำเตือน: เกิดข้อผิดพลาดในการทำนายด้วยโมเดล {model_name} สำหรับ {target_name}: {e}")
            preds_from_all_models.append(0)
            prob_distributions.append(np.full(10, 1/10))

    if preds_from_all_models:
        most_common_digit = Counter(preds_from_all_models).most_common(1)[0][0]
    else:
        most_common_digit = 0
    predicted_digits[target_name] = most_common_digit

    avg_confidence_list = []
    for prob in prob_distributions:
        if most_common_digit < len(prob):
            avg_confidence_list.append(prob[most_common_digit])
        else:
            avg_confidence_list.append(0.0)

    avg_confidence = np.mean(avg_confidence_list) if avg_confidence_list else 0.0
    confidence_scores[target_name] = avg_confidence

print("\n--- ผลการทำนายแต่ละหลักและระดับความเชื่อมั่น ---")
print(f"Predicted R1_Last3_Hundreds: {predicted_digits.get('R1_Last3_Hundreds', 'N/A')} (Confidence: {confidence_scores.get('R1_Last3_Hundreds', 0.0):.2f})")
print(f"Predicted R1_Last3_Tens:       {predicted_digits.get('R1_Last3_Tens', 'N/A')} (Confidence: {confidence_scores.get('R1_Last3_Tens', 0.0):.2f})")
print(f"Predicted R1_Last3_Units:      {predicted_digits.get('R1_Last3_Units', 'N/A')} (Confidence: {confidence_scores.get('R1_Last3_Units', 0.0):.2f})")
print(f"Predicted L2_Tens:             {predicted_digits.get('L2_Tens', 'N/A')} (Confidence: {confidence_scores.get('L2_Tens', 0.0):.2f})")
print(f"Predicted L2_Units:            {predicted_digits.get('L2_Units', 'N/A')} (Confidence: {confidence_scores.get('L2_Units', 0.0):.2f})")

# --- 7. แสดงผลตัวเลขคาดการณ์ที่ดีที่สุดและชุดเพิ่มเติม ---
print("\n--- ตัวเลขคาดการณ์สำหรับงวดถัดไป (" f"{next_draw_date.strftime('%d %b %Y (%A)')}" ") ---")

predicted_r1_last3 = f"{predicted_digits.get('R1_Last3_Hundreds', '')}{predicted_digits.get('R1_Last3_Tens', '')}{predicted_digits.get('R1_Last3_Units', '')}"
print(f"**เลขท้าย 3 ตัวรางวัลที่ 1 ที่คาดการณ์ (ชุดที่ดีที่สุด): {predicted_r1_last3}**")

predicted_l2 = f"{predicted_digits.get('L2_Tens', '')}{predicted_digits.get('L2_Units', '')}"
print(f"**เลขท้าย 2 ตัวล่างที่คาดการณ์ (ชุดที่ดีที่สุด): {predicted_l2}**")

print("\n--- เลขคาดการณ์เพิ่มเติม (จากการวิเคราะห์และโมเดลรอง) ---")

def get_single_model_prediction_scaled(model_dict, target_name, model_type, next_draw_X, le_dict, scaler_dict):
    if target_name in model_dict and model_type in model_dict[target_name] and model_dict[target_name][model_type] is not None:
        model = model_dict[target_name][model_type]
        le = le_dict.get(target_name)
        scaler = scaler_dict.get(target_name, scaler_dict[target_cols[0]]) # Use appropriate scaler
        try:
            next_draw_X_scaled = scaler.transform(next_draw_X)
            pred_encoded = model.predict(next_draw_X_scaled)[0]
            if le:
                return le.inverse_transform([pred_encoded])[0]
            return pred_encoded
        except Exception as e:
            return ''
    return ''

# Retrieve predictions from individual models (RandomForest, XGBoost, MLP)
# RandomForest Predictions
rf_r1_h_pred_single = get_single_model_prediction_scaled(trained_models, 'R1_Last3_Hundreds', 'RandomForest', next_draw_X, label_encoders, scalers)
rf_r1_t_pred_single = get_single_model_prediction_scaled(trained_models, 'R1_Last3_Tens', 'RandomForest', next_draw_X, label_encoders, scalers)
rf_r1_u_pred_single = get_single_model_prediction_scaled(trained_models, 'R1_Last3_Units', 'RandomForest', next_draw_X, label_encoders, scalers)
rf_l2_t_pred_single = get_single_model_prediction_scaled(trained_models, 'L2_Tens', 'RandomForest', next_draw_X, label_encoders, scalers)
rf_l2_u_pred_single = get_single_model_prediction_scaled(trained_models, 'L2_Units', 'RandomForest', next_draw_X, label_encoders, scalers)
predicted_r1_last3_rf = f"{rf_r1_h_pred_single}{rf_r1_t_pred_single}{rf_r1_u_pred_single}"
predicted_l2_rf = f"{rf_l2_t_pred_single}{rf_l2_u_pred_single}"

if predicted_r1_last3_rf != predicted_r1_last3 and predicted_r1_last3_rf.strip():
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก RandomForest): {predicted_r1_last3_rf}")
else:
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก RandomForest): เหมือนชุดที่ดีที่สุด")

if predicted_l2_rf != predicted_l2 and predicted_l2_rf.strip():
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก RandomForest): {predicted_l2_rf}")
else:
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก RandomForest): เหมือนชุดที่ดีที่สุด")

# XGBoost Predictions
xgb_r1_h_pred_single = get_single_model_prediction_scaled(trained_models, 'R1_Last3_Hundreds', 'XGBoost', next_draw_X, label_encoders, scalers)
xgb_r1_t_pred_single = get_single_model_prediction_scaled(trained_models, 'R1_Last3_Tens', 'XGBoost', next_draw_X, label_encoders, scalers)
xgb_r1_u_pred_single = get_single_model_prediction_scaled(trained_models, 'R1_Last3_Units', 'XGBoost', next_draw_X, label_encoders, scalers)
xgb_l2_t_pred_single = get_single_model_prediction_scaled(trained_models, 'L2_Tens', 'XGBoost', next_draw_X, label_encoders, scalers)
xgb_l2_u_pred_single = get_single_model_prediction_scaled(trained_models, 'L2_Units', 'XGBoost', next_draw_X, label_encoders, scalers)
predicted_r1_last3_xgb = f"{xgb_r1_h_pred_single}{xgb_r1_t_pred_single}{xgb_r1_u_pred_single}"
predicted_l2_xgb = f"{xgb_l2_t_pred_single}{xgb_l2_u_pred_single}"

if predicted_r1_last3_xgb != predicted_r1_last3 and predicted_r1_last3_xgb != predicted_r1_last3_rf and predicted_r1_last3_xgb.strip():
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก XGBoost): {predicted_r1_last3_xgb}")
else:
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก XGBoost): เหมือนชุดที่ดีที่สุด/ชุดที่ 2")

if predicted_l2_xgb != predicted_l2 and predicted_l2_xgb != predicted_l2_rf and predicted_l2_xgb.strip():
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก XGBoost): {predicted_l2_xgb}")
else:
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก XGBoost): เหมือนชุดที่ดีที่สุด/ชุดที่ 2")

# MLPClassifier Predictions
mlp_r1_h_pred_single = get_single_model_prediction_scaled(trained_models, 'R1_Last3_Hundreds', 'MLPClassifier', next_draw_X, label_encoders, scalers)
mlp_r1_t_pred_single = get_single_model_prediction_scaled(trained_models, 'R1_Last3_Tens', 'MLPClassifier', next_draw_X, label_encoders, scalers)
mlp_r1_u_pred_single = get_single_model_prediction_scaled(trained_models, 'R1_Last3_Units', 'MLPClassifier', next_draw_X, label_encoders, scalers)
mlp_l2_t_pred_single = get_single_model_prediction_scaled(trained_models, 'L2_Tens', 'MLPClassifier', next_draw_X, label_encoders, scalers)
mlp_l2_u_pred_single = get_single_model_prediction_scaled(trained_models, 'L2_Units', 'MLPClassifier', next_draw_X, label_encoders, scalers)
predicted_r1_last3_mlp = f"{mlp_r1_h_pred_single}{mlp_r1_t_pred_single}{mlp_r1_u_pred_single}"
predicted_l2_mlp = f"{mlp_l2_t_pred_single}{mlp_l2_u_pred_single}"

if predicted_r1_last3_mlp != predicted_r1_last3 and predicted_r1_last3_mlp != predicted_r1_last3_rf and predicted_r1_last3_mlp != predicted_r1_last3_xgb and predicted_r1_last3_mlp.strip():
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก MLPClassifier): {predicted_r1_last3_mlp}")
else:
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก MLPClassifier): เหมือนชุดอื่นๆ")

if predicted_l2_mlp != predicted_l2 and predicted_l2_mlp != predicted_l2_rf and predicted_l2_mlp != predicted_l2_xgb and predicted_l2_mlp.strip():
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก MLPClassifier): {predicted_l2_mlp}")
else:
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก MLPClassifier): เหมือนชุดอื่นๆ")


# --- 8. แสดงผลการ Backtesting ที่ทำได้ ---
print("\n--- สรุปผลการ Backtesting (ประสิทธิภาพของระบบในการทำนายย้อนหลัง) ---")
if not backtest_df.empty:
    total_r1_last3_matches = backtest_df['R1_Last3_Match'].sum()
    total_l2_matches = backtest_df['L2_Match'].sum()
    total_test_samples = len(backtest_df)

    print(f"จำนวนงวดที่ทำการทดสอบย้อนหลัง: {total_test_samples} งวด")
    if total_test_samples > 0:
        print(f"ระบบทำนาย 'เลขท้าย 3 ตัวรางวัลที่ 1' ถูกต้อง: {total_r1_last3_matches} งวด ({total_r1_last3_matches/total_test_samples:.2%} ความแม่นยำ)")
        print(f"ระบบทำนาย 'เลขท้าย 2 ตัวล่าง' ถูกต้อง: {total_l2_matches} งวด ({total_l2_matches/total_test_samples:.2%} ความแม่นยำ)")
    else:
        print("ไม่มีข้อมูลสำหรับการทดสอบย้อนหลัง.")

    print("\n--- ผลการ Backtesting แยกตามงวด ---")
    print(backtest_df[['Date', 'เลขท้าย3ตัวรางวัลที่1_Actual', 'Predicted_R1_Last3', 'R1_Last3_Match',
                       'เลขท้าย2ตัว_Actual', 'Predicted_L2', 'L2_Match']])
else:
    print("ไม่มีข้อมูล Backtesting ที่พร้อมใช้งาน.")


# --- 9. แสดงผลการทำงานของระบบแบบโปร่งใสเป็นขั้นตอนด้วยภาษาไทย ---
print("\n--- สรุปการทำงานของระบบคาดการณ์หวยไทย (Hybrid Intelligence) ---")
print("ระบบนี้ถูกสร้างขึ้นโดยใช้แนวคิด Hybrid Intelligence ซึ่งผสานรวมหลายวิธีการเพื่อเพิ่มความแม่นยำและน่าเชื่อถือ:")
print("1. **การเตรียมข้อมูล:** ทำความสะอาด, แปลงวันที่, และเติมเลขศูนย์ให้ครบหลัก เพื่อให้ข้อมูลพร้อมใช้งาน")
print("2. **วิศวกรรมฟีเจอร์ (Feature Engineering):** สร้างคุณลักษณะใหม่ๆ จากข้อมูล เช่น เลขแต่ละหลัก, ผลรวม, เลขคู่/คี่, วันที่ออก, รวมถึงเลขของงวดก่อนหน้า (เพิ่ม Lag 2 และ Lag 3) เพื่อให้โมเดลมีข้อมูลที่หลากหลายและซับซ้อนขึ้น")
print("3. **การปรับสเกลข้อมูล (Feature Scaling):** ใช้ StandardScaler เพื่อปรับสเกลคุณลักษณะให้โมเดล AI เรียนรู้ได้ดีขึ้น")
print("4. **การสร้างและฝึกโมเดล AI/ML พร้อมการปรับจูน:**")
print("    - ใช้โมเดล Machine Learning หลากหลายชนิด: **Random Forest**, **Gradient Boosting (XGBoost)**, และ **MLP Classifier** (Neural Network)")
print("    - ทำ **Hyperparameter Tuning** ด้วย **GridSearchCV** เพื่อหาค่าพารามิเตอร์ที่ดีที่สุดสำหรับแต่ละโมเดล ทำให้โมเดลมีประสิทธิภาพสูงสุด")
print("    - ใช้ **TimeSeriesSplit Cross-Validation** เพื่อประเมินประสิทธิภาพโมเดลอย่างถูกต้องสำหรับข้อมูลอนุกรมเวลา")
print("    - แต่ละหลักของเลขท้าย 3 ตัว และเลขท้าย 2 ตัว จะถูกทำนายแยกกัน เพื่อความแม่นยำสูงสุด")
print("5. **การรวมโมเดล (Ensemble Method):** ใช้เทคนิค 'Majority Voting' เพื่อรวมผลการทำนายจากโมเดลแต่ละตัวเข้าด้วยกัน ทำให้ผลลัพธ์มีความเสถียรและแม่นยำยิ่งขึ้น")
print("6. **การทดสอบย้อนหลัง (Backtesting):** ระบบจะแบ่งข้อมูลออกเป็นส่วน Train และ Test เพื่อจำลองการทำนายงวดในอดีตและประเมินประสิทธิภาพของโมเดลอย่างละเอียด รวมถึงแสดงรายงานความแม่นยำ")
print("7. **การประเมินความเชื่อมั่น (Confidence Measures):** แสดงค่าความเชื่อมั่นสำหรับเลขแต่ละหลักที่ทำนายได้ โดยอ้างอิงจากความน่าจะเป็นที่โมเดลคาดการณ์")
print("8. **การคาดการณ์สำหรับงวดถัดไป:** ใช้โมเดลที่ฝึกมาทำนายเลขท้าย 3 ตัวรางวัลที่ 1 และเลขท้าย 2 ตัวล่างสำหรับงวดถัดไป พร้อมนำเสนอชุดตัวเลขคาดการณ์จาก Ensemble และจากโมเดลเดี่ยวๆ หากมีความแตกต่าง")
print("\n**ข้อจำกัดที่สำคัญ:** การทำนายหวยเป็นการคาดการณ์จากข้อมูลในอดีต ซึ่งเป็นการสุ่มที่มีความซับซ้อนสูง แม้ระบบจะใช้แนวทางทางสถิติและ AI/ML ขั้นสูง แต่ก็ไม่สามารถรับประกันผลได้ 100% จุดประสงค์หลักคือการช่วยในการตัดสินใจโดยมีข้อมูลและเหตุผลทางสถิติรองรับ โปรดใช้วิจารณญาณในการใช้งาน")

print("\n--- การประมวลผลคำทำนายและแสดงผลลัพธ์เสร็จสิ้น ---")
```

```python
import numpy as np
import pandas as pd # เพิ่ม import pandas
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from collections import defaultdict, Counter
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from scipy.stats import chi2_contingency, norm, poisson, uniform

# ตรวจสอบว่า df_processed มีอยู่ใน global scope หรือไม่
# หากคุณแน่ใจว่าเซลล์ที่ 1 จะสร้าง df_processed ให้ใช้โค้ดนี้
# แต่ถ้าคุณกำลังทดสอบโค้ดเซลล์นี้แยกต่างหาก และ df_processed ยังไม่มี
# คุณอาจจะต้องสร้าง df_processed แบบจำลองขึ้นมาก่อนเพื่อการทดสอบ
if 'df_processed' not in globals(): # ตรวจสอบใน globals() แทน locals() เพื่อความแน่ใจ
    print("! โปรดรันเซลล์ที่ 1 ก่อน เพื่อเตรียมข้อมูล 'df_processed'.")
    # สร้าง df_processed จำลองขึ้นมาเพื่อไม่ให้โค้ดหยุดทำงานทันทีในกรณีทดสอบ (เฉพาะการทดสอบเท่านั้น)
    # ในการใช้งานจริงใน Colab ต้องรันเซลล์ 1 ก่อนเสมอ
    try:
        # ลองโหลดข้อมูลจำลอง (คุณต้องปรับให้เข้ากับข้อมูลจริงของคุณ)
        # ตัวอย่างการสร้าง df_processed แบบจำลอง (สมมติโครงสร้างข้อมูล)
        data = {
            'Date': pd.to_datetime(['2024-01-01', '2024-01-16', '2024-02-01', '2024-02-16', '2024-03-01', '2024-03-16']),
            'เลขหกหลักรางวัลที่1': ['123456', '789012', '345678', '901234', '567890', '123789'],
            'เลขท้าย3ตัวรางวัลที่1': ['456', '012', '678', '234', '890', '789'],
            'เลขท้าย2ตัว': ['56', '12', '78', '34', '90', '89']
        }
        df_processed = pd.DataFrame(data)
        print("💡 สร้าง 'df_processed' จำลองสำหรับการทดสอบชั่วคราวแล้ว โปรดแน่ใจว่าได้รันเซลล์ที่ 1 ในการใช้งานจริง!")
    except Exception as e:
        print(f"❌ ไม่สามารถสร้าง 'df_processed' จำลองได้: {e}")
        exit() # ยังคง exit หากสร้าง df_processed จำลองไม่ได้
else:
    print("✅ 'df_processed' พบแล้ว ดำเนินการต่อ")


df = df_processed.copy()

print("--- เริ่มต้นการวิเคราะห์และสร้างระบบหลัก ---")

# --- 1. Exploratory Data Analysis (EDA) และ Feature Engineering ---
print("\n--- 1.1 การวิเคราะห์และสร้าง Features (Feature Engineering) ---")

# แยกเลขแต่ละหลักสำหรับรางวัลที่ 1 และเลขท้าย 2 ตัว
for i in range(6):
    df[f'R1_Digit_{i+1}'] = df['เลขหกหลักรางวัลที่1'].str[i].astype(int)
for i in range(2):
    df[f'L2_Digit_{i+1}'] = df['เลขท้าย2ตัว'].str[i].astype(int)

# สร้าง Features จากเลขท้าย 3 ตัว และเลขท้าย 2 ตัว
df['R1_Last3_Hundreds'] = df['เลขท้าย3ตัวรางวัลที่1'].str[0].astype(int)
df['R1_Last3_Tens'] = df['เลขท้าย3ตัวรางวัลที่1'].str[1].astype(int)
df['R1_Last3_Units'] = df['เลขท้าย3ตัวรางวัลที่1'].str[2].astype(int)

df['L2_Tens'] = df['เลขท้าย2ตัว'].str[0].astype(int)
df['L2_Units'] = df['เลขท้าย2ตัว'].str[1].astype(int)

# Features ที่เกี่ยวกับวันและเวลา
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek # Monday=0, Sunday=6
df['DayOfMonth'] = df['Date'].dt.day # ซ้ำกับ 'Day' แต่บางครั้งก็แยกใช้ได้

# Features: ผลรวม, เลขคู่/คี่, เลขซ้ำ (Simple features)
df['R1_Sum'] = df['เลขหกหลักรางวัลที่1'].apply(lambda x: sum(int(d) for d in str(x)))
df['L2_Sum'] = df['เลขท้าย2ตัว'].apply(lambda x: sum(int(d) for d in str(x)))

df['R1_Even_Count'] = df['เลขหกหลักรางวัลที่1'].apply(lambda x: sum(1 for d in str(x) if int(d) % 2 == 0))
df['L2_Even_Count'] = df['เลขท้าย2ตัว'].apply(lambda x: sum(1 for d in str(x) if int(d) % 2 == 0))

# Features เชิงลำดับ (Sequential Patterns) - Lookback Features
# เช่น เลขท้าย 2 ตัวของงวดที่แล้ว
df['Prev_L2'] = df['เลขท้าย2ตัว'].shift(1).fillna('00').astype(str)
df['Prev_R1_Last3'] = df['เลขท้าย3ตัวรางวัลที่1'].shift(1).fillna('000').astype(str)

# Features แยกหลักของ Previous L2 และ R1_Last3
df['Prev_L2_Tens'] = df['Prev_L2'].str[0].astype(int)
df['Prev_L2_Units'] = df['Prev_L2'].str[1].astype(int)
df['Prev_R1_Last3_Hundreds'] = df['Prev_R1_Last3'].str[0].astype(int)
df['Prev_R1_Last3_Tens'] = df['Prev_R1_Last3'].str[1].astype(int)
df['Prev_R1_Last3_Units'] = df['Prev_R1_Last3'].str[2].astype(int)


print("ตัวอย่าง Features ที่สร้างขึ้น (5 แถวแรก):")
print(df[['Date', 'เลขท้าย3ตัวรางวัลที่1', 'เลขท้าย2ตัว', 'Month', 'DayOfWeek', 'R1_Sum', 'Prev_L2', 'Prev_R1_Last3']].head())

# --- 1.2 Exploratory Data Analysis (EDA) ---
print("\n--- 1.2 การวิเคราะห์ข้อมูลเบื้องต้น (EDA) ---")

# การกระจายตัวของเลขแต่ละหลักสำหรับเลขท้าย 3 ตัวรางวัลที่ 1
print("\nการกระจายตัวของเลขท้าย 3 ตัวรางวัลที่ 1:")
for col in ['R1_Last3_Hundreds', 'R1_Last3_Tens', 'R1_Last3_Units']:
    print(f"- {col}:")
    print(df[col].value_counts().sort_index())
    # sns.histplot(df[col], bins=10, kde=True)
    # plt.title(f'Distribution of {col}')
    # plt.show() # คอมเมนต์ plt.show() เพื่อให้ Colab แสดงกราฟอัตโนมัติทีละกราฟเมื่อถึงเวลา

# การกระจายตัวของเลขท้าย 2 ตัว
print("\nการกระจายตัวของเลขท้าย 2 ตัว:")
for col in ['L2_Tens', 'L2_Units']:
    print(f"- {col}:")
    print(df[col].value_counts().sort_index())
    # sns.histplot(df[col], bins=10, kde=True)
    # plt.title(f'Distribution of {col}')
    # plt.show() # คอมเมนต์ plt.show()

# ความสัมพันธ์ของเลขท้าย 2 ตัวกับเลขท้าย 2 ตัวงวดก่อนหน้า
print("\nความสัมพันธ์ระหว่างเลขท้าย 2 ตัวงวดปัจจุบันและงวดก่อนหน้า:")
# ตรวจสอบประเภทข้อมูลก่อนคำนวณ correlation
# ถ้ามีค่าที่ไม่ใช่ตัวเลขจะทำให้เกิด error ได้
# ใช้ .dropna() เพื่อจัดการกับ NaN ที่เกิดจาก .shift(1)
if pd.api.types.is_numeric_dtype(df['เลขท้าย2ตัว']) and pd.api.types.is_numeric_dtype(df['Prev_L2']):
    corr_L2_L2 = df['เลขท้าย2ตัว'].corr(df['Prev_L2'])
    print(f"ค่าสหสัมพันธ์ (Correlation) ระหว่างเลขท้าย 2 ตัวกับเลขท้าย 2 ตัวงวดก่อนหน้า: {corr_L2_L2:.4f}")
else:
    # แปลงเป็น int ก่อน ถ้ายังไม่ได้เป็น
    df['เลขท้าย2ตัว_num'] = pd.to_numeric(df['เลขท้าย2ตัว'], errors='coerce')
    df['Prev_L2_num'] = pd.to_numeric(df['Prev_L2'], errors='coerce')
    corr_L2_L2 = df['เลขท้าย2ตัว_num'].corr(df['Prev_L2_num'])
    print(f"ค่าสหสัมพันธ์ (Correlation) ระหว่างเลขท้าย 2 ตัวกับเลขท้าย 2 ตัวงวดก่อนหน้า: {corr_L2_L2:.4f}")


# ความสัมพันธ์ของเลขท้าย 3 ตัวรางวัลที่ 1 กับเลขท้าย 3 ตัวรางวัลที่ 1 งวดก่อนหน้า
if pd.api.types.is_numeric_dtype(df['เลขท้าย3ตัวรางวัลที่1']) and pd.api.types.is_numeric_dtype(df['Prev_R1_Last3']):
    corr_R1_L3_R1_L3 = df['เลขท้าย3ตัวรางวัลที่1'].corr(df['Prev_R1_Last3'])
    print(f"ค่าสหสัมพันธ์ (Correlation) ระหว่างเลขท้าย 3 ตัวรางวัลที่ 1 กับงวดก่อนหน้า: {corr_R1_L3_R1_L3:.4f}")
else:
    df['เลขท้าย3ตัวรางวัลที่1_num'] = pd.to_numeric(df['เลขท้าย3ตัวรางวัลที่1'], errors='coerce')
    df['Prev_R1_Last3_num'] = pd.to_numeric(df['Prev_R1_Last3'], errors='coerce')
    corr_R1_L3_R1_L3 = df['เลขท้าย3ตัวรางวัลที่1_num'].corr(df['Prev_R1_Last3_num'])
    print(f"ค่าสหสัมพันธ์ (Correlation) ระหว่างเลขท้าย 3 ตัวรางวัลที่ 1 กับงวดก่อนหน้า: {corr_R1_L3_R1_L3:.4f}")


# --- 2. Statistical Tests (การทดสอบทางสถิติ) ---
print("\n--- 2.1 การทดสอบทางสถิติเพื่อหาสูตร/รูปแบบ (Statistical Tests for Patterns) ---")

# 2.1.1 Chi-square test for Uniformity (ทดสอบการกระจายตัวว่าเป็นแบบสุ่มหรือไม่)
print("\n--- Chi-square test (การทดสอบความเป็น Uniformity) ---")
# ทดสอบแต่ละหลักของเลขท้าย 3 ตัว รางวัลที่ 1
for col in ['R1_Last3_Hundreds', 'R1_Last3_Tens', 'R1_Last3_Units']:
    observed_frequencies = df[col].value_counts().sort_index()
    expected_frequencies_val = len(df) / 10 # คาดหวังว่าแต่ละเลขจะออกเท่าๆ กัน
    expected_frequencies = pd.Series([expected_frequencies_val] * 10, index=range(10))

    # จัดการกรณีที่บางเลขอาจไม่ออกเลยในข้อมูล (ทำให้ expected/observed เป็น 0)
    observed = [observed_frequencies.get(i, 0) for i in range(10)]
    expected = [expected_frequencies.get(i, 0) for i in range(10)]

    # Filter out categories with zero expected values for chi-square test
    # (เนื่องจาก chi2_contingency ต้องการ observed และ expected ที่มีมิติเดียวกันและไม่มีค่า 0 ใน expected)
    # หากข้อมูลมีขนาดเล็กมาก และบางเลขไม่เคยปรากฏเลย อาจทำให้เกิดปัญหาได้
    # แนะนำให้ใช้ข้อมูลจำนวนมากสำหรับการทดสอบนี้

    # Check if all observed frequencies are zero or if there's only one unique value
    if len(observed_frequencies) <= 1:
        print(f"สำหรับ {col}: มีข้อมูลไม่เพียงพอหรือมีค่าเดียวสำหรับการทดสอบ Chi-square.")
        continue # ข้ามการทดสอบสำหรับคอลัมน์นี้

    # Ensure expected values are not zero, if they are due to data size,
    # consider grouping or add a small epsilon if necessary (but generally not for discrete counts 0-9)
    # The chi2_contingency requires observed and expected to be 2D arrays.
    # For a goodness-of-fit test, we pass [observed, expected] where observed and expected are 1D arrays
    # representing counts for each category.

    # Filter out categories where expected count is too low (e.g., < 5) to avoid chi-square test issues
    # A common rule of thumb is that expected counts should be at least 5 for each category.
    # For 0-9 digits, if N is total observations, expected for each is N/10.
    # If N/10 < 5, the test might be unreliable.

    # Create observed and expected arrays based on common values in both
    all_indices = sorted(list(set(observed_frequencies.index).union(set(expected_frequencies.index))))
    observed_aligned = [observed_frequencies.get(i, 0) for i in all_indices]
    expected_aligned = [expected_frequencies.get(i, 0) for i in all_indices]

    # Chi-square test requires that the sum of observed equals sum of expected
    # For goodness-of-fit to a theoretical distribution (like uniform),
    # it's usually done with `scipy.stats.chisquare`
    from scipy.stats import chisquare

    # Chi-square test for goodness of fit to a uniform distribution
    # If all expected frequencies are 0 (e.g., len(df) is 0), chisquare will fail.
    if all(e == 0 for e in expected_aligned) or sum(observed_aligned) == 0:
        print(f"สำหรับ {col}: ข้อมูลไม่เพียงพอสำหรับการทดสอบ Chi-square (ผลรวมความถี่เป็นศูนย์).")
        continue

    # chisquare function expects only observed frequencies and optionally f_exp
    # f_exp should sum to the same as f_obs
    total_observations = sum(observed_aligned)
    # Recalculate expected based on the total number of observations for this specific column
    # and the number of categories (10 for 0-9 digits)
    expected_for_chisquare = [total_observations / 10] * 10

    # Filter out categories where observed is 0 and expected is also 0 (or very small)
    # This is more robust

    # Ensure all categories (0-9) are present in `observed_frequencies` for consistency
    full_observed = pd.Series(0, index=range(10))
    full_observed.update(observed_frequencies)

    chi2, p_value = chisquare(f_obs=full_observed, f_exp=expected_for_chisquare)

    print(f"สำหรับ {col}: Chi-square statistic = {chi2:.2f}, P-value = {p_value:.3f}")
    if p_value < 0.05:
        print(f"  -> P-value น้อยกว่า 0.05: **มีนัยสำคัญทางสถิติ** การกระจายตัวของ {col} **ไม่เป็นแบบ Uniform (ไม่สุ่มสมบูรณ์)**")
    else:
        print(f"  -> P-value มากกว่า 0.05: **ไม่มีนัยสำคัญทางสถิติ** การกระจายตัวของ {col} **เป็นแบบ Uniform (ใกล้เคียงสุ่ม)**")

# ทดสอบแต่ละหลักของเลขท้าย 2 ตัวล่าง
for col in ['L2_Tens', 'L2_Units']:
    observed_frequencies = df[col].value_counts().sort_index()
    expected_frequencies_val = len(df) / 10
    expected_frequencies = pd.Series([expected_frequencies_val] * 10, index=range(10))

    if len(observed_frequencies) <= 1:
        print(f"สำหรับ {col}: มีข้อมูลไม่เพียงพอหรือมีค่าเดียวสำหรับการทดสอบ Chi-square.")
        continue

    full_observed = pd.Series(0, index=range(10))
    full_observed.update(observed_frequencies)

    total_observations = sum(full_observed)
    expected_for_chisquare = [total_observations / 10] * 10

    if all(e == 0 for e in expected_for_chisquare) or total_observations == 0:
        print(f"สำหรับ {col}: ข้อมูลไม่เพียงพอสำหรับการทดสอบ Chi-square (ผลรวมความถี่เป็นศูนย์).")
        continue

    chi2, p_value = chisquare(f_obs=full_observed, f_exp=expected_for_chisquare)

    print(f"สำหรับ {col}: Chi-square statistic = {chi2:.2f}, P-value = {p_value:.3f}")
    if p_value < 0.05:
        print(f"  -> P-value น้อยกว่า 0.05: **มีนัยสำคัญทางสถิติ** การกระจายตัวของ {col} **ไม่เป็นแบบ Uniform (ไม่สุ่มสมบูรณ์)**")
    else:
        print(f"  -> P-value มากกว่า 0.05: **ไม่มีนัยสำคัญทางสถิติ** การกระจายตัวของ {col} **เป็นแบบ Uniform (ใกล้เคียงสุ่ม)**")


# 2.1.2 Autocorrelation and Periodicity (ความสัมพันธ์ตัวเองและวงรอบ)
print("\n--- Autocorrelation (ACF) และ Partial Autocorrelation (PACF) ---")
print("หากกราฟมีแท่งที่สูงเกินขีดจำกัด แสดงว่ามี Autocorrelation หรือ Periodicity")

# เลขท้าย 3 ตัวรางวัลที่ 1 (รวมเป็นตัวเลข)
# ใช้เลข 3 หลักเป็น Y
r1_last3_numeric = pd.to_numeric(df['เลขท้าย3ตัวรางวัลที่1'], errors='coerce').dropna().astype(int)
print("\nACF/PACF สำหรับ เลขท้าย 3 ตัวรางวัลที่ 1 (เป็นตัวเลขรวม):")
if len(r1_last3_numeric) > 1: # ต้องมีข้อมูลเพียงพอ
    plt.figure(figsize=(12, 4))
    plt.subplot(121)
    plot_acf(r1_last3_numeric, lags=min(20, len(r1_last3_numeric)//2 -1), ax=plt.gca(), title='Autocorrelation Function (ACF) for R1_Last3') # ปรับ lags
    plt.subplot(122)
    plot_pacf(r1_last3_numeric, lags=min(20, len(r1_last3_numeric)//2 -1), ax=plt.gca(), title='Partial Autocorrelation Function (PACF) for R1_Last3') # ปรับ lags
    plt.tight_layout()
    plt.show() # แสดงกราฟ
else:
    print("  - ข้อมูลไม่เพียงพอสำหรับกราฟ ACF/PACF สำหรับ เลขท้าย 3 ตัวรางวัลที่ 1")


# เลขท้าย 2 ตัวล่าง (รวมเป็นตัวเลข)
l2_numeric = pd.to_numeric(df['เลขท้าย2ตัว'], errors='coerce').dropna().astype(int)
print("\nACF/PACF สำหรับ เลขท้าย 2 ตัวล่าง (เป็นตัวเลขรวม):")
if len(l2_numeric) > 1: # ต้องมีข้อมูลเพียงพอ
    plt.figure(figsize=(12, 4))
    plt.subplot(121)
    plot_acf(l2_numeric, lags=min(20, len(l2_numeric)//2 -1), ax=plt.gca(), title='Autocorrelation Function (ACF) for L2')
    plt.subplot(122)
    plot_pacf(l2_numeric, lags=min(20, len(l2_numeric)//2 -1), ax=plt.gca(), title='Partial Autocorrelation Function (PACF) for L2')
    plt.tight_layout()
    plt.show() # แสดงกราฟ
else:
    print("  - ข้อมูลไม่เพียงพอสำหรับกราฟ ACF/PACF สำหรับ เลขท้าย 2 ตัวล่าง")


# 2.1.3 Probability Distribution Modeling
print("\n--- การจำลองการแจกแจงความน่าจะเป็น (Probability Distribution Modeling) ---")
print("เปรียบเทียบการแจกแจงของข้อมูลจริงกับ Uniform/Poisson Distribution")

# สำหรับแต่ละหลักของเลขท้าย 3 ตัวรางวัลที่ 1
for col in ['R1_Last3_Hundreds', 'R1_Last3_Tens', 'R1_Last3_Units']:
    data_points = df[col]
    # Uniform Distribution (เลข 0-9 ควรจะออกเท่าๆ กันถ้าสุ่มจริง)
    uniform_prob = 1/10
    print(f"\n- การแจกแจงของ {col}:")
    print(f"  - ค่าความน่าจะเป็นที่คาดหวังแบบ Uniform: {uniform_prob:.2f} (สำหรับแต่ละเลข)")
    # สามารถคำนวณและแสดงความน่าจะเป็นจริงของแต่ละเลขได้
    actual_probs = data_points.value_counts(normalize=True).sort_index()
    print("  - ความน่าจะเป็นจริงของแต่ละเลข:")
    print(actual_probs)
    # Visual comparison
    plt.figure(figsize=(8, 4))
    sns.histplot(data_points, bins=10, stat='probability', discrete=True)
    plt.axhline(y=uniform_prob, color='r', linestyle='--', label='Uniform Probability')
    plt.title(f'Probability Distribution of {col}')
    plt.legend()
    plt.show() # แสดงกราฟ


# สำหรับแต่ละหลักของเลขท้าย 2 ตัวล่าง
for col in ['L2_Tens', 'L2_Units']:
    data_points = df[col]
    uniform_prob = 1/10
    print(f"\n- การแจกแจงของ {col}:")
    print(f"  - ค่าความน่าจะเป็นที่คาดหวังแบบ Uniform: {uniform_prob:.2f} (สำหรับแต่ละเลข)")
    actual_probs = data_points.value_counts(normalize=True).sort_index()
    print("  - ความน่าจะเป็นจริงของแต่ละเลข:")
    print(actual_probs)
    # Visual comparison
    plt.figure(figsize=(8, 4))
    sns.histplot(data_points, bins=10, stat='probability', discrete=True)
    plt.axhline(y=uniform_prob, color='r', linestyle='--', label='Uniform Probability')
    plt.title(f'Probability Distribution of {col}')
    plt.legend()
    plt.show() # แสดงกราฟ


# --- 3. Multiple ML Models (การสร้างโมเดล Machine Learning) ---
print("\n--- 3.1 การสร้างและฝึกโมเดล Machine Learning ---")

# เตรียมข้อมูลสำหรับ ML Models
# เราจะพยายามทำนายแต่ละหลักแยกกัน หรือทำนายตัวเลขรวม แล้วค่อยแยกหลัก
# สำหรับเลขท้าย 3 ตัวรางวัลที่ 1 และเลขท้าย 2 ตัวล่าง
# Features ที่จะใช้
features = ['Year', 'Month', 'Day', 'DayOfWeek', 'DayOfMonth', 'R1_Sum', 'R1_Even_Count',
            'Prev_L2_Tens', 'Prev_L2_Units',
            'Prev_R1_Last3_Hundreds', 'Prev_R1_Last3_Tens', 'Prev_R1_Last3_Units']

# สร้าง X (Features) และ Y (Target)
# ตรวจสอบและจัดการกับค่า NaN ที่อาจเกิดจาก .shift(1) หรือการแปลงข้อมูล
# ใช้ .iloc[1:] เพื่อข้ามแถวแรกที่มี NaN จากการ shift
df_ml = df.dropna(subset=features + ['R1_Last3_Hundreds', 'R1_Last3_Tens', 'R1_Last3_Units',
                                     'L2_Tens', 'L2_Units']).copy()

# สร้าง X และ Y
X = df_ml[features]
Y_r1_last3_hundreds = df_ml['R1_Last3_Hundreds']
Y_r1_last3_tens = df_ml['R1_Last3_Tens']
Y_r1_last3_units = df_ml['R1_Last3_Units']
Y_l2_tens = df_ml['L2_Tens']
Y_l2_units = df_ml['L2_Units']

# ตรวจสอบว่ามีข้อมูลเพียงพอหรือไม่
if len(X) == 0:
    print("! ไม่มีข้อมูลที่เพียงพอสำหรับฝึกโมเดลหลังจากจัดการค่าว่าง โปรดตรวจสอบข้อมูลเริ่มต้น.")
    exit()

# แบ่งข้อมูลสำหรับ Cross-validation และ Backtesting
# ใช้ Time Series Split: 80% Train, 20% Test
train_size = int(len(X) * 0.8)
if train_size == 0: # ป้องกันกรณีข้อมูลน้อยเกินไป
    print("! ข้อมูลน้อยเกินไปสำหรับแบ่ง Train/Test.")
    exit()

X_train, X_test = X[:train_size], X[train_size:]
Y_r1_last3_hundreds_train, Y_r1_last3_hundreds_test = Y_r1_last3_hundreds[:train_size], Y_r1_last3_hundreds[train_size:]
Y_r1_last3_tens_train, Y_r1_last3_tens_test = Y_r1_last3_tens[:train_size], Y_r1_last3_tens[train_size:]
Y_r1_last3_units_train, Y_r1_last3_units_test = Y_r1_last3_units[:train_size], Y_r1_last3_units[train_size:]
Y_l2_tens_train, Y_l2_tens_test = Y_l2_tens[:train_size], Y_l2_tens[train_size:]
Y_l2_units_train, Y_l2_units_test = Y_l2_units[:train_size], Y_l2_units[train_size:]


print(f"จำนวนข้อมูลสำหรับ Train: {len(X_train)} แถว")
print(f"จำนวนข้อมูลสำหรับ Test: {len(X_test)} แถว")

# --- Function สำหรับฝึกและประเมินโมเดล ---
def train_and_evaluate_model(model_instance, X_train_data, Y_train_data, X_test_data, Y_test_data, target_name):
    print(f"\n--- กำลังฝึกโมเดล {model_instance.__class__.__name__} สำหรับ {target_name} ---")

    # ตรวจสอบว่าชุดข้อมูลทดสอบมีข้อมูลอยู่หรือไม่
    if len(X_test_data) == 0:
        print(f"ข้อมูลสำหรับ Test ของ {target_name} เป็นศูนย์ ไม่สามารถประเมินผลได้.")
        return model_instance, 0.0, np.array([])

    model_instance.fit(X_train_data, Y_train_data)
    y_pred = model_instance.predict(X_test_data)
    accuracy = accuracy_score(Y_test_data, y_pred)
    print(f"Accuracy ของ {target_name}: {accuracy:.4f}")
    # print(f"Classification Report สำหรับ {target_name}:\n{classification_report(Y_test_data, y_pred, zero_division=0)}")
    return model_instance, accuracy, y_pred

# --- โมเดล Machine Learning ---
models_configs = {
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': GradientBoostingClassifier(n_estimators=100, random_state=42), # ใช้ GradientBoostingClassifier แทน XGBoost สำหรับความเข้ากันได้ง่าย
    'MLPClassifier': MLPClassifier(hidden_layer_sizes=(50, 20), max_iter=500, random_state=42)
}

# เก็บผลลัพธ์
model_results = {}
predictions = {}

# ฝึกโมเดลสำหรับ เลขท้าย 3 ตัว รางวัลที่ 1
for name, model_instance in models_configs.items():
    # ต้องสร้าง instance ใหม่ของโมเดลสำหรับแต่ละ target เพื่อไม่ให้ทับกัน
    current_model = type(model_instance)(**model_instance.get_params())
    model_r1_h, acc_r1_h, pred_r1_h = train_and_evaluate_model(current_model, X_train, Y_r1_last3_hundreds_train, X_test, Y_r1_last3_hundreds_test, f'R1_Last3_Hundreds ({name})')
    model_results[f'R1_Last3_Hundreds_{name}'] = acc_r1_h
    predictions[f'R1_Last3_Hundreds_{name}'] = pred_r1_h

    current_model = type(model_instance)(**model_instance.get_params())
    model_r1_t, acc_r1_t, pred_r1_t = train_and_evaluate_model(current_model, X_train, Y_r1_last3_tens_train, X_test, Y_r1_last3_tens_test, f'R1_Last3_Tens ({name})')
    model_results[f'R1_Last3_Tens_{name}'] = acc_r1_t
    predictions[f'R1_Last3_Tens_{name}'] = pred_r1_t

    current_model = type(model_instance)(**model_instance.get_params())
    model_r1_u, acc_r1_u, pred_r1_u = train_and_evaluate_model(current_model, X_train, Y_r1_last3_units_train, X_test, Y_r1_last3_units_test, f'R1_Last3_Units ({name})')
    model_results[f'R1_Last3_Units_{name}'] = acc_r1_u
    predictions[f'R1_Last3_Units_{name}'] = pred_r1_u

# ฝึกโมเดลสำหรับ เลขท้าย 2 ตัวล่าง
for name, model_instance in models_configs.items():
    current_model = type(model_instance)(**model_instance.get_params())
    model_l2_t, acc_l2_t, pred_l2_t = train_and_evaluate_model(current_model, X_train, Y_l2_tens_train, X_test, Y_l2_tens_test, f'L2_Tens ({name})')
    model_results[f'L2_Tens_{name}'] = acc_l2_t
    predictions[f'L2_Tens_{name}'] = pred_l2_t

    current_model = type(model_instance)(**model_instance.get_params())
    model_l2_u, acc_l2_u, pred_l2_u = train_and_evaluate_model(current_model, X_train, Y_l2_units_train, X_test, Y_l2_units_test, f'L2_Units ({name})')
    model_results[f'L2_Units_{name}'] = acc_l2_u
    predictions[f'L2_Units_{name}'] = pred_l2_u

# --- 4. Ensemble Method (วิธีการรวมโมเดล) ---
print("\n--- 3.2 การรวมโมเดล (Ensemble Method) ---")

# เนื่องจากทำนายแต่ละหลักแยกกัน การทำ Ensemble จะทำโดยการโหวตจากโมเดลแต่ละหลัก
# เช่น R1_Last3_Hundreds_Ensemble = โหวตจาก RandomForest_R1_Last3_Hundreds, XGBoost_R1_Last3_Hundreds, MLP_R1_Last3_Hundreds

def ensemble_predict(predictions_dict, prefix, test_size_val): # เปลี่ยนชื่อ test_size เพื่อไม่ให้ชนกับ built-in
    all_preds = []
    for model_name in models_configs.keys(): # ใช้ models_configs.keys() แทน models.keys()
        key = f'{prefix}_{model_name}'
        if key in predictions_dict:
            all_preds.append(predictions_dict[key])
        else:
            # Handle cases where a prediction might be missing
            print(f"คำเตือน: ไม่พบการทำนายสำหรับ {key}")
            # อาจจะใส่ array ว่าง หรือ array ที่มีค่า 0 แทน ขึ้นอยู่กับกลยุทธ์
            all_preds.append(np.zeros(test_size_val)) # ใส่ค่า 0 ถ้าไม่มีข้อมูล

    if not all_preds or all(len(p) == 0 for p in all_preds):
        return np.array([]) # กรณีไม่มีการทำนาย

    # Transpose เพื่อให้แต่ละแถวคือการทำนายของแต่ละโมเดลสำหรับงวดหนึ่งๆ
    ensemble_preds = []
    for i in range(test_size_val):
        daily_preds = [preds[i] for preds in all_preds if i < len(preds)] # ตรวจสอบขอบเขต
        if daily_preds:
            # โหวตหาค่าที่ออกบ่อยที่สุด (mode)
            most_common = Counter(daily_preds).most_common(1)[0][0]
            ensemble_preds.append(most_common)
        else:
            ensemble_preds.append(0) # หรือค่าเริ่มต้นอื่นๆ หากไม่มีการทำนาย
    return np.array(ensemble_preds)

print("\n--- ผลการทำนายของ Ensemble Model ---")

# ตรวจสอบขนาดของชุดทดสอบจริงก่อนส่งให้ ensemble_predict
test_sample_size = len(Y_r1_last3_hundreds_test) # ใช้ขนาดจากชุดทดสอบจริง

# Ensemble สำหรับเลขท้าย 3 ตัว รางวัลที่ 1
ensemble_r1_h_pred = ensemble_predict(predictions, 'R1_Last3_Hundreds', test_sample_size)
ensemble_r1_t_pred = ensemble_predict(predictions, 'R1_Last3_Tens', test_sample_size)
ensemble_r1_u_pred = ensemble_predict(predictions, 'R1_Last3_Units', test_sample_size)

# Ensemble สำหรับเลขท้าย 2 ตัวล่าง
ensemble_l2_t_pred = ensemble_predict(predictions, 'L2_Tens', test_sample_size)
ensemble_l2_u_pred = ensemble_predict(predictions, 'L2_Units', test_sample_size)

# ประเมินผล Ensemble
if test_sample_size > 0:
    acc_ensemble_r1_h = accuracy_score(Y_r1_last3_hundreds_test, ensemble_r1_h_pred)
    acc_ensemble_r1_t = accuracy_score(Y_r1_last3_tens_test, ensemble_r1_t_pred)
    acc_ensemble_r1_u = accuracy_score(Y_r1_last3_units_test, ensemble_r1_u_pred)

    acc_ensemble_l2_t = accuracy_score(Y_l2_tens_test, ensemble_l2_t_pred)
    acc_ensemble_l2_u = accuracy_score(Y_l2_units_test, ensemble_l2_u_pred)

    print(f"Accuracy ของ Ensemble สำหรับ R1_Last3_Hundreds: {acc_ensemble_r1_h:.4f}")
    print(f"Accuracy ของ Ensemble สำหรับ R1_Last3_Tens: {acc_ensemble_r1_t:.4f}")
    print(f"Accuracy ของ Ensemble สำหรับ R1_Last3_Units: {acc_ensemble_r1_u:.4f}")
    print(f"Accuracy ของ Ensemble สำหรับ L2_Tens: {acc_ensemble_l2_t:.4f}")
    print(f"Accuracy ของ Ensemble สำหรับ L2_Units: {acc_ensemble_l2_u:.4f}")
else:
    print("ไม่มีข้อมูลสำหรับทดสอบ (Test set) จึงไม่สามารถประเมินผล Ensemble ได้.")

# --- 5. Backtesting (การทดสอบย้อนหลัง) ---
print("\n--- 4. การทดสอบย้อนหลัง (Backtesting) ---")

# สร้าง DataFrame สำหรับผลการทดสอบ
# ต้องแน่ใจว่า index ของ backtest_df ตรงกับ Y_test
backtest_df = df_ml.iloc[train_size:].copy()
if len(backtest_df) > 0:
    backtest_df['Predicted_R1_Last3_Hundreds'] = ensemble_r1_h_pred
    backtest_df['Predicted_R1_Last3_Tens'] = ensemble_r1_t_pred
    backtest_df['Predicted_R1_Last3_Units'] = ensemble_r1_u_pred
    backtest_df['Predicted_L2_Tens'] = ensemble_l2_t_pred
    backtest_df['Predicted_L2_Units'] = ensemble_l2_u_pred

    # ตรวจสอบและแปลงชนิดข้อมูลของคอลัมน์ที่เป็นการทำนายให้เป็น string ก่อน apply
    for col in ['Predicted_R1_Last3_Hundreds', 'Predicted_R1_Last3_Tens', 'Predicted_R1_Last3_Units',
                'Predicted_L2_Tens', 'Predicted_L2_Units']:
        backtest_df[col] = backtest_df[col].astype(str)

    backtest_df['Predicted_R1_Last3'] = backtest_df.apply(lambda row: f"{row['Predicted_R1_Last3_Hundreds']}{row['Predicted_R1_Last3_Tens']}{row['Predicted_R1_Last3_Units']}", axis=1)
    backtest_df['Predicted_L2'] = backtest_df.apply(lambda row: f"{row['Predicted_L2_Tens']}{row['Predicted_L2_Units']}", axis=1)

    # ตรวจสอบความถูกต้องของการทำนายเลขท้าย 3 ตัว
    backtest_df['R1_Last3_Match'] = (backtest_df['Predicted_R1_Last3'] == backtest_df['เลขท้าย3ตัวรางวัลที่1'])
    # ตรวจสอบความถูกต้องของการทำนายเลขท้าย 2 ตัว
    backtest_df['L2_Match'] = (backtest_df['Predicted_L2'] == backtest_df['เลขท้าย2ตัว'])

    print("\n--- สรุปผล Backtesting (5 แถวสุดท้ายของการทดสอบ) ---")
    print(backtest_df[['Date', 'เลขท้าย3ตัวรางวัลที่1', 'Predicted_R1_Last3', 'R1_Last3_Match',
                       'เลขท้าย2ตัว', 'Predicted_L2', 'L2_Match']].tail())

    total_r1_last3_matches = backtest_df['R1_Last3_Match'].sum()
    total_l2_matches = backtest_df['L2_Match'].sum()
    total_test_samples = len(backtest_df)

    print(f"\nจำนวนงวดที่ทำการทดสอบย้อนหลัง: {total_test_samples} งวด")
    print(f"จำนวนงวดที่ทายเลขท้าย 3 ตัวรางวัลที่ 1 ถูก: {total_r1_last3_matches} งวด ({total_r1_last3_matches/total_test_samples:.2%} ความแม่นยำ)")
    print(f"จำนวนงวดที่ทายเลขท้าย 2 ตัวล่าง ถูก: {total_l2_matches} งวด ({total_l2_matches/total_test_samples:.2%} ความแม่นยำ)")
else:
    print("ไม่สามารถทำการ Backtesting ได้ เนื่องจากไม่มีข้อมูลในชุดทดสอบ.")

# --- 6. Risk Assessment และ Confidence Intervals (การประเมินความเสี่ยงและความเชื่อมั่น) ---
print("\n--- 5. การประเมินความเสี่ยงและความเชื่อมั่น ---")
print("การทำนายหวยเป็นการยากที่จะให้ Confidence Intervals ที่แม่นยำเหมือนโมเดลทั่วไป เนื่องจากเป็นการสุ่ม")
print("อย่างไรก็ตาม เราสามารถใช้ความน่าจะเป็นที่โมเดลคาดการณ์สำหรับแต่ละหลักมาประมาณได้")

# ตัวอย่างการคำนวณความน่าจะเป็นของแต่ละตัวเลข (จาก Train set)
# โดยใช้โมเดล RandomForestClassifier
# เนื่องจาก models_configs เก็บแค่ configuration, ต้องสร้าง instance ใหม่และ train ใหม่
rf_r1_h = RandomForestClassifier(n_estimators=100, random_state=42)
rf_r1_h.fit(X_train, Y_r1_last3_hundreds_train)

# สมมติเราต้องการทำนายงวดถัดไป
# สร้างข้อมูลสำหรับทำนายงวดถัดไป (ปัจจุบันคือ 1 มิ.ย. 68, ดังนั้นงวดถัดไปคือ 16 มิ.ย. 68)
# เราจะต้องใช้ข้อมูลของงวดล่าสุด (1 มิ.ย. 68) เป็น 'Prev_L2', 'Prev_R1_Last3'
if not df.empty:
    last_row = df.iloc[-1]
    # ตรวจสอบและจัดการกับวันที่ในอนาคต (ถ้าจำเป็น)
    next_date_features = {
        'Year': 2025, # ปี 68 -> 2025
        'Month': 6, # มิ.ย.
        'Day': 16, # วันที่ 16
        'DayOfWeek': pd.to_datetime('2025-06-16').dayofweek, # คำนวณ DayOfWeek จริงๆ
        'DayOfMonth': 16,
        'R1_Sum': df['R1_Sum'].mean(), # ใช้ค่าเฉลี่ยหรือค่ากลาง (อาจต้องปรับปรุง)
        'R1_Even_Count': df['R1_Even_Count'].mean(), # ใช้ค่าเฉลี่ย (อาจต้องปรับปรุง)
        'Prev_L2_Tens': int(str(last_row['เลขท้าย2ตัว'])[0]) if pd.notna(last_row['เลขท้าย2ตัว']) and len(str(last_row['เลขท้าย2ตัว'])) > 0 else 0,
        'Prev_L2_Units': int(str(last_row['เลขท้าย2ตัว'])[1]) if pd.notna(last_row['เลขท้าย2ตัว']) and len(str(last_row['เลขท้าย2ตัว'])) > 1 else 0,
        'Prev_R1_Last3_Hundreds': int(str(last_row['เลขท้าย3ตัวรางวัลที่1'])[0]) if pd.notna(last_row['เลขท้าย3ตัวรางวัลที่1']) and len(str(last_row['เลขท้าย3ตัวรางวัลที่1'])) > 0 else 0,
        'Prev_R1_Last3_Tens': int(str(last_row['เลขท้าย3ตัวรางวัลที่1'])[1]) if pd.notna(last_row['เลขท้าย3ตัวรางวัลที่1']) and len(str(last_row['เลขท้าย3ตัวรางวัลที่1'])) > 1 else 0,
        'Prev_R1_Last3_Units': int(str(last_row['เลขท้าย3ตัวรางวัลที่1'])[2]) if pd.notna(last_row['เลขท้าย3ตัวรางวัลที่1']) and len(str(last_row['เลขท้าย3ตัวรางวัลที่1'])) > 2 else 0
    }

    # สร้าง DataFrame สำหรับข้อมูลงวดถัดไป
    next_data_point = pd.DataFrame([next_date_features])
    next_data_point_X = next_data_point[features]

    print("\n--- ความน่าจะเป็นในการทำนายเลขแต่ละหลักของงวดถัดไป (จาก RandomForest) ---")

    # Predict_proba สำหรับ เลขท้าย 3 ตัว รางวัลที่ 1
    # ตรวจสอบว่าโมเดลถูกฝึกมาแล้วหรือไม่
    if hasattr(rf_r1_h, 'classes_'): # ตรวจสอบว่าโมเดลได้รับการ fit แล้ว
        prob_r1_h = rf_r1_h.predict_proba(next_data_point_X)[0]
        print(f"ความน่าจะเป็นของ R1_Last3_Hundreds: {np.round(prob_r1_h, 3)}")
        # ตัวอย่าง: เลข 0 มีโอกาสออก prob_r1_h[0]*100%
    else:
        print("โมเดล RandomForest สำหรับ R1_Last3_Hundreds ยังไม่ได้ถูกฝึก.")
else:
    print("ไม่สามารถทำนายงวดถัดไปได้ เนื่องจาก df ว่างเปล่า.")


print("\n--- การวิเคราะห์และสร้างระบบหลักเสร็จสิ้น ---")

# เก็บโมเดลและข้อมูลที่จำเป็นสำหรับเซลล์ถัดไป
trained_models = {}
for target_col in ['R1_Last3_Hundreds', 'R1_Last3_Tens', 'R1_Last3_Units', 'L2_Tens', 'L2_Units']:
    trained_models[target_col] = {}
    Y_target_train = df_ml[target_col].iloc[:train_size] # ใช้ df_ml
    for name, model_instance in models_configs.items():
        current_model = type(model_instance)(**model_instance.get_params())
        trained_models[target_col][name] = current_model.fit(X_train, Y_target_train)

# ตรวจสอบว่า backtest_df ถูกสร้างขึ้นหรือไม่ ก่อนที่จะเก็บใน global_prediction_data
if 'backtest_df' in locals():
    global_prediction_data = {
        'df': df, # เก็บ df_original ไว้
        'df_ml': df_ml, # เก็บ df_ml ที่ใช้ train/test ไว้ด้วย
        'features': features,
        'trained_models': trained_models,
        'backtest_df': backtest_df # เพื่อให้เซลล์ 3 สามารถแสดงผล Backtest ได้อีกครั้ง
    }
else:
    global_prediction_data = {
        'df': df,
        'df_ml': df_ml,
        'features': features,
        'trained_models': trained_models,
        'backtest_df': None # หรือจัดการตามความเหมาะสม
    }
```

```python
import pandas as pd
import numpy as np
from collections import Counter
from datetime import datetime, timedelta

# ตรวจสอบว่า global_prediction_data มีอยู่ใน global scope หรือไม่
# ใช้ globals() เพื่อตรวจสอบใน global scope
if 'global_prediction_data' not in globals() or global_prediction_data is None:
    print("! โปรดรันเซลล์ที่ 1 และ 2 ก่อน เพื่อให้มีข้อมูลสำหรับทำนายและแสดงผลลัพธ์.")
    # หากไม่มีข้อมูล ให้สร้างข้อมูลจำลองเพื่อให้โค้ดรันต่อได้ (สำหรับการทดสอบเท่านั้น)
    print("💡 กำลังสร้าง 'global_prediction_data' จำลองสำหรับการทดสอบชั่วคราว...")
    # สร้าง df_processed จำลอง
    data = {
        'Date': pd.to_datetime(['2024-01-01', '2024-01-16', '2024-02-01', '2024-02-16', '2024-03-01', '2024-03-16', '2024-04-01', '2024-04-16', '2024-05-02', '2024-05-16']),
        'เลขหกหลักรางวัลที่1': ['123456', '789012', '345678', '901234', '567890', '123789', '987654', '321098', '654321', '098765'],
        'เลขท้าย3ตัวรางวัลที่1': ['456', '012', '678', '234', '890', '789', '654', '098', '321', '765'],
        'เลขท้าย2ตัว': ['56', '12', '78', '34', '90', '89', '54', '98', '21', '65']
    }
    df_mock = pd.DataFrame(data)

    # จำลองการสร้าง features (จากเซลล์ 2)
    for i in range(6):
        df_mock[f'R1_Digit_{i+1}'] = df_mock['เลขหกหลักรางวัลที่1'].str[i].astype(int)
    for i in range(2):
        df_mock[f'L2_Digit_{i+1}'] = df_mock['เลขท้าย2ตัว'].str[i].astype(int)
    df_mock['R1_Last3_Hundreds'] = df_mock['เลขท้าย3ตัวรางวัลที่1'].str[0].astype(int)
    df_mock['R1_Last3_Tens'] = df_mock['เลขท้าย3ตัวรางวัลที่1'].str[1].astype(int)
    df_mock['R1_Last3_Units'] = df_mock['เลขท้าย3ตัวรางวัลที่1'].str[2].astype(int)
    df_mock['L2_Tens'] = df_mock['เลขท้าย2ตัว'].str[0].astype(int)
    df_mock['L2_Units'] = df_mock['เลขท้าย2ตัว'].str[1].astype(int)
    df_mock['Year'] = df_mock['Date'].dt.year
    df_mock['Month'] = df_mock['Date'].dt.month
    df_mock['Day'] = df_mock['Date'].dt.day
    df_mock['DayOfWeek'] = df_mock['Date'].dt.dayofweek
    df_mock['DayOfMonth'] = df_mock['Date'].dt.day
    df_mock['R1_Sum'] = df_mock['เลขหกหลักรางวัลที่1'].apply(lambda x: sum(int(d) for d in str(x)))
    df_mock['L2_Sum'] = df_mock['เลขท้าย2ตัว'].apply(lambda x: sum(int(d) for d in str(x)))
    df_mock['R1_Even_Count'] = df_mock['เลขหกหลักรางวัลที่1'].apply(lambda x: sum(1 for d in str(x) if int(d) % 2 == 0))
    df_mock['L2_Even_Count'] = df_mock['เลขท้าย2ตัว'].apply(lambda x: sum(1 for d in str(x) if int(d) % 2 == 0))
    df_mock['Prev_L2'] = df_mock['เลขท้าย2ตัว'].shift(1).fillna('00').astype(str)
    df_mock['Prev_R1_Last3'] = df_mock['เลขท้าย3ตัวรางวัลที่1'].shift(1).fillna('000').astype(str)
    df_mock['Prev_L2_Tens'] = df_mock['Prev_L2'].str[0].astype(int)
    df_mock['Prev_L2_Units'] = df_mock['Prev_L2'].str[1].astype(int)
    df_mock['Prev_R1_Last3_Hundreds'] = df_mock['Prev_R1_Last3'].str[0].astype(int)
    df_mock['Prev_R1_Last3_Tens'] = df_mock['Prev_R1_Last3'].str[1].astype(int)
    df_mock['Prev_R1_Last3_Units'] = df_mock['Prev_R1_Last3'].str[2].astype(int)

    # จำลอง trained_models และ backtest_df (เพื่อให้โค้ดรันได้แม้ไม่มีการฝึกจริง)
    from sklearn.ensemble import RandomForestClassifier # ต้อง import โมเดล
    from sklearn.neural_network import MLPClassifier
    from sklearn.ensemble import GradientBoostingClassifier

    features_mock = ['Year', 'Month', 'Day', 'DayOfWeek', 'DayOfMonth', 'R1_Sum', 'R1_Even_Count',
                     'Prev_L2_Tens', 'Prev_L2_Units',
                     'Prev_R1_Last3_Hundreds', 'Prev_R1_Last3_Tens', 'Prev_R1_Last3_Units']

    # สร้าง X, Y จาก df_mock และ dropna
    df_ml_mock = df_mock.dropna(subset=features_mock + ['R1_Last3_Hundreds', 'R1_Last3_Tens', 'R1_Last3_Units',
                                                         'L2_Tens', 'L2_Units']).copy()

    if len(df_ml_mock) > 1: # ต้องมีข้อมูลเพียงพอสำหรับการแบ่ง train/test
        train_size_mock = int(len(df_ml_mock) * 0.8)
        X_train_mock, Y_train_r1_h_mock = df_ml_mock[features_mock][:train_size_mock], df_ml_mock['R1_Last3_Hundreds'][:train_size_mock]
        Y_train_r1_t_mock = df_ml_mock['R1_Last3_Tens'][:train_size_mock]
        Y_train_r1_u_mock = df_ml_mock['R1_Last3_Units'][:train_size_mock]
        Y_train_l2_t_mock = df_ml_mock['L2_Tens'][:train_size_mock]
        Y_train_l2_u_mock = df_ml_mock['L2_Units'][:train_size_mock]

        mock_models = {
            'RandomForest': RandomForestClassifier(n_estimators=10, random_state=42), # n_estimators น้อยลงเพื่อความเร็ว
            'XGBoost': GradientBoostingClassifier(n_estimators=10, random_state=42),
            'MLPClassifier': MLPClassifier(hidden_layer_sizes=(10,), max_iter=50, random_state=42) # max_iter และ hidden_layer_sizes น้อยลง
        }

        trained_models_mock = {
            'R1_Last3_Hundreds': {name: type(m)(**m.get_params()).fit(X_train_mock, Y_train_r1_h_mock) for name, m in mock_models.items()},
            'R1_Last3_Tens': {name: type(m)(**m.get_params()).fit(X_train_mock, Y_train_r1_t_mock) for name, m in mock_models.items()},
            'R1_Last3_Units': {name: type(m)(**m.get_params()).fit(X_train_mock, Y_train_r1_u_mock) for name, m in mock_models.items()},
            'L2_Tens': {name: type(m)(**m.get_params()).fit(X_train_mock, Y_train_l2_t_mock) for name, m in mock_models.items()},
            'L2_Units': {name: type(m)(**m.get_params()).fit(X_train_mock, Y_train_l2_u_mock) for name, m in mock_models.items()}
        }

        # สร้าง backtest_df จำลอง
        X_test_mock = df_ml_mock[features_mock][train_size_mock:]
        Y_test_r1_h_mock = df_ml_mock['R1_Last3_Hundreds'][train_size_mock:]
        Y_test_l2_t_mock = df_ml_mock['L2_Tens'][train_size_mock:]

        backtest_df_mock = df_ml_mock.iloc[train_size_mock:].copy()
        if not X_test_mock.empty:
            # ทำนายด้วยโมเดลสุ่มๆ เพื่อให้มีค่า
            backtest_df_mock['Predicted_R1_Last3_Hundreds'] = np.random.randint(0, 10, size=len(X_test_mock))
            backtest_df_mock['Predicted_R1_Last3_Tens'] = np.random.randint(0, 10, size=len(X_test_mock))
            backtest_df_mock['Predicted_R1_Last3_Units'] = np.random.randint(0, 10, size=len(X_test_mock))
            backtest_df_mock['Predicted_L2_Tens'] = np.random.randint(0, 10, size=len(X_test_mock))
            backtest_df_mock['Predicted_L2_Units'] = np.random.randint(0, 10, size=len(X_test_mock))

            # ตรวจสอบและแปลงชนิดข้อมูลของคอลัมน์ที่เป็นการทำนายให้เป็น string ก่อน apply
            for col in ['Predicted_R1_Last3_Hundreds', 'Predicted_R1_Last3_Tens', 'Predicted_R1_Last3_Units',
                        'Predicted_L2_Tens', 'Predicted_L2_Units']:
                backtest_df_mock[col] = backtest_df_mock[col].astype(str)

            backtest_df_mock['Predicted_R1_Last3'] = backtest_df_mock.apply(lambda row: f"{row['Predicted_R1_Last3_Hundreds']}{row['Predicted_R1_Last3_Tens']}{row['Predicted_R1_Last3_Units']}", axis=1)
            backtest_df_mock['Predicted_L2'] = backtest_df_mock.apply(lambda row: f"{row['Predicted_L2_Tens']}{row['Predicted_L2_Units']}", axis=1)
            backtest_df_mock['R1_Last3_Match'] = (backtest_df_mock['Predicted_R1_Last3'] == backtest_df_mock['เลขท้าย3ตัวรางวัลที่1'])
            backtest_df_mock['L2_Match'] = (backtest_df_mock['Predicted_L2'] == backtest_df_mock['เลขท้าย2ตัว'])
        else:
            backtest_df_mock = pd.DataFrame(columns=df_ml_mock.columns.tolist() + ['Predicted_R1_Last3_Hundreds', 'Predicted_R1_Last3_Tens', 'Predicted_R1_Last3_Units',
                                                                                 'Predicted_L2_Tens', 'Predicted_L2_Units', 'Predicted_R1_Last3', 'Predicted_L2', 'R1_Last3_Match', 'L2_Match'])
    else: # กรณีที่ df_ml_mock ไม่มีข้อมูลเพียงพอ
        trained_models_mock = {}
        backtest_df_mock = pd.DataFrame(columns=df_mock.columns.tolist() + ['Predicted_R1_Last3_Hundreds', 'Predicted_R1_Last3_Tens', 'Predicted_R1_Last3_Units',
                                                                                 'Predicted_L2_Tens', 'Predicted_L2_Units', 'Predicted_R1_Last3', 'Predicted_L2', 'R1_Last3_Match', 'L2_Match'])

    global_prediction_data = {
        'df': df_mock,
        'features': features_mock,
        'trained_models': trained_models_mock,
        'backtest_df': backtest_df_mock
    }
    print("✅ 'global_prediction_data' จำลองถูกสร้างเรียบร้อยแล้ว")

# ดึงข้อมูลจาก global_prediction_data
df = global_prediction_data['df']
features = global_prediction_data['features']
trained_models = global_prediction_data['trained_models']
backtest_df = global_prediction_data['backtest_df']

# ตรวจสอบว่า df มีข้อมูลหรือไม่
if df.empty:
    print("! DataFrame 'df' ว่างเปล่า ไม่สามารถดำเนินการต่อได้.")
    exit()

print("--- เริ่มต้นการประมวลผลคำทำนายและแสดงผลลัพธ์ ---")

# --- 1. กำหนดงวดที่จะทำนาย (ปัจจุบันคือ 1 มิ.ย. 68 -> ทำนาย 16 มิ.ย. 68) ---
last_draw_date = df['Date'].max()
print(f"\nงวดล่าสุดในข้อมูลคือ: {last_draw_date.strftime('%d %b %Y (%A)')}")

# คำนวณวันที่ของงวดถัดไป
# หวยออกวันที่ 1 และ 16 ของทุกเดือน (ยกเว้นเดือนที่มีวันหยุดพิเศษ)
# Logic ง่ายๆ: ถ้าวันล่าสุดคือ 1 ให้ทำนาย 16, ถ้าวันล่าสุดคือ 16 หรือ 17 (ม.ค. ฯลฯ) ให้ทำนายวันที 1 ของเดือนถัดไป
if last_draw_date.day == 1 or last_draw_date.day == 2: # 2 พ.ค.
    next_draw_date = last_draw_date.replace(day=16)
elif last_draw_date.day == 16 or last_draw_date.day == 17: # 17 ม.ค.
    # ไปเดือนถัดไป วันที่ 1
    if last_draw_date.month == 12:
        next_draw_date = last_draw_date.replace(year=last_draw_date.year + 1, month=1, day=1)
    else:
        next_draw_date = last_draw_date.replace(month=last_draw_date.month + 1, day=1)
else: # กรณีอื่นๆ เช่น 30/31 ธ.ค., 31 ก.ค.
    # ลองหา 16 หรือ 1 ของเดือนถัดไป หรือถัดไปอีก
    temp_date = last_draw_date + timedelta(days=1)
    while temp_date.day not in [1, 16]:
        temp_date += timedelta(days=1)
    next_draw_date = temp_date


print(f"งวดที่จะทำนายคือ: {next_draw_date.strftime('%d %b %Y (%A)')}")


# --- 2. สร้างข้อมูล Feature สำหรับงวดที่จะทำนาย ---
last_row_data = df.iloc[-1]

# ปรับปรุงการดึงค่า Prev_L2 และ Prev_R1_Last3 เพื่อจัดการกับค่าว่างหรือความยาวไม่พอ
def safe_int_extract(s, index, default=0):
    s = str(s) if pd.notna(s) else ''
    if len(s) > index:
        try:
            return int(s[index])
        except ValueError:
            return default
    return default

prev_l2_tens = safe_int_extract(last_row_data.get('เลขท้าย2ตัว'), 0)
prev_l2_units = safe_int_extract(last_row_data.get('เลขท้าย2ตัว'), 1)
prev_r1_last3_hundreds = safe_int_extract(last_row_data.get('เลขท้าย3ตัวรางวัลที่1'), 0)
prev_r1_last3_tens = safe_int_extract(last_row_data.get('เลขท้าย3ตัวรางวัลที่1'), 1)
prev_r1_last3_units = safe_int_extract(last_row_data.get('เลขท้าย3ตัวรางวัลที่1'), 2)


# สร้าง Feature สำหรับงวดถัดไป
# ค่า R1_Sum, R1_Even_Count เป็นค่าสถิติที่อาจจะไม่แม่นยำสำหรับการทำนายงวดเดียว
# สำหรับการทำนายงวดเดียว อาจใช้ค่าเฉลี่ยหรือค่าที่พบบ่อยจากข้อมูลย้อนหลัง
# หรือพิจารณารูปแบบแนวโน้มเพิ่มเติม หากมีข้อมูลเพียงพอ
# ในที่นี้จะใช้ค่าเฉลี่ยจากข้อมูลย้อนหลังเพื่อเป็นค่าเริ่มต้น
next_data_features_input = {
    'Year': next_draw_date.year,
    'Month': next_draw_date.month,
    'Day': next_draw_date.day,
    'DayOfWeek': next_draw_date.dayofweek,
    'DayOfMonth': next_draw_date.day,
    'R1_Sum': df['R1_Sum'].mean() if 'R1_Sum' in df and not df['R1_Sum'].empty else 0, # ใช้ค่าเฉลี่ยจากข้อมูลย้อนหลัง
    'R1_Even_Count': df['R1_Even_Count'].mean() if 'R1_Even_Count' in df and not df['R1_Even_Count'].empty else 0, # ใช้ค่าเฉลี่ยจากข้อมูลย้อนหลัง
    'Prev_L2_Tens': prev_l2_tens,
    'Prev_L2_Units': prev_l2_units,
    'Prev_R1_Last3_Hundreds': prev_r1_last3_hundreds,
    'Prev_R1_Last3_Tens': prev_r1_last3_tens,
    'Prev_R1_Last3_Units': prev_r1_last3_units
}
next_draw_X = pd.DataFrame([next_data_features_input])[features]

print("\n--- ข้อมูล Features สำหรับงวดที่จะทำนาย ---")
print(next_draw_X)

# --- 3. ทำนายผลโดยใช้ Ensemble Model (โหวตจากโมเดลทั้งหมด) ---
print("\n--- กำลังทำนายตัวเลขสำหรับงวดถัดไป... ---")

predicted_digits = {}
confidence_scores = {} # เก็บค่าความเชื่อมั่น (เช่น ความน่าจะเป็นสูงสุด)

for target_name in trained_models.keys():
    preds_from_all_models = []
    prob_distributions = [] # เก็บ predict_proba จากแต่ละโมเดล

    # ตรวจสอบว่า target_name มีโมเดลที่ถูกฝึกหรือไม่
    if not trained_models[target_name]:
        print(f"คำเตือน: ไม่มีโมเดลที่ถูกฝึกสำหรับ {target_name}. จะใช้ค่าเริ่มต้น 0.")
        predicted_digits[target_name] = 0
        confidence_scores[target_name] = 0.0
        continue # ข้ามไปยัง target_name ถัดไป

    for model_name, model in trained_models[target_name].items():
        try:
            pred_digit = model.predict(next_draw_X)[0]
            preds_from_all_models.append(pred_digit)

            # ตรวจสอบว่าโมเดลมี predict_proba และสามารถใช้ได้
            if hasattr(model, 'predict_proba'):
                prob_dist = model.predict_proba(next_draw_X)[0]
                prob_distributions.append(prob_dist)
            else:
                # ถ้าโมเดลไม่มี predict_proba, ให้ใช้ค่าเริ่มต้น
                prob_distributions.append(np.full(10, 1/10)) # ค่าเริ่มต้นเป็น Uniform

        except Exception as e:
            print(f"คำเตือน: เกิดข้อผิดพลาดในการทำนายด้วยโมเดล {model_name} สำหรับ {target_name}: {e}")
            # หากเกิดข้อผิดพลาดในการทำนาย ให้เพิ่มค่าเริ่มต้น (เช่น 0)
            preds_from_all_models.append(0)
            prob_distributions.append(np.full(10, 1/10)) # ค่าเริ่มต้นเป็น Uniform


    # Ensemble: Majority Vote
    if preds_from_all_models:
        most_common_digit = Counter(preds_from_all_models).most_common(1)[0][0]
    else:
        most_common_digit = 0 # ถ้าไม่มีการทำนายใดๆ ให้เป็น 0
    predicted_digits[target_name] = most_common_digit

    # Calculate Ensemble Confidence (using average probability of the predicted digit)
    # หาค่าเฉลี่ยความน่าจะเป็นของเลขที่ทำนายได้จากทุกโมเดล
    avg_confidence_list = []
    for prob in prob_distributions:
        # ตรวจสอบขนาดของ prob และ most_common_digit
        if most_common_digit < len(prob):
            avg_confidence_list.append(prob[most_common_digit])
        else:
            # ถ้า index อยู่นอกขอบเขต ให้เพิ่มค่าเริ่มต้น เช่น 0 หรือ 1/จำนวนคลาสทั้งหมด
            avg_confidence_list.append(0.0) # หรือ 1/10 สำหรับเลข 0-9
            print(f"คำเตือน: Index {most_common_digit} อยู่นอกขอบเขตของ prob array (ขนาด {len(prob)}) สำหรับ {target_name}. ใช้ความเชื่อมั่น 0.0")

    if avg_confidence_list:
        avg_confidence = np.mean(avg_confidence_list)
    else:
        avg_confidence = 0.0 # ถ้าไม่มี prob_distributions เลย

    confidence_scores[target_name] = avg_confidence

print("\n--- ผลการทำนายแต่ละหลักและระดับความเชื่อมั่น ---")
print(f"Predicted R1_Last3_Hundreds: {predicted_digits.get('R1_Last3_Hundreds', 'N/A')} (Confidence: {confidence_scores.get('R1_Last3_Hundreds', 0.0):.2f})")
print(f"Predicted R1_Last3_Tens:       {predicted_digits.get('R1_Last3_Tens', 'N/A')} (Confidence: {confidence_scores.get('R1_Last3_Tens', 0.0):.2f})")
print(f"Predicted R1_Last3_Units:      {predicted_digits.get('R1_Last3_Units', 'N/A')} (Confidence: {confidence_scores.get('R1_Last3_Units', 0.0):.2f})")
print(f"Predicted L2_Tens:             {predicted_digits.get('L2_Tens', 'N/A')} (Confidence: {confidence_scores.get('L2_Tens', 0.0):.2f})")
print(f"Predicted L2_Units:            {predicted_digits.get('L2_Units', 'N/A')} (Confidence: {confidence_scores.get('L2_Units', 0.0):.2f})")


# --- 4. แสดงผลตัวเลขคาดการณ์ที่ดีที่สุด (4 ชุดตัวเลข) ---
print("\n--- ตัวเลขคาดการณ์สำหรับงวดถัดไป (" f"{next_draw_date.strftime('%d %b %Y (%A)')}" ") ---")

# สำหรับเลขท้าย 3 ตัวรางวัลที่ 1
predicted_r1_last3 = f"{predicted_digits.get('R1_Last3_Hundreds', '')}{predicted_digits.get('R1_Last3_Tens', '')}{predicted_digits.get('R1_Last3_Units', '')}"
print(f"**เลขท้าย 3 ตัวรางวัลที่ 1 ที่คาดการณ์ (ชุดที่ดีที่สุด): {predicted_r1_last3}**")

# สำหรับเลขท้าย 2 ตัวล่าง
predicted_l2 = f"{predicted_digits.get('L2_Tens', '')}{predicted_digits.get('L2_Units', '')}"
print(f"**เลขท้าย 2 ตัวล่างที่คาดการณ์ (ชุดที่ดีที่สุด): {predicted_l2}**")

# การสร้าง "4 ชุดตัวเลขคาดการณ์ที่ดีที่สุด"
print("\n--- เลขคาดการณ์เพิ่มเติม (จากการวิเคราะห์และโมเดลรอง) ---")

# ฟังก์ชันช่วยในการทำนายจากโมเดลเดี่ยว
def get_single_model_prediction(model_dict, target_name, model_type, next_draw_X):
    if target_name in model_dict and model_type in model_dict[target_name]:
        model = model_dict[target_name][model_type]
        try:
            return model.predict(next_draw_X)[0]
        except Exception as e:
            print(f"คำเตือน: ไม่สามารถทำนายด้วยโมเดล {model_type} สำหรับ {target_name}: {e}")
            return '' # คืนค่าว่างถ้าเกิดข้อผิดพลาด
    return '' # คืนค่าว่างถ้าไม่พบโมเดล

# ชุดที่ 2: จาก RandomForest โดยตรง (หากต่างจาก Ensemble)
rf_r1_h_pred_single = get_single_model_prediction(trained_models, 'R1_Last3_Hundreds', 'RandomForest', next_draw_X)
rf_r1_t_pred_single = get_single_model_prediction(trained_models, 'R1_Last3_Tens', 'RandomForest', next_draw_X)
rf_r1_u_pred_single = get_single_model_prediction(trained_models, 'R1_Last3_Units', 'RandomForest', next_draw_X)
rf_l2_t_pred_single = get_single_model_prediction(trained_models, 'L2_Tens', 'RandomForest', next_draw_X)
rf_l2_u_pred_single = get_single_model_prediction(trained_models, 'L2_Units', 'RandomForest', next_draw_X)

predicted_r1_last3_rf = f"{rf_r1_h_pred_single}{rf_r1_t_pred_single}{rf_r1_u_pred_single}"
predicted_l2_rf = f"{rf_l2_t_pred_single}{rf_l2_u_pred_single}"

if predicted_r1_last3_rf != predicted_r1_last3 and predicted_r1_last3_rf: # เพิ่ม check ว่าไม่ว่างเปล่า
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก RandomForest): {predicted_r1_last3_rf}")
else:
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก RandomForest): เหมือนชุดที่ดีที่สุด")

if predicted_l2_rf != predicted_l2 and predicted_l2_rf: # เพิ่ม check ว่าไม่ว่างเปล่า
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก RandomForest): {predicted_l2_rf}")
else:
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก RandomForest): เหมือนชุดที่ดีที่สุด")


# ชุดที่ 3: จาก XGBoost (GradientBoosting) โดยตรง (หากต่างจาก Ensemble)
xgb_r1_h_pred_single = get_single_model_prediction(trained_models, 'R1_Last3_Hundreds', 'XGBoost', next_draw_X)
xgb_r1_t_pred_single = get_single_model_prediction(trained_models, 'R1_Last3_Tens', 'XGBoost', next_draw_X)
xgb_r1_u_pred_single = get_single_model_prediction(trained_models, 'R1_Last3_Units', 'XGBoost', next_draw_X)
xgb_l2_t_pred_single = get_single_model_prediction(trained_models, 'L2_Tens', 'XGBoost', next_draw_X)
xgb_l2_u_pred_single = get_single_model_prediction(trained_models, 'L2_Units', 'XGBoost', next_draw_X)

predicted_r1_last3_xgb = f"{xgb_r1_h_pred_single}{xgb_r1_t_pred_single}{xgb_r1_u_pred_single}"
predicted_l2_xgb = f"{xgb_l2_t_pred_single}{xgb_l2_u_pred_single}"

if predicted_r1_last3_xgb != predicted_r1_last3 and predicted_r1_last3_xgb != predicted_r1_last3_rf and predicted_r1_last3_xgb:
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก XGBoost): {predicted_r1_last3_xgb}")
else:
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก XGBoost): เหมือนชุดที่ดีที่สุด/ชุดที่ 2")

if predicted_l2_xgb != predicted_l2 and predicted_l2_xgb != predicted_l2_rf and predicted_l2_xgb:
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก XGBoost): {predicted_l2_xgb}")
else:
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก XGBoost): เหมือนชุดที่ดีที่สุด/ชุดที่ 2")


# ชุดที่ 4: จาก MLPClassifier โดยตรง (หากต่างจาก Ensemble)
mlp_r1_h_pred_single = get_single_model_prediction(trained_models, 'R1_Last3_Hundreds', 'MLPClassifier', next_draw_X)
mlp_r1_t_pred_single = get_single_model_prediction(trained_models, 'R1_Last3_Tens', 'MLPClassifier', next_draw_X)
mlp_r1_u_pred_single = get_single_model_prediction(trained_models, 'R1_Last3_Units', 'MLPClassifier', next_draw_X)
mlp_l2_t_pred_single = get_single_model_prediction(trained_models, 'L2_Tens', 'MLPClassifier', next_draw_X)
mlp_l2_u_pred_single = get_single_model_prediction(trained_models, 'L2_Units', 'MLPClassifier', next_draw_X)

predicted_r1_last3_mlp = f"{mlp_r1_h_pred_single}{mlp_r1_t_pred_single}{mlp_r1_u_pred_single}"
predicted_l2_mlp = f"{mlp_l2_t_pred_single}{mlp_l2_u_pred_single}"

if predicted_r1_last3_mlp != predicted_r1_last3 and predicted_r1_last3_mlp != predicted_r1_last3_rf and predicted_r1_last3_mlp != predicted_r1_last3_xgb and predicted_r1_last3_mlp:
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก MLPClassifier): {predicted_r1_last3_mlp}")
else:
    print(f"  - เลขท้าย 3 ตัวรางวัลที่ 1 (จาก MLPClassifier): เหมือนชุดอื่นๆ")

if predicted_l2_mlp != predicted_l2 and predicted_l2_mlp != predicted_l2_rf and predicted_l2_mlp != predicted_l2_xgb and predicted_l2_mlp:
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก MLPClassifier): {predicted_l2_mlp}")
else:
    print(f"  - เลขท้าย 2 ตัวล่าง (จาก MLPClassifier): เหมือนชุดอื่นๆ")


# --- 5. แสดงผลการ Backtesting ที่ทำได้ ---
print("\n--- สรุปผลการ Backtesting (ประสิทธิภาพของระบบในการทำนายย้อนหลัง) ---")
if backtest_df is not None and not backtest_df.empty:
    total_r1_last3_matches = backtest_df['R1_Last3_Match'].sum()
    total_l2_matches = backtest_df['L2_Match'].sum()
    total_test_samples = len(backtest_df)

    print(f"จำนวนงวดที่ทำการทดสอบย้อนหลัง: {total_test_samples} งวด")
    if total_test_samples > 0:
        print(f"ระบบทำนาย 'เลขท้าย 3 ตัวรางวัลที่ 1' ถูกต้อง: {total_r1_last3_matches} งวด ({total_r1_last3_matches/total_test_samples:.2%} ความแม่นยำ)")
        print(f"ระบบทำนาย 'เลขท้าย 2 ตัวล่าง' ถูกต้อง: {total_l2_matches} งวด ({total_l2_matches/total_test_samples:.2%} ความแม่นยำ)")
    else:
        print("ไม่มีข้อมูลสำหรับการทดสอบย้อนหลัง.")

    print("\n--- ผลการ Backtesting แยกตามงวด (5 งวดล่าสุด) ---")
    if not backtest_df.empty:
        print(backtest_df[['Date', 'เลขท้าย3ตัวรางวัลที่1', 'Predicted_R1_Last3', 'R1_Last3_Match',
                           'เลขท้าย2ตัว', 'Predicted_L2', 'L2_Match']].tail())
    else:
        print("ไม่มีข้อมูล Backtesting ให้แสดง.")
else:
    print("ไม่มีข้อมูล Backtesting ที่พร้อมใช้งาน.")

# --- 6. แสดงผลการทำงานของระบบแบบโปร่งใสเป็นขั้นตอนด้วยภาษาไทย ---
print("\n--- สรุปการทำงานของระบบคาดการณ์หวยไทย (Hybrid Intelligence) ---")
print("ระบบนี้ถูกสร้างขึ้นโดยใช้แนวคิด Hybrid Intelligence ซึ่งผสานรวมหลายวิธีการเพื่อเพิ่มความแม่นยำและน่าเชื่อถือ:")
print("1. **การเตรียมข้อมูล:** ทำความสะอาด, แปลงวันที่, และเติมเลขศูนย์ให้ครบหลัก เพื่อให้ข้อมูลพร้อมใช้งาน")
print("2. **วิศวกรรมฟีเจอร์ (Feature Engineering):** สร้างคุณลักษณะใหม่ๆ จากข้อมูล เช่น เลขแต่ละหลัก, ผลรวม, เลขคู่/คี่, วันที่ออก, รวมถึงเลขของงวดก่อนหน้า เพื่อให้โมเดลมีข้อมูลที่หลากหลาย")
print("3. **การวิเคราะห์ทางสถิติ (Statistical Analysis):**")
print("    - **EDA:** วิเคราะห์การกระจายตัวของตัวเลข, ความถี่, และความสัมพันธ์เบื้องต้น")
print("    - **Chi-square Test:** ทดสอบว่าการกระจายตัวของเลขแต่ละหลักเป็นแบบสุ่ม (Uniform) หรือไม่ หาก P-value น้อยกว่า 0.05 แสดงว่าอาจมีแนวโน้มที่ไม่ใช่การสุ่มสมบูรณ์")
print("    - **Autocorrelation (ACF)/PACF:** วิเคราะห์หาความสัมพันธ์ของตัวเลขกับงวดก่อนหน้า เพื่อดูว่ามี Pattern ที่ซ้ำกันในระยะเวลาหนึ่งหรือไม่")
print("    - **Probability Distribution Modeling:** วิเคราะห์ความน่าจะเป็นของแต่ละเลขที่ออก เพื่อเปรียบเทียบกับการกระจายตัวแบบสุ่ม")
print("4. **การสร้างและฝึกโมเดล AI/ML:**")
print("    - ใช้โมเดล Machine Learning หลากหลายชนิด เช่น **Random Forest**, **Gradient Boosting (XGBoost)**, และ **MLP Classifier** (Neural Network) เพื่อเรียนรู้ Pattern จาก Features ที่สร้างขึ้น")
print("    - แต่ละหลักของเลขท้าย 3 ตัว และเลขท้าย 2 ตัว จะถูกทำนายแยกกัน เพื่อความแม่นยำสูงสุด")
print("5. **การรวมโมเดล (Ensemble Method):** ใช้เทคนิค 'Majority Voting' เพื่อรวมผลการทำนายจากโมเดลแต่ละตัวเข้าด้วยกัน ทำให้ผลลัพธ์มีความเสถียรและแม่นยำยิ่งขึ้น")
print("6. **การทดสอบย้อนหลัง (Backtesting):** ระบบจะแบ่งข้อมูลออกเป็นส่วน Train และ Test (โดยใช้ 80% แรกสำหรับ Train และ 20% หลังสำหรับ Test ตามลำดับเวลา) เพื่อจำลองการทำนายงวดในอดีตและประเมินประสิทธิภาพของโมเดล")
print("7. **การประเมินความเชื่อมั่น (Confidence Measures):** แสดงค่าความเชื่อมั่นสำหรับเลขแต่ละหลักที่ทำนายได้ โดยอ้างอิงจากความน่าจะเป็นที่โมเดลคาดการณ์")
print("8. **การคาดการณ์สำหรับงวดถัดไป:** ใช้โมเดลที่ฝึกมาทำนายเลขท้าย 3 ตัวรางวัลที่ 1 และเลขท้าย 2 ตัวล่างสำหรับงวดถัดไป พร้อมนำเสนอชุดตัวเลขคาดการณ์ 4 ชุด")
print("\n**หมายเหตุ:** การทำนายหวยเป็นการคาดการณ์จากข้อมูลในอดีต ซึ่งเป็นการสุ่มที่มีความซับซ้อนสูง แม้ระบบจะใช้แนวทางทางสถิติและ AI/ML ขั้นสูง แต่ก็ไม่สามารถรับประกันผลได้ 100% จุดประสงค์หลักคือการช่วยในการตัดสินใจโดยมีข้อมูลและเหตุผลทางสถิติรองรับ")

print("\n--- การประมวลผลคำทำนายและแสดงผลลัพธ์เสร็จสิ้น ---")
```
