```python
# ---- 0. เตรียมไลบรารี ----
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter, defaultdict
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from datetime import timedelta # เพิ่มเข้ามาสำหรับจัดการวันที่

# ตรวจสอบและจัดการการ import Keras/TensorFlow
# หาก Colab มีติดตั้ง TensorFlow อยู่แล้ว สามารถ import ได้เลย
KERAS_AVAILABLE = False
try:
    import tensorflow as tf
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import LSTM, Dense, Embedding
    from tensorflow.keras.utils import to_categorical
    # ตรวจสอบเวอร์ชัน Keras (TensorFlow 2.x ใช้ Keras ในตัว)
    if tf.__version__.startswith('2'):
        print(f"TensorFlow version {tf.__version__} is installed and Keras is available.")
        KERAS_AVAILABLE = True
except ImportError:
    print("TensorFlow/Keras not found. LSTM related functionalities will be skipped.")

import warnings

# ปิด warning สำหรับ to_categorical (เป็นข้อควรระวังเรื่องค่าที่หายไป)
# และ warning อื่นๆ ที่ไม่จำเป็น
warnings.filterwarnings('ignore', category=FutureWarning, module='tensorflow')
warnings.filterwarnings('ignore', category=UserWarning, module='keras')

# --- ข้อมูล CSV ที่เราใช้มาตลอด (ข้อมูลใหม่ที่คุณให้มา) ---
csv_content = """วันที่,รางวัลที่ 1 (6 หลัก),เลข 2 ตัวล่าง
2025/07/01,949246,91
2025/06/16,507392,06
2025/06/01,559352,20
2025/05/16,251309,87
2025/05/02,213388,06
2025/04/16,266227,85
2025/04/01,669687,36
2025/03/16,757563,32
2025/03/01,818894,54
2025/02/16,847377,50
2025/02/01,558700,51
2025/01/17,807779,23
2025/01/02,730209,51
2024/12/16,097863,21
2024/12/01,669843,61
2024/11/16,187221,38
2024/11/01,536044,32
2024/10/16,482962,00
2024/10/01,718665,59
2024/09/16,608662,37
2024/09/01,199606,94
2024/08/16,095867,28
2024/08/01,407041,46
2024/07/16,367336,21
2024/07/01,434503,89
2024/06/16,518504,31
2024/06/01,530593,42
2024/05/16,205690,60
2024/05/02,980116,17
2024/04/16,943598,79
2024/04/01,803481,90
2024/03/16,997626,78
2024/03/01,253603,79
2024/02/16,941395,43
2024/02/01,607063,09
2024/01/17,105979,61
2023/12/30,625544,89
2023/12/16,356757,85
2023/12/01,251097,91
2023/11/16,557990,14
2023/11/01,743951,63
2023/10/16,931446,44
2023/10/01,727202,66
2023/09/16,320812,46
2023/09/01,915478,91
2023/08/16,471782,67
2023/07/31,260453,11
2023/07/16,169530,62
2023/07/01,922605,16
2023/06/16,264872,30
2023/06/01,125272,09
2023/05/16,132903,99
2023/05/02,843019,65
2023/04/16,984906,71
2023/04/01,087907,99
2023/03/16,025873,73
2023/03/01,417652,55
2023/02/16,590417,80
2023/02/01,297411,92
2023/01/17,812519,47
2022/12/30,157196,58
2022/12/16,845093,14
2022/12/01,375805,08
2022/11/16,121789,64
2022/11/01,913106,70
2022/10/16,613106,15
2022/10/01,484669,50
2022/09/16,943703,75
2022/09/01,929332,83
2022/08/16,331583,42
2022/08/01,436594,14
2022/07/16,620405,53
2022/07/01,981417,61
2022/06/16,361807,92
2022/06/01,319196,02
2022/05/16,155012,06
2022/05/02,658642,09
2022/04/16,395919,58
2022/04/01,970618,10
2022/03/16,737867,03
2022/03/01,061905,07
2022/02/17,098597,57
2022/02/01,944308,30
2022/01/17,880159,92
2021/12/30,819068,36
2021/12/16,639235,83
2021/12/01,077258,82
2021/11/16,032761,57
2021/11/01,045037,95
2021/10/16,386372,38
2021/10/01,578171,83
2021/09/16,070935,90
2021/09/01,114475,79
2021/08/16,046750,23
2021/08/01,910261,69
2021/07/16,556725,70
2021/07/01,713517,29
2021/06/16,691861,17
2021/06/01,292972,45
2021/05/16,684579,14
2021/05/02,501272,18
2021/04/16,100787,56
2021/04/01,472270,05
2021/03/16,890422,19
2021/03/01,835538,73
2021/02/16,424603,39
2021/02/01,912307,97
2021/01/17,384395,15
2020/12/30,803628,19
2020/12/16,201303,70
2020/12/01,100994,84
2020/11/16,972661,46
2020/11/01,506404,40
2020/10/16,286051,38
2020/10/01,837893,59
2020/09/16,244083,57
2020/09/01,999997,98
2020/08/16,945811,88
2020/08/01,569391,92
2020/07/16,873286,53
2020/07/01,347258,83
2020/06/16,516967,64
2020/06/01,831567,24
2020/04/01,051095,22
2020/03/16,503446,77
2020/03/01,875938,98
2020/02/16,781403,94
2020/02/01,589227,06
2020/01/17,491774,68
2019/12/30,510541,81
2019/12/16,529924,97
2019/12/01,453522,81
2019/11/16,017223,32
2019/11/01,967375,79
2019/10/16,812564,15
2019/10/01,691197,59
2019/09/16,340388,85
2019/09/01,798787,20
2019/08/16,775476,89
2019/08/01,387006,58
2019/07/15,369765,88
2019/07/01,943647,86
2019/06/16,174055,29
2019/06/01,516461,46
2019/05/16,962526,71
2019/05/02,061324,25
2019/04/16,570331,23
2019/04/01,109767,52
2019/03/16,724628,64
2019/03/01,345650,65
2019/02/16,074824,56
2019/02/01,967134,04
2019/01/17,197079,65
2018/12/30,735867,02
2018/12/16,356564,62
2018/12/01,021840,67
2018/11/16,989903,16
2018/11/01,149840,58
2018/10/16,200515,93
2018/10/01,452643,99
2018/09/16,149760,79
2018/09/01,734510,26
2018/08/16,586117,10
2018/08/01,386602,78
2018/07/16,596324,27
2018/07/01,963623,83
2018/06/16,223131,46
2018/06/01,988117,95
2018/05/16,075629,20
2018/05/02,248038,85
2018/04/16,739229,60
2018/04/01,412073,85
2018/03/16,218559,82
2018/03/02,759415,29
2018/02/16,309915,39
2018/02/01,026853,31
2018/01/17,203823,50
2017/12/30,911234,98
2017/12/16,955596,89
2017/12/01,451005,33
2017/11/16,292391,98
2017/11/01,533726,85
2017/10/16,413494,86
2017/10/01,880714,52
2017/09/16,170143,71
2017/09/01,143224,65
2017/08/16,715431,37
2017/08/01,756519,36
2017/07/16,820327,87
2017/07/01,112360,26
2017/06/16,943142,47
2017/06/01,053630,61
2017/05/16,454891,53
2017/05/02,008656,35
2017/04/16,816729,40
2017/04/01,392785,80
2017/03/16,273863,92
2017/03/01,978453,78
2017/02/16,229116,14
2017/02/01,054672,42
2017/01/17,145157,25
2016/12/30,377712,46
2016/12/16,435286,35
2016/12/01,086069,77
2016/11/16,858383,44
2016/11/01,785438,86
2016/10/16,571947,98
2016/10/01,887102,33
2016/09/16,240650,42
2016/09/01,638684,62
2016/08/16,254004,33
2016/08/01,272932,57
2016/07/16,449764,55
2016/07/01,082460,53
2016/06/16,073816,79
2016/06/01,511825,14
2016/05/16,141737,98
2016/05/02,399459,02
2016/04/16,221609,87
2016/04/01,066720,92
2016/03/16,134918,32
2016/03/01,439686,06
2016/02/16,356364,98
2016/02/01,927800,09
2016/01/17,304371,50
2015/12/30,008217,02
2015/12/17,930255,08
2015/12/01,915350,78
2015/11/16,795283,03
2015/11/01,361211,45
2015/10/16,968630,62
2015/10/01,594825,07
2015/09/16,743148,06
2015/09/01,021094,89
2015/08/16,033363,40
2015/08/01,518677,53
2015/07/16,121507,49
2015/07/01,759049,26
2015/06/16,644742,05
2015/06/02,388881,65
2015/05/16,011421,38
2015/05/02,543466,30
2015/04/16,506260,38
2015/04/01,605704,70
2015/03/16,048151,92
2015/03/01,240237,34
2015/02/16,001864,90
2015/02/01,155537,79
2015/01/16,244351,74
2014/12/30,461704,57
2014/12/16,948354,90
2014/12/01,480449,11
2014/11/16,479804,25
2014/11/01,206608,44
2014/10/16,656409,94
2014/10/01,375615,44
2014/09/16,772269,35
2014/09/01,856763,22
2014/08/16,662842,91
2014/08/01,766391,82
2014/07/16,468728,45
2014/07/01,378477,39
2014/06/16,673920,95
2014/06/01,781198,18
2014/05/16,087523,20
2014/05/02,103297,52
2014/04/16,153406,26
2014/04/01,028866,95
2014/03/16,531404,79
2014/03/01,906318,35
2014/02/16,384245,01
2014/02/01,180149,95
2014/01/16,306902,52
2013/12/30,561072,48
2013/12/16,341767,79
2013/12/01,168795,27
2013/11/16,806925,28
2013/11/01,739804,47
2013/10/16,963289,60
2013/10/01,647882,14
2013/09/16,562684,63
2013/09/01,548123,05
2013/08/16,321327,20
2013/08/01,356435,82
2013/07/16,566996,86
2013/07/01,646905,51
2013/06/16,289673,69
2013/06/01,935489,90
2013/05/16,687125,56
2013/05/02,603458,07
2013/04/16,843846,86
2013/04/01,571688,53
2013/03/16,968433,52
2013/03/01,976241,37
2013/02/16,368257,09
2013/02/01,565566,66
2013/01/16,820981,08
2012/12/30,302358,00
2012/12/16,529524,72
2012/12/01,110443,43
2012/11/16,639500,15
2012/11/01,524694,63
2012/10/16,281343,28
2012/10/01,124025,58
2012/09/16,540143,79
2012/09/01,329997,07
2012/08/16,683877,28
2012/08/01,895590,50
2012/07/16,904050,11
2012/07/01,915900,60
2012/06/16,159373,51
2012/06/01,882727,38
2012/05/16,814418,31
2012/05/02,889501,29
2012/04/16,583470,62
2012/04/01,257562,69
2012/03/16,607064,08
2012/03/01,222518,79
2012/02/16,648684,18
2012/02/01,320605,32
2012/01/16,451445,81
2011/12/30,526402,65
2011/12/16,884178,21
2011/12/01,408147,02
2011/11/16,997777,57
2011/11/01,805540,54
2011/10/16,955756,83
2011/10/01,511052,15
2011/09/16,731198,28
2011/09/01,724533,85
2011/08/16,536960,62
2011/08/01,218756,12
2011/07/16,116556,12
2011/07/01,622953,51
2011/06/16,351276,88
2011/06/01,562370,46
2011/05/16,406417,05
2011/05/02,054136,85
2011/04/16,825988,44
2011/04/01,814931,01
2011/03/16,593331,96
2011/03/01,656037,97
2011/02/16,481746,27
2011/02/01,610089,55
2011/01/16,281062,23
2010/12/30,884112,49
2010/12/16,334380,24
2010/12/01,181752,09
2010/11/16,813993,43
2010/11/01,191100,59
2010/10/16,621377,42
2010/10/01,488372,02
2010/09/16,017422,66
2010/09/01,354656,11
2010/08/16,911097,64
2010/08/01,210008,10
2010/07/16,180387,34
2010/07/01,480239,68
2010/06/16,500104,73
2010/06/01,444874,81
2010/05/16,480012,12
2010/05/02,360371,06
2010/04/16,211743,96
2010/04/01,959517,22
2010/03/16,364222,97
2010/03/01,215227,97
2010/02/16,133707,03
2010/02/01,186312,14
2010/01/16,073577,67
2009/12/30,994304,87
2009/12/16,685141,05
2009/12/01,776980,59
2009/11/16,055986,58
2009/11/01,689140,85
2009/10/16,258487,00
2009/10/01,169387,06
2009/09/16,202912,48
2009/09/01,015865,32
2009/08/16,462933,96
2009/08/01,154986,92
2009/07/16,000816,94
2009/07/01,207542,66
2009/06/16,930456,15
2009/06/01,777661,26
2009/05/16,111411,54
2009/05/02,294452,11
2009/04/16,368415,33
2009/04/01,816578,50
2009/03/16,268812,36
2009/03/01,553091,67
2009/02/16,038730,93
2009/02/01,534533,69
2009/01/16,743212,25
2008/12/30,218596,22
2008/12/16,074114,25
2008/12/01,205434,05
2008/11/16,002612,20
2008/11/01,272028,76
2008/10/16,431277,98
2008/10/01,882911,67
2008/09/16,012377,56
2008/09/01,695993,09
2008/08/16,380377,36
2008/08/01,850348,11
2008/07/16,257374,41
2008/07/01,943671,50
2008/06/16,729111,75
2008/06/01,414875,35
2008/05/16,329231,69
2008/05/02,453011,62
2008/04/16,982800,64
2008/04/01,012653,71
2008/03/16,074946,33
2008/03/01,936685,05
2008/02/16,137054,80
2008/02/01,212684,26
2008/01/16,556010,81
2007/12/30,595411,81
2007/12/16,513501,96
2007/12/01,113410,18
2007/11/16,562481,73
2007/11/01,927907,88
2007/10/16,032988,48
2007/10/01,430667,76
2007/09/16,499336,45
2007/09/01,331810,69
2007/08/16,476207,93
2007/08/01,429924,29
2007/07/16,527384,77
2007/07/01,565151,76
2007/06/16,393194,41
2007/06/01,836393,05
2007/05/16,232897,25
2007/05/02,430374,81
2007/04/16,405105,63
2007/04/01,622780,93
2007/03/16,876763,85
2007/03/01,742425,61
2007/02/16,229116,14
2007/02/01,054672,42
2007/01/17,145157,25
2006/12/30,377712,46
2006/12/16,435286,35
2006/12/01,086069,77
2006/11/16,858383,44
2006/11/01,785438,86
2006/10/16,571947,98
2006/10/01,887102,33
2006/09/16,240650,42
2006/09/01,638684,62
2006/08/16,254004,33
2006/08/01,272932,57
2006/07/16,449764,55
2006/07/01,082460,53
2006/06/16,073816,79
2006/06/01,511825,14
2006/05/16,141737,98
2006/05/02,399459,02
2006/04/16,221609,87
2006/04/01,066720,92
2006/03/16,134918,32
2006/03/01,439686,06
2006/02/16,356364,98
2006/02/01,927800,09
2006/01/16,304371,50
2005/12/30,008217,02
2005/12/17,930255,08
2005/12/01,915350,78
2005/11/16,795283,03
2005/11/01,361211,45
2005/10/16,968630,62
2005/10/01,594825,07
2005/09/16,743148,06
2005/09/01,021094,89
2005/08/16,033363,40
2005/08/01,518677,53
2005/07/16,121507,49
2005/07/01,759049,26
2005/06/16,644742,05
2005/06/02,388881,65
2005/05/16,011421,38
2005/05/02,543466,30
2005/04/16,506260,38
2005/04/01,605704,70
2005/03/16,048151,92
2005/03/01,240237,34
2005/02/16,001864,90
2005/02/01,155537,79
2005/01/16,244351,74
2004/12/30,461704,57
2004/12/16,948354,90
2004/12/01,480449,11
2004/11/16,479804,25
2004/11/01,206608,44
2004/10/16,656409,94
2004/10/01,375615,44
2004/09/16,772269,35
2004/09/01,856763,22
2004/08/16,662842,91
2004/08/01,766391,82
2004/07/16,468728,45
2004/07/01,378477,39
2004/06/16,673920,95
2004/06/01,781198,18
2004/05/16,087523,20
2004/05/02,103297,52
2004/04/16,153406,26
2004/04/01,028866,95
2004/03/16,531404,79
2004/03/01,906318,35
2004/02/16,384245,01
2004/02/01,180149,95
2004/01/16,306902,52
2003/12/30,561072,48
2003/12/16,341767,79
2003/12/01,168795,27
2003/11/16,238511,68
2003/11/01,941438,47
2003/10/16,305500,03
2003/10/01,912040,43
2003/09/16,600589,53
2003/09/01,187813,92
2003/08/16,354771,00
2003/08/01,766098,91
2003/07/16,679545,75
"""

# --- คลาส LotteryDataProcessor ที่ปรับปรุงใหม่ ---
class LotteryDataProcessor:
    def __init__(self):
        pass

    def _get_digit_sum_and_parity(self, number_str):
        """คำนวณผลรวมของหลัก และตรวจสอบว่าเป็นคู่หรือคี่"""
        digits = [int(d) for d in str(number_str) if d.isdigit()]
        s = sum(digits)
        return s, 'even' if s % 2 == 0 else 'odd'

    def _get_digit_frequency(self, number_str):
        """นับความถี่ของแต่ละหลักในตัวเลข"""
        freq = {str(i): 0 for i in range(10)}
        for d in str(number_str):
            if d.isdigit():
                freq[d] += 1
        return freq

    def _get_consecutive_count(self, number_str):
        """นับจำนวนตัวเลขที่เรียงติดกัน (เช่น 12, 23)"""
        s_num = str(number_str)
        count = 0
        for i in range(len(s_num) - 1):
            if int(s_num[i+1]) - int(s_num[i]) == 1:
                count += 1
        return count

    def _get_repeated_digit_count(self, number_str):
        """นับจำนวนตัวเลขที่ซ้ำกัน (เช่น 11, 344)"""
        s_num = str(number_str)
        counts = {}
        for d in s_num:
            if d.isdigit():
                counts[d] = counts.get(d, 0) + 1
        return sum(1 for d in counts if counts[d] > 1)

    def generate_features(self, df_raw_data, window_size=8):
        """
        สร้างฟีเจอร์เชิงลึกจากข้อมูล DataFrame ที่ประมวลผลแล้ว
        โดยพิจารณาข้อมูลในอดีต (window_size)
        """
        if df_raw_data.empty:
            return pd.DataFrame()

        feature_list = []

        # Make sure data is sorted by date
        df = df_raw_data.sort_values(by='date', ascending=True).copy()

        for i in range(len(df)):
            current_entry = df.iloc[i]

            # ดึงข้อมูลย้อนหลังตาม window_size
            # เราจะใช้ข้อมูลจากงวดก่อนหน้าถึงงวดที่ i-1 สำหรับ feature ที่เป็น cumulative/historical
            start_idx = max(0, i - window_size)
            window_data = df.iloc[start_idx:i] # ไม่รวมงวดปัจจุบัน

            features = {
                'draw_date': current_entry['date'],
                'full_prize_1': current_entry['prize1'],
                'bottom2_full': current_entry['bottom2'], # เปลี่ยนชื่อให้ชัดเจน
                'prize1_3d_full': current_entry['prize1_3d'], # ใช้ prize1_3d ที่เตรียมไว้แล้ว
            }

            # --- Features สำหรับเลข 2 ตัวล่าง (Current Draw) ---
            # ใช้ current_entry['bottom2'] แทน 'two_digit_bottom'
            sum_2d, parity_2d = self._get_digit_sum_and_parity(current_entry['bottom2'])
            features['2d_sum'] = sum_2d
            features['2d_parity'] = 1 if parity_2d == 'even' else 0 # แปลงเป็นตัวเลข
            features['2d_consecutive_count'] = self._get_consecutive_count(current_entry['bottom2'])
            features['2d_repeated_digit_count'] = self._get_repeated_digit_count(current_entry['bottom2'])

            # --- Features สำหรับเลขท้าย 3 ตัว (รางวัลที่ 1) (Current Draw) ---
            # ใช้ current_entry['prize1_3d'] แทน 'three_digit_top'
            sum_3d, parity_3d = self._get_digit_sum_and_parity(current_entry['prize1_3d'])
            features['3d_sum'] = sum_3d
            features['3d_parity'] = 1 if parity_3d == 'even' else 0 # แปลงเป็นตัวเลข
            features['3d_consecutive_count'] = self._get_consecutive_count(current_entry['prize1_3d'])
            features['3d_repeated_digit_count'] = self._get_repeated_digit_count(current_entry['prize1_3d'])

            # --- Cumulative Features จากข้อมูลย้อนหลัง (ถ้ามี) ---
            if not window_data.empty:
                # Cumulative features for 2D
                features['avg_2d_sum_prev_window'] = window_data['bottom2'].apply(lambda x: self._get_digit_sum_and_parity(x)[0]).mean()
                features['std_2d_sum_prev_window'] = window_data['bottom2'].apply(lambda x: self._get_digit_sum_and_parity(x)[0]).std()
                features['2d_even_parity_ratio_prev_window'] = (window_data['bottom2'].apply(lambda x: self._get_digit_sum_and_parity(x)[1] == 'even')).mean()

                # Cumulative features for 3D
                features['avg_3d_sum_prev_window'] = window_data['prize1_3d'].apply(lambda x: self._get_digit_sum_and_parity(x)[0]).mean()
                features['std_3d_sum_prev_window'] = window_data['prize1_3d'].apply(lambda x: self._get_digit_sum_and_parity(x)[0]).std()
                features['3d_even_parity_ratio_prev_window'] = (window_data['prize1_3d'].apply(lambda x: self._get_digit_sum_and_parity(x)[1] == 'even')).mean()

                # --- Frequency of individual digits in previous window ---
                # สำหรับ 2D
                all_2d_digits = ''.join(str(x) for x in window_data['bottom2'])
                freq_2d_digits = self._get_digit_frequency(all_2d_digits)
                for d in range(10):
                    features[f'2d_digit_{d}_freq_prev_window'] = freq_2d_digits.get(str(d), 0)

                # สำหรับ 3D
                all_3d_digits = ''.join(str(x) for x in window_data['prize1_3d'])
                freq_3d_digits = self._get_digit_frequency(all_3d_digits)
                for d in range(10):
                    features[f'3d_digit_{d}_freq_prev_window'] = freq_3d_digits.get(str(d), 0)

            else: # ถ้าไม่มีข้อมูลย้อนหลัง (งวดแรกๆ) ให้ใส่ค่า 0 หรือ NaN
                features['avg_2d_sum_prev_window'] = 0
                features['std_2d_sum_prev_window'] = 0
                features['2d_even_parity_ratio_prev_window'] = 0
                features['avg_3d_sum_prev_window'] = 0
                features['std_3d_sum_prev_window'] = 0
                features['3d_even_parity_ratio_prev_window'] = 0
                for d in range(10):
                    features[f'2d_digit_{d}_freq_prev_window'] = 0
                    features[f'3d_digit_{d}_freq_prev_window'] = 0

            feature_list.append(features)

        # สร้าง DataFrame จาก feature_list
        features_df = pd.DataFrame(feature_list)

        # แทนที่ค่า NaN ที่อาจเกิดจาก std() ของข้อมูลน้อย
        features_df = features_df.fillna(0) # หรือใช้ features_df.replace([np.inf, -np.inf], np.nan).fillna(0) ถ้ามีค่า inf

        return features_df

# ---- 1. โหลดข้อมูลและประมวลผลเบื้องต้น (ย้ายส่วนนี้จากด้านล่างขึ้นมา) ----
df_raw = pd.read_csv(io.StringIO(csv_content.strip()), header=0)
df_raw.columns = ['date','prize1','bottom2']

df_raw['prize1'] = df_raw['prize1'].astype(str).str.zfill(6)
df_raw['bottom2'] = df_raw['bottom2'].astype(str).str.zfill(2)

df_raw['date'] = pd.to_datetime(df_raw['date'], format='%Y/%m/%d')
df_raw = df_raw.sort_values(by='date', ascending=True).reset_index(drop=True)

df_raw['year'] = df_raw['date'].dt.year
df_raw['month'] = df_raw['date'].dt.month
df_raw['day'] = df_raw['date'].dt.day
df_raw['prize1_3d'] = df_raw['prize1'].apply(lambda x: str(x)[-3:])

# เพิ่มการเตรียมข้อมูลสำหรับเมธอดอื่นๆ (ยังคงทำเหมือนเดิม)
# แตกตัวเลข 2 ตัวล่างเป็นหลักสิบและหลักหน่วย
df_raw['bottom2_digit1'] = df_raw['bottom2'].apply(lambda x: int(x[0]))
df_raw['bottom2_digit2'] = df_raw['bottom2'].apply(lambda x: int(x[1]))

# แตกตัวเลข 3 ตัวท้ายรางวัลที่ 1 เป็นแต่ละหลัก
df_raw['prize1_3d_digit1'] = df_raw['prize1_3d'].apply(lambda x: int(x[0]))
df_raw['prize1_3d_digit2'] = df_raw['prize1_3d'].apply(lambda x: int(x[1]))
df_raw['prize1_3d_digit3'] = df_raw['prize1_3d'].apply(lambda x: int(x[2]))

print("Data loading and initial processing complete in Cell 0.")
print(df_raw.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# --- 0. เตรียมข้อมูลดิบ (หากยังไม่มี df) ---
# *** สำคัญมาก: แทนที่ด้วยการโหลดหรือสร้าง DataFrame 'df' ของคุณจริง ๆ ***
# ตัวอย่างการสร้าง DataFrame 'df' จำลอง เพื่อให้โค้ดรันได้
# ในการใช้งานจริง: df = pd.read_csv('your_lottery_data.csv') หรือโหลดจากแหล่งอื่น
if 'df' not in locals() or df.empty: # ตรวจสอบว่า df มีอยู่แล้วหรือไม่ และไม่ว่างเปล่า
    print("Creating mock 'df' for demonstration. Please replace with your actual data loading.")
    num_rows = 525 # จำนวนงวดที่คุณต้องการ
    mock_data = {
        'Prize1_Full': [str(np.random.randint(0, 1000)).zfill(3) for _ in range(num_rows)],
        'Bottom2_Full': [str(np.random.randint(0, 100)).zfill(2) for _ in range(num_rows)],
        # สมมติว่ามีคอลัมน์ D1, D2, D3, B1, B2 จากการแยกหลักแล้ว
        'Prize1_D1': np.random.randint(0, 10, num_rows),
        'Prize1_D2': np.random.randint(0, 10, num_rows),
        'Prize1_D3': np.random.randint(0, 10, num_rows),
        'Bottom2_D1': np.random.randint(0, 10, num_rows),
        'Bottom2_D2': np.random.randint(0, 10, num_rows),
    }
    df = pd.DataFrame(mock_data)

    # แปลง Prize1_Full และ Bottom2_Full เป็นตัวเลขเพื่อให้คำนวณ Sum ได้ถูกต้อง (หากเป็น string)
    df['Prize1_Full_Int'] = df['Prize1_Full'].astype(int)
    df['Bottom2_Full_Int'] = df['Bottom2_Full'].astype(int)

print(f"Original DataFrame 'df' shape: {df.shape}")

# --- 1. การสร้าง Feature Engineering Functions ---

def get_sum(number_str):
    return sum(int(d) for d in str(number_str))

def get_parity(number_str):
    s = get_sum(number_str)
    return 1 if s % 2 == 0 else 0 # 1 for even, 0 for odd

def get_consecutive_count(number_str):
    count = 0
    s = str(number_str)
    for i in range(len(s) - 1):
        if abs(int(s[i]) - int(s[i+1])) == 1:
            count += 1
    return count

def get_repeated_digit_count(number_str):
    s = str(number_str)
    return len(s) - len(set(s))

# --- 2. สร้าง Features สำหรับปัจจุบัน (เลข Prize 1 และ 2-Digit Bottom) ---

features_df = pd.DataFrame(index=df.index)

# Features สำหรับ 2-Digit Bottom
features_df['2d_sum'] = df['Bottom2_Full_Int'].apply(get_sum)
features_df['2d_parity'] = df['Bottom2_Full_Int'].apply(get_parity)
features_df['2d_consecutive_count'] = df['Bottom2_Full_Int'].apply(get_consecutive_count)
features_df['2d_repeated_digit_count'] = df['Bottom2_Full_Int'].apply(get_repeated_digit_count)

# Features สำหรับ Prize 1 (3D)
features_df['3d_sum'] = df['Prize1_Full_Int'].apply(get_sum)
features_df['3d_parity'] = df['Prize1_Full_Int'].apply(get_parity)
features_df['3d_consecutive_count'] = df['Prize1_Full_Int'].apply(get_consecutive_count)
features_df['3d_repeated_digit_count'] = df['Prize1_Full_Int'].apply(get_repeated_digit_count)

# --- 3. สร้าง Lagged Features (Previous Window Statistics) ---
window_size = 10 # ปรับ window size เป็น 10 เพื่อให้โมเดลมีข้อมูลประวัติมากขึ้น

for col in ['2d_sum', '2d_parity', '2d_consecutive_count', '2d_repeated_digit_count',
            '3d_sum', '3d_parity', '3d_consecutive_count', '3d_repeated_digit_count']:
    features_df[f'avg_{col}_prev_window'] = features_df[col].rolling(window=window_size, closed='left').mean()
    features_df[f'std_{col}_prev_window'] = features_df[col].rolling(window=window_size, closed='left').std()
    features_df[f'min_{col}_prev_window'] = features_df[col].rolling(window=window_size, closed='left').min()
    features_df[f'max_{col}_prev_window'] = features_df[col].rolling(window=window_size, closed='left').max()

# เพิ่มความถี่ของแต่ละหลักในเลข 2d และ 3d ใน window ก่อนหน้า (รวมทุกหลัก)
for i in range(10): # สำหรับตัวเลข 0-9
    features_df[f'2d_digit_{i}_freq_overall_prev_window'] = 0.0
    features_df[f'3d_digit_{i}_freq_overall_prev_window'] = 0.0

    for idx in range(window_size, len(df)):
        recent_2d_numbers = df['Bottom2_Full'].iloc[idx-window_size:idx].tolist()
        all_2d_digits_in_window = ''.join(recent_2d_numbers)
        features_df.loc[idx, f'2d_digit_{i}_freq_overall_prev_window'] = all_2d_digits_in_window.count(str(i)) / (window_size * 2)

        recent_3d_numbers = df['Prize1_Full'].iloc[idx-window_size:idx].tolist()
        all_3d_digits_in_window = ''.join(recent_3d_numbers)
        features_df.loc[idx, f'3d_digit_{i}_freq_overall_prev_window'] = all_3d_digits_in_window.count(str(i)) / (window_size * 3)


# --- เพิ่มฟีเจอร์ใหม่: การปรากฏของแต่ละหลักในตำแหน่งเฉพาะ (positional frequency) ---
for i in range(10): # สำหรับตัวเลข 0-9
    # Prize 1 (3D)
    features_df[f'3d_d1_digit_{i}_freq_prev_window'] = 0.0 # หลักร้อย
    features_df[f'3d_d2_digit_{i}_freq_prev_window'] = 0.0 # หลักสิบ
    features_df[f'3d_d3_digit_{i}_freq_prev_window'] = 0.0 # หลักหน่วย
    # 2-Digit Bottom
    features_df[f'2d_d1_digit_{i}_freq_prev_window'] = 0.0 # หลักสิบ
    features_df[f'2d_d2_digit_{i}_freq_prev_window'] = 0.0 # หลักหน่วย

    for idx in range(window_size, len(df)):
        # Prize 1
        recent_p1_d1 = df['Prize1_D1'].iloc[idx-window_size:idx].tolist()
        recent_p1_d2 = df['Prize1_D2'].iloc[idx-window_size:idx].tolist()
        recent_p1_d3 = df['Prize1_D3'].iloc[idx-window_size:idx].tolist()

        features_df.loc[idx, f'3d_d1_digit_{i}_freq_prev_window'] = recent_p1_d1.count(i) / window_size
        features_df.loc[idx, f'3d_d2_digit_{i}_freq_prev_window'] = recent_p1_d2.count(i) / window_size
        features_df.loc[idx, f'3d_d3_digit_{i}_freq_prev_window'] = recent_p1_d3.count(i) / window_size

        # 2-Digit Bottom
        recent_b2_d1 = df['Bottom2_D1'].iloc[idx-window_size:idx].tolist()
        recent_b2_d2 = df['Bottom2_D2'].iloc[idx-window_size:idx].tolist()

        features_df.loc[idx, f'2d_d1_digit_{i}_freq_prev_window'] = recent_b2_d1.count(i) / window_size
        features_df.loc[idx, f'2d_d2_digit_{i}_freq_prev_window'] = recent_b2_d2.count(i) / window_size

# --- เพิ่มฟีเจอร์ใหม่: Last Seen / Consecutive Missing for Each Digit ---
# เราจะนับว่าเลขแต่ละตัว (0-9) ออกครั้งสุดท้ายเมื่อไหร่ในตำแหน่งที่ออกไป
for i in range(10): # สำหรับตัวเลข 0-9
    # Initialise last seen to a large number (meaning not seen recently)
    last_seen_p1_d1 = {j: num_rows for j in range(10)} # สำหรับแต่ละหลัก 0-9 ในตำแหน่ง D1
    last_seen_p1_d2 = {j: num_rows for j in range(10)}
    last_seen_p1_d3 = {j: num_rows for j in range(10)}
    last_seen_b2_d1 = {j: num_rows for j in range(10)}
    last_seen_b2_d2 = {j: num_rows for j in range(10)}

    # Initialize consecutive missing for each digit
    consecutive_missing_p1_d1 = {j: 0 for j in range(10)}
    consecutive_missing_p1_d2 = {j: 0 for j in range(10)}
    consecutive_missing_p1_d3 = {j: 0 for j in range(10)}
    consecutive_missing_b2_d1 = {j: 0 for j in range(10)}
    consecutive_missing_b2_d2 = {j: 0 for j in range(10)}

    features_df[f'p1_d1_last_seen_digit_{i}'] = 0
    features_df[f'p1_d2_last_seen_digit_{i}'] = 0
    features_df[f'p1_d3_last_seen_digit_{i}'] = 0
    features_df[f'b2_d1_last_seen_digit_{i}'] = 0
    features_df[f'b2_d2_last_seen_digit_{i}'] = 0

    features_df[f'p1_d1_consec_missing_digit_{i}'] = 0
    features_df[f'p1_d2_consec_missing_digit_{i}'] = 0
    features_df[f'p1_d3_consec_missing_digit_{i}'] = 0
    features_df[f'b2_d1_consec_missing_digit_{i}'] = 0
    features_df[f'b2_d2_consec_missing_digit_{i}'] = 0

    for idx in range(len(df)):
        current_p1_d1 = df.loc[idx, 'Prize1_D1']
        current_p1_d2 = df.loc[idx, 'Prize1_D2']
        current_p1_d3 = df.loc[idx, 'Prize1_D3']
        current_b2_d1 = df.loc[idx, 'Bottom2_D1']
        current_b2_d2 = df.loc[idx, 'Bottom2_D2']

        # Update last_seen and consecutive_missing for next row's features
        for digit_val in range(10):
            # Prize 1 - D1
            if current_p1_d1 == digit_val:
                last_seen_p1_d1[digit_val] = 0
                consecutive_missing_p1_d1[digit_val] = 0
            else:
                last_seen_p1_d1[digit_val] += 1
                consecutive_missing_p1_d1[digit_val] += 1
            features_df.loc[idx, f'p1_d1_last_seen_digit_{digit_val}'] = last_seen_p1_d1[digit_val]
            features_df.loc[idx, f'p1_d1_consec_missing_digit_{digit_val}'] = consecutive_missing_p1_d1[digit_val]

            # Prize 1 - D2
            if current_p1_d2 == digit_val:
                last_seen_p1_d2[digit_val] = 0
                consecutive_missing_p1_d2[digit_val] = 0
            else:
                last_seen_p1_d2[digit_val] += 1
                consecutive_missing_p1_d2[digit_val] += 1
            features_df.loc[idx, f'p1_d2_last_seen_digit_{digit_val}'] = last_seen_p1_d2[digit_val]
            features_df.loc[idx, f'p1_d2_consec_missing_digit_{digit_val}'] = consecutive_missing_p1_d2[digit_val]

            # Prize 1 - D3
            if current_p1_d3 == digit_val:
                last_seen_p1_d3[digit_val] = 0
                consecutive_missing_p1_d3[digit_val] = 0
            else:
                last_seen_p1_d3[digit_val] += 1
                consecutive_missing_p1_d3[digit_val] += 1
            features_df.loc[idx, f'p1_d3_last_seen_digit_{digit_val}'] = last_seen_p1_d3[digit_val]
            features_df.loc[idx, f'p1_d3_consec_missing_digit_{digit_val}'] = consecutive_missing_p1_d3[digit_val]

            # Bottom 2 - D1
            if current_b2_d1 == digit_val:
                last_seen_b2_d1[digit_val] = 0
                consecutive_missing_b2_d1[digit_val] = 0
            else:
                last_seen_b2_d1[digit_val] += 1
                consecutive_missing_b2_d1[digit_val] += 1
            features_df.loc[idx, f'b2_d1_last_seen_digit_{digit_val}'] = last_seen_b2_d1[digit_val]
            features_df.loc[idx, f'b2_d1_consec_missing_digit_{digit_val}'] = consecutive_missing_b2_d1[digit_val]

            # Bottom 2 - D2
            if current_b2_d2 == digit_val:
                last_seen_b2_d2[digit_val] = 0
                consecutive_missing_b2_d2[digit_val] = 0
            else:
                last_seen_b2_d2[digit_val] += 1
                consecutive_missing_b2_d2[digit_val] += 1
            features_df.loc[idx, f'b2_d2_last_seen_digit_{digit_val}'] = last_seen_b2_d2[digit_val]
            features_df.loc[idx, f'b2_d2_consec_missing_digit_{digit_val}'] = consecutive_missing_b2_d2[digit_val]


# เติมค่า NaN ที่เกิดจาก rolling window ด้วย 0 หรือวิธีอื่นที่เหมาะสม
features_df = features_df.fillna(0) # ควรทำหลังจากการคำนวณทั้งหมด

# --- 4. การปรับขนาดข้อมูล (Min-Max Scaling) ---
scaler = MinMaxScaler()
scaled_features_array = scaler.fit_transform(features_df)
scaled_features_df = pd.DataFrame(scaled_features_array, columns=features_df.columns, index=features_df.index)

print("\nData scaling complete for model input.")
print(scaled_features_df.head())
print(f"Shape of scaled_features_df: {scaled_features_df.shape}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split # สำหรับแบ่งข้อมูล

# ตรวจสอบว่า df และ scaled_features_df มีอยู่แล้วหรือไม่
# หากคุณลบเซลล์ที่ 1 ไปด้วย หรือรันไม่ครบ อาจต้องรันเซลล์ที่ 1 ใหม่อีกครั้งก่อน
if 'df' not in locals() or df.empty:
    print("Error: 'df' DataFrame not found. Please ensure Cell 1 (Feature Engineering & Scaling) was run successfully.")
    # หากรันใน Colab และลบไปแล้ว ให้ไปรันเซลล์ที่ 1 ใหม่อีกครั้ง
    raise SystemExit("Exiting: 'df' not found. Please run Cell 1.")

if 'scaled_features_df' not in locals() or scaled_features_df.empty:
    print("Error: 'scaled_features_df' DataFrame not found. Please ensure Cell 1 (Feature Engineering & Scaling) was run successfully.")
    raise SystemExit("Exiting: 'scaled_features_df' not found. Please run Cell 1.")

# --- 1. กำหนดค่าพารามิเตอร์ ---
# n_steps: จำนวนงวดย้อนหลังที่ใช้เป็น Input (window size)
# ต้องตรงกับ window_size ที่ใช้ในเซลล์ 1 (ตอนนี้เราตั้งเป็น 10)
n_steps = 10
# n_features: จำนวนฟีเจอร์ที่คุณมี (จะถูกกำหนดอัตโนมัติจาก scaled_features_df.shape[1])
n_features = scaled_features_df.shape[1]

print(f"Parameters set: n_steps = {n_steps}, n_features = {n_features}")

# --- 2. สร้าง Time Series Sequences (X_lstm) ---
def create_sequences(data, n_steps):
    X_seq = []
    # Loop ตั้งแต่ index ที่ n_steps-1 จนถึงสุดท้ายของข้อมูล
    # เพื่อให้แต่ละ sequence มี n_steps ตัวอย่าง และมี target สำหรับตัวอย่างถัดไป
    for i in range(len(data) - n_steps): # เปลี่ยนจาก len(data) - n_steps + 1 เป็น len(data) - n_steps
        X_seq.append(data[i:(i + n_steps)])
    return np.array(X_seq)

# แปลง scaled_features_df เป็น NumPy array เพื่อสร้าง sequences
X_features_array = scaled_features_df.values
X_lstm = create_sequences(X_features_array, n_steps)


print(f"LSTM sequences created with sequence_length = {n_steps}.")
print(f"Shape of X_lstm: {X_lstm.shape}") # (จำนวนตัวอย่าง, n_steps, n_features)

# --- 3. เตรียม Target Variables (y_lstm) สำหรับแต่ละหลัก ---
# Target สำหรับ Prize 1 (3D) - D1, D2, D3
# Target สำหรับ Bottom 2 - D1, D2
# ต้องเลื่อน Target ให้สอดคล้องกับ X_lstm
# ถ้า X_lstm[i] คือข้อมูลตั้งแต่ index i ถึง i+n_steps-1
# Target คือผลของงวดที่ i+n_steps
y_prize1_3d_d1_lstm = df['Prize1_D1'].values[n_steps:]
y_prize1_3d_d2_lstm = df['Prize1_D2'].values[n_steps:]
y_prize1_3d_d3_lstm = df['Prize1_D3'].values[n_steps:]

y_bottom2_d1_lstm = df['Bottom2_D1'].values[n_steps:]
y_bottom2_d2_lstm = df['Bottom2_D2'].values[n_steps:]

print(f"Shape of y_prize1_3d_d1_lstm: {y_prize1_3d_d1_lstm.shape}")
print(f"Shape of y_bottom2_d1_lstm: {y_bottom2_d1_lstm.shape}")

# ตรวจสอบให้แน่ใจว่า X_lstm และ y_lstm มีจำนวนตัวอย่างเท่ากัน
if X_lstm.shape[0] != y_prize1_3d_d1_lstm.shape[0]:
    print(f"Warning: X_lstm samples ({X_lstm.shape[0]}) and y_lstm samples ({y_prize1_3d_d1_lstm.shape[0]}) do not match. Adjusting...")
    # ปรับขนาดให้เท่ากัน โดยยึดตามจำนวนตัวอย่างที่น้อยกว่า
    min_samples = min(X_lstm.shape[0], y_prize1_3d_d1_lstm.shape[0])
    X_lstm = X_lstm[:min_samples]
    y_prize1_3d_d1_lstm = y_prize1_3d_d1_lstm[:min_samples]
    y_prize1_3d_d2_lstm = y_prize1_3d_d2_lstm[:min_samples]
    y_prize1_3d_d3_lstm = y_prize1_3d_d3_lstm[:min_samples]
    y_bottom2_d1_lstm = y_bottom2_d1_lstm[:min_samples]
    y_bottom2_d2_lstm = y_bottom2_d2_lstm[:min_samples]
    print(f"Adjusted shapes: X_lstm {X_lstm.shape}, y_lstm {y_prize1_3d_d1_lstm.shape}")


# --- 4. แบ่งข้อมูลเป็น Train, Validation, Test Sets ---
# สำหรับ Time Series ควรแบ่งโดยใช้ train_test_split แบบ shuffle=False เพื่อรักษาลำดับเวลา
test_size = 0.15 # 15% สำหรับ Test Set
val_size = 0.15  # 15% สำหรับ Validation Set (จากข้อมูลที่เหลือหลังหัก Test)

# แบ่งข้อมูลหลัก (X_lstm, y_...) เป็น Train/Val และ Test
X_train_val, X_test, y_train_val_d1, y_test_d1 = train_test_split(
    X_lstm, y_prize1_3d_d1_lstm, test_size=test_size, shuffle=False
)
_, _, y_train_val_d2, y_test_d2 = train_test_split(
    y_prize1_3d_d2_lstm, y_prize1_3d_d2_lstm, test_size=test_size, shuffle=False # ใช้ y ตัวเองเพื่อให้ได้ส่วนแบ่งเท่ากัน
)
_, _, y_train_val_d3, y_test_d3 = train_test_split(
    y_prize1_3d_d3_lstm, y_prize1_3d_d3_lstm, test_size=test_size, shuffle=False
)
_, _, y_train_val_b1, y_test_b1 = train_test_split(
    y_bottom2_d1_lstm, y_bottom2_d1_lstm, test_size=test_size, shuffle=False
)
_, _, y_train_val_b2, y_test_b2 = train_test_split(
    y_bottom2_d2_lstm, y_bottom2_d2_lstm, test_size=test_size, shuffle=False
)

# แบ่ง Train/Val อีกครั้งจาก X_train_val
# คำนวณอัตราส่วนของ Validation ในชุด Train/Val
val_ratio_in_train_val = val_size / (1 - test_size)

X_train, X_val, y_train_d1, y_val_d1 = train_test_split(
    X_train_val, y_train_val_d1, test_size=val_ratio_in_train_val, shuffle=False
)
_, _, y_train_d2, y_val_d2 = train_test_split(
    y_train_val_d2, y_train_val_d2, test_size=val_ratio_in_train_val, shuffle=False
)
_, _, y_train_d3, y_val_d3 = train_test_split(
    y_train_val_d3, y_train_val_d3, test_size=val_ratio_in_train_val, shuffle=False
)
_, _, y_train_b1, y_val_b1 = train_test_split(
    y_train_val_b1, y_train_val_b1, test_size=val_ratio_in_train_val, shuffle=False
)
_, _, y_train_b2, y_val_b2 = train_test_split(
    y_train_val_b2, y_train_val_b2, test_size=val_ratio_in_train_val, shuffle=False
)


print(f"\nData split into Train, Validation, Test:")
print(f"X_train shape: {X_train.shape}, y_train_d1 shape: {y_train_d1.shape}")
print(f"X_val shape: {X_val.shape}, y_val_d1 shape: {y_val_d1.shape}")
print(f"X_test shape: {X_test.shape}, y_test_d1 shape: {y_test_d1.shape}")

# --- 5. เตรียมข้อมูลสำหรับทำนายงวดถัดไป (X_new_scaled) ---
# X_new_scaled คือ sequence n_steps ล่าสุดจาก scaled_features_df
# ใช้ .iloc[-n_steps:] เพื่อให้ได้ข้อมูล n_steps สุดท้ายที่ถูกต้อง
X_new_scaled = scaled_features_df.iloc[-n_steps:].values.reshape(1, n_steps, n_features)
print(f"\nShape of X_new_scaled for next prediction: {X_new_scaled.shape}")

print("\nCell 2 (Data Preprocessing) completed successfully!")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers # Import regularizers
import matplotlib.pyplot as plt

# --- 1. ฟังก์ชันสำหรับสร้างและฝึกโมเดล LSTM (ปรับปรุงเพื่อควบคุม Overfitting) ---
def build_and_train_lstm_model(X_train, y_train, X_val, y_val, n_steps, n_features, y_test,
                                output_classes=10, epochs=200, batch_size=32, model_name="",
                                lstm_units_1=128, lstm_units_2=64, dense_units=32, # Reduced default units
                                dropout_rate_lstm=0.5, dropout_rate_dense=0.4, # Increased default dropout
                                learning_rate=0.0001, patience=30, use_bidirectional=True): # Adjusted defaults for learning rate, patience, and bidirectional
    model = Sequential()

    # Add L2 regularization to LSTM layers
    if use_bidirectional:
        # แก้ไข: ย้าย input_shape เข้าไปใน Bidirectional() โดยตรง
        model.add(Bidirectional(LSTM(lstm_units_1, activation='relu', return_sequences=True,
                                     kernel_regularizer=regularizers.l2(0.001)),
                                input_shape=(n_steps, n_features))) # <<-- แก้ไขตรงนี้!
    else:
        # หากไม่ใช้ Bidirectional, input_shape จะอยู่ใน LSTM โดยตรง
        model.add(LSTM(lstm_units_1, activation='relu', input_shape=(n_steps, n_features),
                       return_sequences=True, kernel_regularizer=regularizers.l2(0.001)))
    model.add(Dropout(dropout_rate_lstm))

    model.add(LSTM(lstm_units_2, activation='relu', kernel_regularizer=regularizers.l2(0.001))) # L2 Regularization
    model.add(Dropout(dropout_rate_lstm))

    # Add L2 regularization to Dense layer
    model.add(Dense(dense_units, activation='relu', kernel_regularizer=regularizers.l2(0.001))) # L2 Regularization
    model.add(Dropout(dropout_rate_dense))

    model.add(Dense(output_classes, activation='softmax'))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=patience, # Increased patience slightly to 30
        restore_best_weights=True
    )

    print(f"\n--- Training Model: {model_name} (Input Shape: {X_train.shape}, Output Classes: {output_classes}) ---")
    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping],
        verbose=1
    )

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} - Loss History')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} - Accuracy History')
    plt.legend()
    plt.tight_layout()
    plt.show()

    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
    print(f"  {model_name} - Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")

    return model, history


# --- 2. ฝึกโมเดลสำหรับแต่ละหลัก (ด้วยพารามิเตอร์ที่ปรับปรุง) ---
print("Starting training with enhanced features and tuned hyperparameters (to combat overfitting)...")

# Prize 1 (3D) Models: Reduced units, increased dropout, L2 regularization, lower learning rate
model_prize1_d1, hist_p1d1 = build_and_train_lstm_model(X_train, y_train_d1, X_val, y_val_d1, n_steps, n_features, y_test=y_test_d1,
                                                        model_name="Prize 1 - Digit 1 (Hundred)",
                                                        lstm_units_1=128, lstm_units_2=64, dense_units=32, # Reduced units
                                                        dropout_rate_lstm=0.4, dropout_rate_dense=0.3, # Increased dropout
                                                        learning_rate=0.0001, batch_size=32, use_bidirectional=True) # Lower LR, Bidirectional by default

model_prize1_d2, hist_p1d2 = build_and_train_lstm_model(X_train, y_train_d2, X_val, y_val_d2, n_steps, n_features, y_test=y_test_d2,
                                                        model_name="Prize 1 - Digit 2 (Ten)",
                                                        lstm_units_1=128, lstm_units_2=64, dense_units=32,
                                                        dropout_rate_lstm=0.4, dropout_rate_dense=0.3,
                                                        learning_rate=0.0001, batch_size=32, use_bidirectional=True)

model_prize1_d3, hist_p1d3 = build_and_train_lstm_model(X_train, y_train_d3, X_val, y_val_d3, n_steps, n_features, y_test=y_test_d3,
                                                        model_name="Prize 1 - Digit 3 (Unit)",
                                                        lstm_units_1=128, lstm_units_2=64, dense_units=32,
                                                        dropout_rate_lstm=0.4, dropout_rate_dense=0.3,
                                                        learning_rate=0.0001, batch_size=32, use_bidirectional=True)

# 2-Digit Bottom Models: Further reduced units, increased dropout, L2 regularization, lower learning rate
model_bottom2_d1, hist_b2d1 = build_and_train_lstm_model(X_train, y_train_b1, X_val, y_val_b1, n_steps, n_features, y_test=y_test_b1,
                                                          model_name="2-Digit Bottom - Digit 1",
                                                          lstm_units_1=64, lstm_units_2=32, dense_units=16, # Further reduced units for 2D
                                                          dropout_rate_lstm=0.4, dropout_rate_dense=0.3,
                                                          learning_rate=0.0001, batch_size=32, use_bidirectional=True)

model_bottom2_d2, hist_b2d2 = build_and_train_lstm_model(X_train, y_train_b2, X_val, y_val_b2, n_steps, n_features, y_test=y_test_b2,
                                                          model_name="2-Digit Bottom - Digit 2",
                                                          lstm_units_1=64, lstm_units_2=32, dense_units=16,
                                                          dropout_rate_lstm=0.4, dropout_rate_dense=0.3,
                                                          learning_rate=0.0001, batch_size=32, use_bidirectional=True)
```

```python
import numpy as np

# --- 1. ฟังก์ชันสำหรับทำนาย Top-K ---
def get_top_k_predictions(model, input_data, k=5):
    """
    ทำนายและคืนค่า K อันดับแรกของตัวเลขที่มีความน่าจะเป็นสูงสุดพร้อมเปอร์เซ็นต์
    """
    # ตรวจสอบ input_data.shape ให้เป็น (1, n_steps, n_features)
    if input_data.ndim == 2:
        input_data = input_data.reshape(1, input_data.shape[0], input_data.shape[1])
    elif input_data.ndim == 1:
        # หาก input เป็น 1D, ควรระบุ n_steps และ n_features ให้ถูกต้อง
        # ในบริบทนี้ X_new_scaled ควรเป็น (1, n_steps, n_features) อยู่แล้ว
        pass # ไม่ต้องทำอะไร ถ้า X_new_scaled ถูกต้อง

    probabilities = model.predict(input_data, verbose=0)[0]
    top_k_indices = np.argsort(probabilities)[-k:][::-1]
    top_k_probs = probabilities[top_k_indices]
    return top_k_indices, top_k_probs

# --- 2. กำหนดค่า K สำหรับ Top-K Predictions ---
k_value = 5 # จำนวนตัวเลือก Top-K ที่ต้องการ

print("\n--- Predicting Top-K Digits for the Next Draw ---")

# สำหรับ Prize 1 (3D)
top_k_p1_d1_indices, top_k_p1_d1_probs = get_top_k_predictions(model_prize1_d1, X_new_scaled, k=k_value)
top_k_p1_d2_indices, top_k_p1_d2_probs = get_top_k_predictions(model_prize1_d2, X_new_scaled, k=k_value)
top_k_p1_d3_indices, top_k_p1_d3_probs = get_top_k_predictions(model_prize1_d3, X_new_scaled, k=k_value)

print(f"Prize 1 (Hundred) - Top {k_value}: {top_k_p1_d1_indices} (Probs: {[f'{p*100:.2f}%' for p in top_k_p1_d1_probs]})")
print(f"Prize 1 (Ten) - Top {k_value}: {top_k_p1_d2_indices} (Probs: {[f'{p*100:.2f}%' for p in top_k_p1_d2_probs]})")
print(f"Prize 1 (Unit) - Top {k_value}: {top_k_p1_d3_indices} (Probs: {[f'{p*100:.2f}%' for p in top_k_p1_d3_probs]})")

# สำหรับ 2-Digit Bottom
top_k_b2_d1_indices, top_k_b2_d1_probs = get_top_k_predictions(model_bottom2_d1, X_new_scaled, k=k_value)
top_k_b2_d2_indices, top_k_b2_d2_probs = get_top_k_predictions(model_bottom2_d2, X_new_scaled, k=k_value)

print(f"\n2-Digit Bottom (Digit 1) - Top {k_value}: {top_k_b2_d1_indices} (Probs: {[f'{p*100:.2f}%' for p in top_k_b2_d1_probs]})")
print(f"2-Digit Bottom (Digit 2) - Top {k_value}: {top_k_b2_d2_indices} (Probs: {[f'{p*100:.2f}%' for p in top_k_b2_d2_probs]})")


# --- 3. ฟังก์ชันสำหรับรวม Top-K และคำนวณคะแนนรวม ---
def get_all_possible_combinations_with_scores(digit_indices_list, digit_probs_list, prediction_type="3D"):
    """
    สร้างชุดตัวเลขที่เป็นไปได้ทั้งหมดจาก Top-K ของแต่ละหลัก
    พร้อมคำนวณคะแนนรวมจากความน่าจะเป็นของแต่ละหลักในชุดนั้น
    """
    all_combinations_with_scores = []

    # สร้าง Dictionary สำหรับการค้นหา Probability ได้ง่ายขึ้น
    probs_map_list = []
    for indices, probs in zip(digit_indices_list, digit_probs_list):
        probs_map_list.append(dict(zip(indices, probs)))

    if prediction_type == "3D":
        for d1_idx in digit_indices_list[0]: # top_k_p1_d1_indices
            for d2_idx in digit_indices_list[1]: # top_k_p1_d2_indices
                for d3_idx in digit_indices_list[2]: # top_k_p1_d3_indices
                    number_str = f"{d1_idx}{d2_idx}{d3_idx}"
                    # ใช้ผลรวมของความน่าจะเป็น (หรือจะลองคูณก็ได้หากต้องการเปรียบเทียบ)
                    total_prob_score = (probs_map_list[0].get(d1_idx, 0) +
                                        probs_map_list[1].get(d2_idx, 0) +
                                        probs_map_list[2].get(d3_idx, 0))
                    all_combinations_with_scores.append((number_str, total_prob_score))
    elif prediction_type == "2D":
        for d1_idx in digit_indices_list[0]: # top_k_b2_d1_indices
            for d2_idx in digit_indices_list[1]: # top_k_b2_d2_indices
                number_str = f"{d1_idx}{d2_idx}"
                total_prob_score = (probs_map_list[0].get(d1_idx, 0) +
                                    probs_map_list[1].get(d2_idx, 0))
                all_combinations_with_scores.append((number_str, total_prob_score))
    else:
        raise ValueError("prediction_type ต้องเป็น '3D' หรือ '2D'")

    return all_combinations_with_scores

# --- 4. สร้างชุดตัวเลขที่เป็นไปได้ทั้งหมดพร้อมคะแนน ---
all_possible_prize1_combinations = get_all_possible_combinations_with_scores(
    [top_k_p1_d1_indices, top_k_p1_d2_indices, top_k_p1_d3_indices],
    [top_k_p1_d1_probs, top_k_p1_d2_probs, top_k_p1_d3_probs],
    prediction_type="3D"
)
print(f"\nGenerated {len(all_possible_prize1_combinations)} Possible Prize 1 (3D) numbers with scores.")

all_possible_bottom2_combinations = get_all_possible_combinations_with_scores(
    [top_k_b2_d1_indices, top_k_b2_d2_indices],
    [top_k_b2_d1_probs, top_k_b2_d2_probs],
    prediction_type="2D"
)
print(f"Generated {len(all_possible_bottom2_combinations)} Possible 2-Digit Bottom numbers with scores.")
```

```python
import pandas as pd

# --- สมมติว่าคุณมีตัวแปรเหล่านี้จากเซลล์ก่อนหน้า ---
# all_possible_prize1_combinations (รายการของ (เลข, คะแนน) สำหรับ Prize 1)
# all_possible_bottom2_combinations (รายการของ (เลข, คะแนน) สำหรับ 2-Digit Bottom)

# --- คุณต้องมี DataFrame หลักของข้อมูลรางวัล (df_original_data) ---
# ที่มีคอลัมน์ของรางวัลที่ 1 และเลขล่าง 2 หลัก ในรูปแบบตัวเลขหรือ string ที่เปรียบเทียบได้
# สมมติว่าคอลัมน์ชื่อ 'Prize1_Full' (เช่น '123') และ 'Bottom2_Full' (เช่น '45')
# และมีข้อมูลการออกรางวัลเรียงตามลำดับเวลา (งวดล่าสุดอยู่ล่างสุด)

# *** สำคัญมาก: แทนที่ด้วยการโหลดหรือเข้าถึง DataFrame ข้อมูลจริงของคุณ ***
# ตัวอย่างการสร้าง DataFrame ข้อมูลหลอก (Mock Data) สำหรับสาธิต
# เพื่อให้รันโค้ดได้ครบ
# ในการใช้งานจริง: df_original_data = pd.read_csv('your_cleaned_lottery_data.csv')
# ต้องมีคอลัมน์ 'Prize1_Full' และ 'Bottom2_Full' ที่เป็น string
data_for_history_size = 550
np.random.seed(42)
mock_prize1_full = [str(np.random.randint(0, 10)) + str(np.random.randint(0, 10)) + str(np.random.randint(0, 10)) for _ in range(data_for_history_size)]
mock_bottom2_full = [str(np.random.randint(0, 10)) + str(np.random.randint(0, 10)) for _ in range(data_for_history_size)]
df_original_data = pd.DataFrame({
    'Prize1_Full': mock_prize1_full,
    'Bottom2_Full': mock_bottom2_full
})
# เพิ่มเลขที่ใกล้เคียงคำทำนายเพื่อสาธิตว่าเคยออก
df_original_data.loc[540, 'Prize1_Full'] = '086' # ตัวอย่างเลขที่เคยออก
df_original_data.loc[530, 'Prize1_Full'] = '050'
df_original_data.loc[545, 'Bottom2_Full'] = '80'
df_original_data.loc[535, 'Bottom2_Full'] = '50'

# --- 1. ฟังก์ชันสำหรับตรวจสอบประวัติและวิเคราะห์แพทเทิร์นตัวเลขเดี่ยว ---

def analyze_and_filter_predictions(all_combinations_with_scores, historical_df, prize_column_name, num_recent_rows=8, num_to_select=5):
    """
    วิเคราะห์แต่ละชุดตัวเลขทำนายว่าเคยออกหรือไม่, ระยะห่าง,
    และตรวจสอบว่าหลักแต่ละตัวในเลขทำนายปรากฏในงวดล่าสุดหรือไม่
    จากนั้นคัดเลือกชุดที่ดีที่สุดตามคะแนนรวม

    Args:
        all_combinations_with_scores (list): รายการของ tuple (ชุดตัวเลข, คะแนนรวม)
        historical_df (pd.DataFrame): DataFrame ที่มีประวัติการออกรางวัลทั้งหมด
        prize_column_name (str): ชื่อคอลัมน์ใน historical_df ที่เก็บเลขรางวัล
        num_recent_rows (int): จำนวนแถวล่าสุดที่ใช้ในการตรวจสอบ "แพทเทิร์นตัวเลขเดี่ยว"
        num_to_select (int): จำนวนชุดทำนายที่ดีที่สุดที่ต้องการ

    Returns:
        list: รายการของ dict ที่มีรายละเอียดการวิเคราะห์สำหรับชุดทำนายที่ดีที่สุด
    """
    analyzed_predictions = []

    # ดึงข้อมูลประวัติการออกรางวัลสำหรับคอลัมน์ที่ระบุ
    prize_history = historical_df[prize_column_name].tolist()

    # ดึงตัวเลขทั้งหมดจาก N แถวล่าสุด (เป็น string เช่น '392', '06')
    recent_history_numbers_str = historical_df[prize_column_name].tail(num_recent_rows).tolist()

    # รวบรวมหลัก "ทั้งหมด" ที่ปรากฏใน N งวดย้อนหลัง (ในทุกตำแหน่ง)
    # เช่น ถ้า 392, 06 -> จะได้ {0, 2, 3, 6, 9}
    all_unique_digits_in_recent_history = set()
    for num_str in recent_history_numbers_str:
        for digit_char in num_str:
            all_unique_digits_in_recent_history.add(digit_char)

    for number_str, score in all_combinations_with_scores:
        found_at_index = -1
        distance = -1

        # ตรวจสอบว่าเคยออกหรือไม่ (จากประวัติทั้งหมด)
        if number_str in prize_history:
            last_occurrence_index = len(prize_history) - 1 - prize_history[::-1].index(number_str)
            found_at_index = last_occurrence_index
            distance = (len(prize_history) - 1) - last_occurrence_index

        # ตรวจสอบ "แพทเทิร์นตัวเลขเดี่ยว" (All digits present in last N draws)
        # ตรวจสอบว่า "แต่ละหลัก" ในเลขทำนาย มีอยู่ใน "ชุดตัวเลขเดี่ยวที่ปรากฏใน N งวดย้อนหลัง" หรือไม่
        predicted_digits_chars = set([d for d in number_str]) # แปลงเป็น set เพื่อเปรียบเทียบง่ายขึ้น

        all_predicted_digits_in_recent_history = True
        for digit_char in predicted_digits_chars:
            if digit_char not in all_unique_digits_in_recent_history:
                all_predicted_digits_in_recent_history = False
                break

        analyzed_predictions.append({
            'Number': number_str,
            'Score': score,
            'Ever_Appeared': "Yes" if found_at_index != -1 else "No",
            'Distance_From_Current': distance if found_at_index != -1 else "N/A",
            f'All_Digits_In_Last_{num_recent_rows}_Draws': "Yes" if all_predicted_digits_in_recent_history else "No"
        })

    # จัดเรียงตามคะแนนรวมจากมากไปน้อยก่อนคัดเลือก
    analyzed_predictions.sort(key=lambda x: x['Score'], reverse=True)

    # คัดเลือก num_to_select ชุดที่ดีที่สุด
    return analyzed_predictions[:num_to_select]


# --- 2. วิเคราะห์และคัดเลือก Best 5 สำหรับ Prize 1 (3D) ---
print("\n--- Analysis of Best 5 Prize 1 (3D) Predictions ---")
final_prize1_predictions = analyze_and_filter_predictions(
    all_possible_prize1_combinations,
    df_original_data, # ใช้ DataFrame ข้อมูลประวัติของคุณ
    'Prize1_Full',     # ชื่อคอลัมน์เลขท้ายรางวัลที่ 1
    num_recent_rows=8, # ตรวจสอบ 8 งวดย้อนหลัง
    num_to_select=5    # เลือก 5 ชุดที่ดีที่สุด
)

for data in final_prize1_predictions:
    print(f"Number: {data['Number']} (Score: {data['Score']:.4f})")
    print(f"  Ever Appeared: {data['Ever_Appeared']}")
    if data['Ever_Appeared'] == "Yes":
        print(f"  Distance from Current Draw: {data['Distance_From_Current']} draws ago")
    print(f"  All digits of this number present in last 8 draws: {data['All_Digits_In_Last_8_Draws']}")
    print("-" * 30)


# --- 3. วิเคราะห์และคัดเลือก Best 5 สำหรับ 2-Digit Bottom ---
print("\n--- Analysis of Best 5 2-Digit Bottom Predictions ---")
final_bottom2_predictions = analyze_and_filter_predictions(
    all_possible_bottom2_combinations,
    df_original_data, # ใช้ DataFrame ข้อมูลประวัติของคุณ
    'Bottom2_Full',   # ชื่อคอลัมน์เลขล่าง 2 หลัก
    num_recent_rows=8, # ตรวจสอบ 8 งวดย้อนหลัง
    num_to_select=5    # เลือก 5 ชุดที่ดีที่สุด
)

for data in final_bottom2_predictions:
    print(f"Number: {data['Number']} (Score: {data['Score']:.4f})")
    print(f"  Ever Appeared: {data['Ever_Appeared']}")
    if data['Ever_Appeared'] == "Yes":
        print(f"  Distance from Current Draw: {data['Distance_From_Current']} draws ago")
    print(f"  All digits of this number present in last 8 draws: {data['All_Digits_In_Last_8_Draws']}")
    print("-" * 30)
```
