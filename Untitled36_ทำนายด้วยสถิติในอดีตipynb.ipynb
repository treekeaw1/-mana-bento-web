```python
import pandas as pd
import numpy as np
import re
from datetime import datetime, timedelta

# --- ‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1: ‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡πÅ‡∏•‡∏∞‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• ---

print("--- ‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1: ‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡πÅ‡∏•‡∏∞‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• ---")
print("üîÑ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÇ‡∏´‡∏•‡∏î‡πÅ‡∏•‡∏∞‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å '‡πÑ‡∏ü‡∏•‡πå‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô.csv'...")

file_path = '‡πÑ‡∏ü‡∏•‡πå‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô.csv'
df = None
original_df_rows_count = 0 # ‡∏à‡∏∞‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÅ‡∏ñ‡∏ß‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏î‡∏¥‡∏ö‡∏Å‡πà‡∏≠‡∏ô‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•

try:
    # 1. ‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå CSV ‡πÇ‡∏î‡∏¢‡∏Ç‡πâ‡∏≤‡∏°‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÅ‡∏£‡∏Å (metadata header)
    # ‡πÉ‡∏ä‡πâ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏ó‡∏µ‡πà‡∏™‡∏≠‡∏á (‡∏ã‡∏∂‡πà‡∏á‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÅ‡∏£‡∏Å‡∏´‡∏•‡∏±‡∏á skip) ‡πÄ‡∏õ‡πá‡∏ô Header ‡∏à‡∏£‡∏¥‡∏á‡πÜ
    # ‡πÅ‡∏•‡∏∞‡∏Å‡∏≥‡∏´‡∏ô‡∏î dtype ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô string ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡πâ‡∏≠‡∏á‡∏Å‡∏±‡∏ô .0
    encodings_to_try = ['utf-8', 'tis-620', 'cp874', 'latin1', 'iso-8859-1']

    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏Ñ‡∏≤‡∏î‡∏´‡∏ß‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏†‡∏≤‡∏û‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á
    expected_header_names = ['‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà', '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (6 ‡∏´‡∏•‡∏±‡∏Å)', '‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á']

    for encoding in encodings_to_try:
        try:
            # skiprows=1: ‡∏Ç‡πâ‡∏≤‡∏°‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÅ‡∏£‡∏Å‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô metadata
            # header=0: ‡πÉ‡∏ä‡πâ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÅ‡∏£‡∏Å‡∏´‡∏•‡∏±‡∏á skip (‡∏Ñ‡∏∑‡∏≠‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏ó‡∏µ‡πà‡∏™‡∏≠‡∏á‡∏Ç‡∏≠‡∏á‡πÑ‡∏ü‡∏•‡πå‡πÄ‡∏î‡∏¥‡∏°) ‡πÄ‡∏õ‡πá‡∏ô Header
            # dtype: ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏Ñ‡∏≤‡∏î‡∏ß‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç ‡∏ñ‡∏π‡∏Å‡∏≠‡πà‡∏≤‡∏ô‡πÄ‡∏õ‡πá‡∏ô string ‡∏ó‡∏±‡∏ô‡∏ó‡∏µ
            # (‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏£‡∏π‡πâ‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏à‡∏£‡∏¥‡∏á‡∏ó‡∏µ‡πà Pandas ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏î‡πâ‡πÉ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ ‡πÅ‡∏ï‡πà‡∏£‡∏π‡πâ‡∏ß‡πà‡∏≤‡∏°‡∏±‡∏ô‡∏Ñ‡∏∑‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà 1 ‡πÅ‡∏•‡∏∞ 2)
            # ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏Å‡∏≥‡∏´‡∏ô‡∏î dtype ‡πÄ‡∏õ‡πá‡∏ô 'object' (‡∏ã‡∏∂‡πà‡∏á‡∏Å‡πá‡∏Ñ‡∏∑‡∏≠ string) ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà 1 ‡πÅ‡∏•‡∏∞ 2
            df_temp = pd.read_csv(file_path,
                                  encoding=encoding,
                                  skiprows=1,
                                  header=0,
                                  dtype={1: str, 2: str}) # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà 1 ‡πÅ‡∏•‡∏∞ 2 ‡πÄ‡∏õ‡πá‡∏ô string

            # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡πÄ‡∏û‡∏µ‡∏¢‡∏á‡∏û‡∏≠‡πÅ‡∏•‡∏∞‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å‡πÑ‡∏°‡πà‡∏ß‡πà‡∏≤‡∏á‡πÄ‡∏õ‡∏•‡πà‡∏≤
            if df_temp.shape[1] >= 3 and pd.notna(df_temp.iloc[0, 0]):
                df = df_temp
                print(f"‚úÖ ‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå‡∏î‡πâ‡∏ß‡∏¢ encoding '{encoding}' ‡πÇ‡∏î‡∏¢‡∏Ç‡πâ‡∏≤‡∏° 1 ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÅ‡∏•‡∏∞‡πÉ‡∏ä‡πâ Header ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à")
                break
            else:
                print(f"‚ùå ‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå‡∏î‡πâ‡∏ß‡∏¢ encoding '{encoding}' ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à ‡πÅ‡∏ï‡πà‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏î‡∏π‡πÑ‡∏°‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á (‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÑ‡∏°‡πà‡∏û‡∏≠ ‡∏´‡∏£‡∏∑‡∏≠‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å‡∏ß‡πà‡∏≤‡∏á)")

        except Exception as e:
            # print(f"‚ùå ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå‡∏î‡πâ‡∏ß‡∏¢ encoding '{encoding}': {e}") # ‡∏õ‡∏¥‡∏î log error ‡∏ã‡πâ‡∏≥‡πÜ
            continue

    if df is None:
        raise Exception("‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡πÇ‡∏´‡∏•‡∏î‡πÑ‡∏ü‡∏•‡πå CSV ‡∏î‡πâ‡∏ß‡∏¢ encoding ‡πÉ‡∏î‡πÜ ‡∏ó‡∏µ‡πà‡∏•‡∏≠‡∏á‡πÑ‡∏î‡πâ ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏ü‡∏•‡πå‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á")

    # 2. ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô ‡πÅ‡∏•‡∏∞‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå
    if len(df.columns) < 3:
        raise ValueError(f"‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏û‡∏ö ({len(df.columns)}) ‡πÑ‡∏°‡πà‡∏û‡∏≠‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà, ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1, ‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á (‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 3 ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå)")

    # ‡πÉ‡∏ä‡πâ df.columns[0], df.columns[1], df.columns[2] ‡∏ã‡∏∂‡πà‡∏á‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà Pandas ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏î‡πâ (‡πÅ‡∏°‡πâ‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏≠‡∏±‡∏Å‡∏Ç‡∏£‡∏∞‡πÅ‡∏õ‡∏•‡∏Å‡πÜ)
    # ‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡∏Ç‡∏≠‡∏á‡πÄ‡∏£‡∏≤‡∏ó‡∏±‡∏ô‡∏ó‡∏µ
    df_selected = df.iloc[:, [0, 1, 2]].copy() # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÅ‡∏Ñ‡πà 3 ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÅ‡∏£‡∏Å
    df_selected.columns = ['Date_Raw', 'Prize1_Raw', 'Last2Digits_Raw']

    print(f"\n‚úÖ ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå '{df.columns[0]}' ‡πÄ‡∏õ‡πá‡∏ô ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà")
    print(f"‚úÖ ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå '{df.columns[1]}' ‡πÄ‡∏õ‡πá‡∏ô ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (6 ‡∏´‡∏•‡∏±‡∏Å)")
    print(f"‚úÖ ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå '{df.columns[2]}' ‡πÄ‡∏õ‡πá‡∏ô ‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á")

    # 3. ‡∏•‡∏ö‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô footer (‡∏ñ‡πâ‡∏≤‡∏°‡∏µ)
    last_row_date_val = str(df_selected.iloc[-1]['Date_Raw']).strip()
    if "artifact" in last_row_date_val.lower():
        print(f"‚ÑπÔ∏è ‡∏ï‡∏£‡∏ß‡∏à‡∏û‡∏ö‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢ ('{last_row_date_val[:50]}...') ‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô Footer/Metadata ‡∏à‡∏∞‡∏•‡∏ö‡∏≠‡∏≠‡∏Å")
        df_selected = df_selected.iloc[:-1].copy().reset_index(drop=True)

    original_df_rows_count = len(df_selected) # ‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÅ‡∏ñ‡∏ß‡πÉ‡∏ô df_selected ‡∏Å‡πà‡∏≠‡∏ô‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î

    # 4. ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏ï‡πà‡∏•‡∏∞‡πÅ‡∏ñ‡∏ß (‡∏õ‡∏£‡∏±‡∏ö‡∏õ‡∏£‡∏∏‡∏á‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£ NaN, '-' ‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏¥‡∏° 0)
    cleaned_data = []
    seen_ids = set() # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡πâ‡∏≥

    for index, row in df_selected.iterrows():
        try:
            # 4.1 ‡πÅ‡∏õ‡∏•‡∏á‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà (‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏≤‡∏¢‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡πÅ‡∏•‡∏∞‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ñ‡πà‡∏≤‡∏ß‡πà‡∏≤‡∏á)
            date_raw = str(row['Date_Raw']).strip()
            if not date_raw or date_raw.lower() == 'nan' or date_raw == '-':
                continue

            date_obj = None
            date_formats = [
                '%Y/%m/%d', '%Y-%m-%d', '%d/%m/%Y', '%d-%m-%Y',
                '%d/%m/%y', '%d-%m-%y', # DD/MM/YY
                '%d %b %Y', '%d %B %Y', # DD MonYYYY (e.g., 1 ‡∏°.‡∏Ñ. 2565)
            ]

            # ‡∏Å‡∏≤‡∏£‡πÅ‡∏õ‡∏•‡∏á‡∏õ‡∏µ ‡∏û.‡∏®. ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô ‡∏Ñ.‡∏®. (‡∏ñ‡πâ‡∏≤‡∏û‡∏ö)
            if re.search(r'\b25\d{2}\b', date_raw):
                date_raw_parts = date_raw.split()
                for i, part in enumerate(date_raw_parts):
                    if part.isdigit() and len(part) == 4 and part.startswith('25'):
                        date_raw_parts[i] = str(int(part) - 543)
                date_raw = ' '.join(date_raw_parts)

            for fmt in date_formats:
                try:
                    date_obj = datetime.strptime(date_raw, fmt)
                    break
                except ValueError:
                    continue

            if date_obj is None:
                print(f"‚ö†Ô∏è ‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà {index} (‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏î‡∏¥‡∏ö‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà '{row['Date_Raw']}'): ‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö ‡∏Ç‡πâ‡∏≤‡∏°‡πÅ‡∏ñ‡∏ß‡∏ô‡∏µ‡πâ")
                continue

            # 4.2 ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î Prize1_Raw (6 ‡∏´‡∏•‡∏±‡∏Å) ‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏¥‡∏° 0
            prize1_raw = str(row['Prize1_Raw']).strip()
            if not prize1_raw or prize1_raw.lower() == 'nan' or prize1_raw == '-':
                continue

            prize1_cleaned = re.sub(r'[^\d]', '', prize1_raw) # ‡∏•‡∏ö‡∏≠‡∏±‡∏Å‡∏Ç‡∏£‡∏∞‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç

            # *** ‡∏à‡∏∏‡∏î‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç: ‡πÄ‡∏ï‡∏¥‡∏° 0 ‡∏î‡πâ‡∏≤‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡πÉ‡∏´‡πâ‡∏Ñ‡∏£‡∏ö 6 ‡∏´‡∏•‡∏±‡∏Å ***
            prize1_cleaned = prize1_cleaned.zfill(6)

            if len(prize1_cleaned) != 6:
                # ‡∏ñ‡πâ‡∏≤‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å zfill ‡πÅ‡∏•‡πâ‡∏ß‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà 6 ‡∏´‡∏•‡∏±‡∏Å ‡πÅ‡∏™‡∏î‡∏á‡∏ß‡πà‡∏≤‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥‡∏°‡∏≤‡∏Å
                print(f"‚ö†Ô∏è ‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà {index} (Prize1_Raw '{row['Prize1_Raw']}'): ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡πÑ‡∏°‡πà‡∏Ñ‡∏£‡∏ö 6 ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏•‡∏±‡∏á‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏¥‡∏° 0 ({len(prize1_cleaned)} ‡∏´‡∏•‡∏±‡∏Å) ‡∏Ç‡πâ‡∏≤‡∏°‡πÅ‡∏ñ‡∏ß‡∏ô‡∏µ‡πâ")
                continue

            digits = [int(d) for d in prize1_cleaned]

            # 4.3 ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î Last2Digits_Raw (2 ‡∏´‡∏•‡∏±‡∏Å) ‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏¥‡∏° 0
            last2_cleaned = None
            last2_raw = str(row['Last2Digits_Raw']).strip()

            if not last2_raw or last2_raw.lower() == 'nan' or last2_raw == '-':
                last2_cleaned = prize1_cleaned[-2:] # ‡πÉ‡∏ä‡πâ 2 ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢‡∏Ç‡∏≠‡∏á Prize1_Raw ‡∏´‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ß‡πà‡∏≤‡∏á
            else:
                last2_cleaned = re.sub(r'[^\d]', '', last2_raw)

            # *** ‡∏à‡∏∏‡∏î‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç: ‡πÄ‡∏ï‡∏¥‡∏° 0 ‡∏î‡πâ‡∏≤‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡πÉ‡∏´‡πâ‡∏Ñ‡∏£‡∏ö 2 ‡∏´‡∏•‡∏±‡∏Å ***
            last2_cleaned = last2_cleaned.zfill(2)

            if len(last2_cleaned) != 2:
                # ‡∏ñ‡πâ‡∏≤‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å zfill ‡πÅ‡∏•‡πâ‡∏ß‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà 2 ‡∏´‡∏•‡∏±‡∏Å ‡πÅ‡∏™‡∏î‡∏á‡∏ß‡πà‡∏≤‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥‡∏°‡∏≤‡∏Å
                print(f"‚ö†Ô∏è ‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà {index} (Last2Digits_Raw '{row.get('Last2Digits_Raw', 'N/A')}'): ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á‡πÑ‡∏°‡πà‡∏Ñ‡∏£‡∏ö 2 ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏•‡∏±‡∏á‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏¥‡∏° 0 ‡∏Ç‡πâ‡∏≤‡∏°‡πÅ‡∏ñ‡∏ß‡∏ô‡∏µ‡πâ")
                continue

            # ‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡πâ‡∏≥‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà + Prize1 ‡πÄ‡∏õ‡πá‡∏ô ID
            row_id = f"{date_obj.isoformat()}_{prize1_cleaned}"
            if row_id in seen_ids:
                print(f"‚ö†Ô∏è ‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà {index}: ‡∏û‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡πâ‡∏≥ (‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà: {date_obj.isoformat()}, Prize1: {prize1_cleaned}) ‡∏Ç‡πâ‡∏≤‡∏°‡πÅ‡∏ñ‡∏ß‡∏ô‡∏µ‡πâ")
                continue
            seen_ids.add(row_id)

            cleaned_data.append({
                'id': row_id,
                'date': date_obj,
                'dateStr': date_obj.strftime('%Y-%m-%d'),
                'prize1': prize1_cleaned,
                'last2': last2_cleaned,
                'digits': digits,
                'sum': sum(digits),
                'evenCount': sum(1 for d in digits if d % 2 == 0),
                'oddCount': sum(1 for d in digits if d % 2 == 1),
                'last2Int': int(last2_cleaned)
            })

        except Exception as e:
            print(f"‚ùå ‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏Ñ‡∏≤‡∏î‡∏Ñ‡∏¥‡∏î‡πÉ‡∏ô‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà {index}: {e}. ‡∏Ç‡πâ‡∏≤‡∏°‡πÅ‡∏ñ‡∏ß‡∏ô‡∏µ‡πâ")
            continue

    # 5. ‡∏à‡∏±‡∏î‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡∏≤‡∏°‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà
    processed_data = sorted(cleaned_data, key=lambda x: x['date'])

    if not processed_data:
        raise Exception("‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î ‡πÇ‡∏õ‡∏£‡∏î‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå CSV ‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì")

    print(f"\n‚úÖ ‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡πÅ‡∏•‡∏∞‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à: ‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÑ‡∏î‡πâ {len(processed_data)} ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£ (‡∏à‡∏≤‡∏Å {original_df_rows_count} ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏°‡∏≤)")
    print(f"‡∏ä‡πà‡∏ß‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•: ‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏ï‡πà {processed_data[0]['dateStr']} ‡∏ñ‡∏∂‡∏á {processed_data[-1]['dateStr']}")

    _processed_data_cell1 = processed_data
    _original_df_rows_cell1 = original_df_rows_count

except FileNotFoundError:
    print(f"‚ùå ‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î: ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå '{file_path}' ‡πÇ‡∏õ‡∏£‡∏î‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ä‡∏∑‡πà‡∏≠‡πÅ‡∏•‡∏∞‡∏ó‡∏µ‡πà‡∏ï‡∏±‡πâ‡∏á‡∏Ç‡∏≠‡∏á‡πÑ‡∏ü‡∏•‡πå")
except ValueError as ve:
    print(f"‚ùå ‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•: {ve}")
except Exception as e:
    print(f"‚ùå ‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î‡∏ó‡∏±‡πà‡∏ß‡πÑ‡∏õ: {e}")

```

```python
import pandas as pd
import numpy as np
import re
from datetime import datetime, timedelta

# --- ‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 2: ‡∏£‡∏∞‡∏ö‡∏ö‡∏´‡∏•‡∏±‡∏Å ---

print("\n--- ‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 2: ‡∏£‡∏∞‡∏ö‡∏ö‡∏´‡∏•‡∏±‡∏Å (Feature Engineering ‡πÅ‡∏•‡∏∞ Model Logic) ---")

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≤‡∏Å‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1 ‡∏°‡∏µ‡∏≠‡∏¢‡∏π‡πà
if '_processed_data_cell1' not in globals() or not _processed_data_cell1:
    print("‚ùå ‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÅ‡∏•‡πâ‡∏ß‡∏à‡∏≤‡∏Å‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1 ‡∏´‡∏£‡∏∑‡∏≠‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ß‡πà‡∏≤‡∏á‡πÄ‡∏õ‡∏•‡πà‡∏≤ ‡πÇ‡∏õ‡∏£‡∏î‡∏£‡∏±‡∏ô‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1 ‡∏Å‡πà‡∏≠‡∏ô")
else:
    processed_data = _processed_data_cell1

    def calculate_std_dev(arr):
        """Calculates the standard deviation of a list of numbers."""
        if len(arr) < 2:
            return 0.0
        mean = np.mean(arr)
        return np.std(arr, ddof=1) # ddof=1 for sample standard deviation

    def extract_sequential_features(data):
        """
        ‡∏™‡∏£‡πâ‡∏≤‡∏á Features ‡πÄ‡∏ä‡∏¥‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÅ‡∏•‡πâ‡∏ß
        """
        print("üîÑ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á Sequential Features...")
        features_list = []

        for index, item in enumerate(data):
            features = item.copy() # ‡∏Ñ‡∏±‡∏î‡∏•‡∏≠‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô

            # Dynamic lookback window: ‡πÉ‡∏ä‡πâ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î 10 ‡∏á‡∏ß‡∏î
            lookback = min(10, index)
            history = data[max(0, index - lookback) : index]

            if history: # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥
                # Features ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏•‡∏Ç 6 ‡∏´‡∏•‡∏±‡∏Å (Position-specific)
                for pos in range(6):
                    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤ 'digits' ‡∏°‡∏µ‡∏≠‡∏¢‡∏π‡πà‡πÅ‡∏•‡∏∞‡∏°‡∏µ‡∏Ç‡∏ô‡∏≤‡∏î‡∏û‡∏≠
                    if 'digits' in item and len(item['digits']) > pos:
                        pos_history = [h['digits'][pos] for h in history if 'digits' in h and len(h['digits']) > pos]

                        features[f'pos{pos}_lastValue'] = pos_history[-1] if pos_history else 0
                        features[f'pos{pos}_trend'] = (pos_history[-1] - pos_history[-2]) if len(pos_history) > 1 else 0
                        features[f'pos{pos}_stdDev'] = calculate_std_dev(pos_history) if pos_history else 0.0
                    else:
                        # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏ñ‡πâ‡∏≤ 'digits' ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏≠‡∏¢‡∏π‡πà‡∏´‡∏£‡∏∑‡∏≠‡∏Ç‡∏ô‡∏≤‡∏î‡πÑ‡∏°‡πà‡∏û‡∏≠
                        features[f'pos{pos}_lastValue'] = 0
                        features[f'pos{pos}_trend'] = 0
                        features[f'pos{pos}_stdDev'] = 0.0

                # Features ‡∏ó‡∏±‡πà‡∏ß‡πÑ‡∏õ
                features['avgSum'] = np.mean([h['sum'] for h in history]) if history else 0
                features['avgEvenCount'] = np.mean([h['evenCount'] for h in history]) if history else 0

                # Features ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á
                last2_history = [h['last2Int'] for h in history if 'last2Int' in h]
                features['last2_lastValue'] = last2_history[-1] if last2_history else 0
                features['last2_frequency_in_history'] = last2_history.count(item['last2Int']) if 'last2Int' in item else 0

                # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì 'gap' ‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏•‡∏∞‡πÄ‡∏≠‡∏µ‡∏¢‡∏î
                last_same_index = -1
                if 'last2Int' in item:
                    for i in range(index - 1, -1, -1):
                        if data[i]['last2Int'] == item['last2Int']:
                            last_same_index = i
                            break
                features['last2_gap'] = index - last_same_index if last_same_index >= 0 else index + 1

            else: # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥ (‡πÄ‡∏ä‡πà‡∏ô ‡πÅ‡∏ñ‡∏ß‡πÅ‡∏£‡∏Å‡πÜ)
                for pos in range(6):
                    features[f'pos{pos}_lastValue'] = 0
                    features[f'pos{pos}_trend'] = 0
                    features[f'pos{pos}_stdDev'] = 0.0
                features['avgSum'] = 0
                features['avgEvenCount'] = 0
                features['last2_lastValue'] = 0
                features['last2_frequency_in_history'] = 0
                features['last2_gap'] = index + 1

            # Features ‡∏ï‡∏≤‡∏°‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà
            features['dayOfWeek'] = item['date'].weekday() if 'date' in item else 0
            features['month'] = item['date'].month if 'date' in item else 0
            features['dayOfMonth'] = item['date'].day if 'date' in item else 0
            features['weekOfYear'] = item['date'].isocalendar()[1] if 'date' in item and hasattr(item['date'], 'isocalendar') else 0

            features_list.append(features)

        print(f"‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á Sequential Features ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à: {len(features_list)} ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£")
        return features_list

    def predict_last2_digits(features):
        """
        ‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏ó‡∏≤‡∏á
        """
        print("üéØ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á (Specialized Model - Second Draft)...")

        if len(features) < 10:
            print("‚ö†Ô∏è ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÑ‡∏°‡πà‡πÄ‡∏û‡∏µ‡∏¢‡∏á‡∏û‡∏≠‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á (‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 10 ‡∏á‡∏ß‡∏î)")
            return {'predictions': [], 'analysis': {}}

        pair_freq = {} # ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà
        pair_last_seen = {} # ‡∏á‡∏ß‡∏î‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡∏ó‡∏µ‡πà‡πÄ‡∏´‡πá‡∏ô
        pair_gaps_history = {} # ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏ä‡πà‡∏ß‡∏á‡∏´‡πà‡∏≤‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏£‡∏±‡πâ‡∏á

        # ‡∏Å‡∏£‡∏≠‡∏á‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏µ 'last2' ‡πÅ‡∏•‡∏∞ 'last2Int'
        valid_features = [f for f in features if 'last2' in f and 'last2Int' in f]

        # ‚úÖ ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏ï‡∏£‡∏á‡∏ô‡∏µ‡πâ: ‡∏¢‡πâ‡∏≤‡∏¢‡∏Å‡∏≤‡∏£‡∏Å‡∏≥‡∏´‡∏ô‡∏î recent_features ‡∏Ç‡∏∂‡πâ‡∏ô‡∏°‡∏≤‡∏Å‡πà‡∏≠‡∏ô‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô
        recent_features = [f for f in valid_features[-20:] if 'last2Int' in f]

        for index, item in enumerate(valid_features):
            pair = item['last2']
            pair_freq[pair] = pair_freq.get(pair, 0) + 1

            if pair in pair_last_seen:
                gap = index - pair_last_seen[pair]
                pair_gaps_history.setdefault(pair, []).append(gap)
            pair_last_seen[pair] = index

        # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏¢
        if not pair_freq:
            return {'predictions': [], 'analysis': {'totalUniquePairs': 0}}


        # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏ä‡πà‡∏ß‡∏á‡∏´‡πà‡∏≤‡∏á (periodicity)
        avg_gaps = {pair: np.mean(gaps) if gaps else len(valid_features) + 1 for pair, gaps in pair_gaps_history.items()}

        sorted_pairs_info = []
        for pair, freq in pair_freq.items():
            current_gap = len(valid_features) - pair_last_seen[pair] - 1 # ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏á‡∏ß‡∏î‡∏ó‡∏µ‡πà‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏≠‡∏≠‡∏Å
            expected_gap = avg_gaps.get(pair, len(valid_features) + 1)

            # ‚úÖ NEW: Ensemble Scoring
            score = 0.0
            score += freq * 0.3 # ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà
            score += (100 - min(100, current_gap)) * 0.3 # ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡∏Ñ‡∏ß‡∏≤‡∏°‡πÉ‡∏´‡∏°‡πà (‡∏¢‡∏¥‡πà‡∏á‡∏´‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ ‡∏¢‡∏¥‡πà‡∏á‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏™‡∏π‡∏á)

            # Due factor (‡∏ñ‡πâ‡∏≤‡∏´‡πà‡∏≤‡∏á‡πÄ‡∏Å‡∏¥‡∏ô‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏°‡∏≤‡∏Å ‡∏≠‡∏≤‡∏à‡∏à‡∏∞ "‡∏ñ‡∏∂‡∏á‡πÄ‡∏ß‡∏•‡∏≤" ‡∏≠‡∏≠‡∏Å)
            if current_gap > expected_gap * 1.5:
                score += (current_gap - expected_gap) * 0.2
            elif current_gap < expected_gap * 0.5: # ‡∏ñ‡πâ‡∏≤‡∏≠‡∏≠‡∏Å‡∏ö‡πà‡∏≠‡∏¢‡πÄ‡∏Å‡∏¥‡∏ô‡πÑ‡∏õ ‡∏≠‡∏≤‡∏à‡∏à‡∏∞‡∏•‡∏î‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏™‡∏ô‡πÉ‡∏à
                score -= (expected_gap - current_gap) * 0.1

            # Pattern strength (‡πÄ‡∏ä‡πà‡∏ô ‡πÄ‡∏•‡∏Ç‡∏´‡∏•‡∏±‡∏Å‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏ã‡πâ‡∏≥ ‡∏´‡∏£‡∏∑‡∏≠‡∏°‡∏µ‡πÅ‡∏ô‡∏ß‡πÇ‡∏ô‡πâ‡∏°‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏á‡∏ß‡∏î‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤)
            # ‚úÖ ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏ï‡∏£‡∏á‡∏ô‡∏µ‡πâ: ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö recent_features ‡∏Å‡πà‡∏≠‡∏ô‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô
            if recent_features:
                last_draw_last2 = recent_features[-1]['last2']
                if pair[0] == last_draw_last2[0] or pair[1] == last_draw_last2[1]:
                    score += 5 # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡πÄ‡∏•‡πá‡∏Å‡∏ô‡πâ‡∏≠‡∏¢‡∏´‡∏≤‡∏Å‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏´‡∏•‡∏±‡∏Å‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ö‡∏á‡∏ß‡∏î‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î

            sorted_pairs_info.append({
                'pair': pair,
                'freq': freq,
                'currentGap': current_gap,
                'expectedGap': expected_gap,
                'score': max(0, min(100, score)) # ‡∏à‡∏≥‡∏Å‡∏±‡∏î‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô 0-100
            })

        sorted_pairs_info.sort(key=lambda x: x['score'], reverse=True)

        # sum_mod_10_trend ‡πÅ‡∏•‡∏∞ even_odd_trend ‡πÉ‡∏ä‡πâ recent_features ‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÑ‡∏ß‡πâ‡πÅ‡∏•‡πâ‡∏ß
        sum_mod_10_trend = [f['last2Int'] % 10 for f in recent_features][-5:] if recent_features else []
        even_odd_trend = [f['last2Int'] % 2 for f in recent_features][-5:] if recent_features else []

        # ‚úÖ NEW: Anomaly detection for cold numbers
        all_pairs = [str(i).zfill(2) for i in range(100)]
        cold_pairs = []
        for p in all_pairs:
            if p not in pair_freq: # ‡πÑ‡∏°‡πà‡πÄ‡∏Ñ‡∏¢‡∏≠‡∏≠‡∏Å‡πÄ‡∏•‡∏¢
                cold_pairs.append(p)
            elif p in pair_last_seen:
                # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏ô‡∏≤‡∏ô‡∏°‡∏≤‡∏Å ‡πÅ‡∏•‡∏∞‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏ï‡πà‡∏≥
                current_gap = len(valid_features) - pair_last_seen[p] - 1 # ‡πÉ‡∏ä‡πâ valid_features ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö current_gap
                if current_gap > len(valid_features) * 0.7 and pair_freq[p] < 3: # 70% ‡∏Ç‡∏≠‡∏á‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥ ‡πÅ‡∏•‡∏∞‡∏≠‡∏≠‡∏Å‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ 3 ‡∏Ñ‡∏£‡∏±‡πâ‡∏á
                    cold_pairs.append(p)

        top_predictions = [
            {
                'value': p['pair'],
                'confidence': min(0.99, p['score'] / 100),
                'reason': f"‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô: {p['score']:.1f}, ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà: {p['freq']}, ‡∏´‡πà‡∏≤‡∏á: {p['currentGap']} ‡∏á‡∏ß‡∏î, ‡∏Ñ‡∏≤‡∏î‡∏´‡∏ß‡∏±‡∏á: {p['expectedGap']:.1f} ‡∏á‡∏ß‡∏î"
            } for p in sorted_pairs_info[:10]
        ]

        return {
            'predictions': top_predictions,
            'analysis': {
                'totalUniquePairs': len(pair_freq),
                'mostFrequent': sorted_pairs_info[0] if sorted_pairs_info else None,
                'longestCurrentGap': max([p['currentGap'] for p in sorted_pairs_info]) if sorted_pairs_info else 0,
                'patterns': {
                    'sumMod10Trend': sum_mod_10_trend,
                    'evenOddTrend': even_odd_trend
                },
                'coldPairs': cold_pairs[:5] # ‡πÅ‡∏™‡∏î‡∏á 5 ‡∏Ñ‡∏π‡πà‡∏ó‡∏µ‡πà‡πÄ‡∏¢‡πá‡∏ô‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
            }
        }

    def predict_sequential_6_digits(features):
        """
        ‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç 6 ‡∏´‡∏•‡∏±‡∏Å‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏• Sequential (Markov Chain like)
        """
        print("üéØ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç 6 ‡∏´‡∏•‡∏±‡∏Å‡πÅ‡∏ö‡∏ö Sequential (Second Draft)...")

        if len(features) < 30:
            print("‚ö†Ô∏è ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÑ‡∏°‡πà‡πÄ‡∏û‡∏µ‡∏¢‡∏á‡∏û‡∏≠‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç 6 ‡∏´‡∏•‡∏±‡∏Å (‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ 30 ‡∏á‡∏ß‡∏î)")
            return {'predictions': [], 'positionAnalysis': []}

        recent_draws = [f for f in features[-30:] if 'digits' in f and len(f['digits']) == 6] # ‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏´‡πâ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤‡∏°‡∏µ 'digits' ‡πÅ‡∏•‡∏∞‡∏Ñ‡∏£‡∏ö 6 ‡∏´‡∏•‡∏±‡∏Å
        if not recent_draws:
            print("‚ö†Ô∏è ‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• 6 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ô 30 ‡∏á‡∏ß‡∏î‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î")
            return {'predictions': [], 'positionAnalysis': []}

        position_predictions = []

        for pos in range(6): # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á (0 ‡∏ñ‡∏∂‡∏á 5)
            pos_values = [d['digits'][pos] for d in recent_draws if len(d['digits']) > pos]
            if not pos_values: # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏ô‡∏µ‡πâ‡πÄ‡∏•‡∏¢
                position_predictions.append({
                    'position': pos,
                    'candidates': [{'digit': i, 'weightedCount': 0, 'probability': 0.1} for i in range(10)], # Default candidates
                    'bestGuess': 0
                })
                continue

            transitions = {} # from_digit -> to_digit : weighted_count

            # ‡∏™‡∏£‡πâ‡∏≤‡∏á Transition Matrix ‡πÅ‡∏ö‡∏ö‡∏ñ‡πà‡∏ß‡∏á‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡∏ï‡∏≤‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡πÉ‡∏´‡∏°‡πà
            for i in range(1, len(pos_values)):
                from_digit = pos_values[i-1]
                to_digit = pos_values[i]
                weight = i / len(pos_values) # ‡∏á‡∏ß‡∏î‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡∏°‡∏µ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤
                key = (from_digit, to_digit)
                transitions[key] = transitions.get(key, 0) + weight

            current_value = pos_values[-1] # ‡πÄ‡∏•‡∏Ç‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡πÉ‡∏ô‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏ô‡∏µ‡πâ

            candidate_next_digits = {} # digit : weighted_prob
            total_weight_from_current = 0

            # ‡∏´‡∏≤‡πÄ‡∏•‡∏Ç‡∏ñ‡∏±‡∏î‡πÑ‡∏õ‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏õ‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å‡πÄ‡∏•‡∏Ç‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô
            for (from_d, to_d), weighted_count in transitions.items():
                if from_d == current_value:
                    candidate_next_digits[to_d] = candidate_next_digits.get(to_d, 0) + weighted_count
                    total_weight_from_current += weighted_count

            next_candidates_for_pos = []
            if total_weight_from_current > 0:
                for digit, weighted_sum in candidate_next_digits.items():
                    next_candidates_for_pos.append({
                        'digit': digit,
                        'weightedCount': weighted_sum,
                        'probability': weighted_sum / total_weight_from_current
                    })

            # Fallback: ‡∏´‡∏≤‡∏Å‡πÑ‡∏°‡πà‡∏°‡∏µ transition ‡∏à‡∏≤‡∏Å‡πÄ‡∏•‡∏Ç‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô ‡∏´‡∏£‡∏∑‡∏≠ confidence ‡∏ï‡πà‡∏≥‡∏°‡∏≤‡∏Å
            if not next_candidates_for_pos or next_candidates_for_pos[0]['probability'] < 0.1:
                digit_freq = {}
                for i, d in enumerate(pos_values):
                    weight = (i + 1) / len(pos_values) # ‡πÉ‡∏´‡πâ‡∏á‡∏ß‡∏î‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡∏°‡∏µ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡∏°‡∏≤‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô
                    digit_freq[d] = digit_freq.get(d, 0) + weight

                total_freq_weight = sum(digit_freq.values())
                if total_freq_weight > 0:
                    next_candidates_for_pos = sorted(
                        [{'digit': d, 'weightedCount': w, 'probability': w / total_freq_weight}
                         for d, w in digit_freq.items()],
                        key=lambda x: x['weightedCount'], reverse=True
                    )[:5] # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å 5 ‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏ñ‡πà‡∏ß‡∏á‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î
                else: # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏•‡∏¢
                    next_candidates_for_pos = [{'digit': i, 'weightedCount': 0, 'probability': 0.1} for i in range(10)]

            next_candidates_for_pos.sort(key=lambda x: x['probability'], reverse=True)
            position_predictions.append({
                'position': pos,
                'candidates': next_candidates_for_pos[:3], # ‡πÄ‡∏≠‡∏≤ 3 ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
                'bestGuess': next_candidates_for_pos[0]['digit'] if next_candidates_for_pos else 0
            })

        # ‡∏£‡∏ß‡∏°‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡πÄ‡∏Ç‡πâ‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡∏∏‡∏î‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç 6 ‡∏´‡∏•‡∏±‡∏Å
        sequences = []

        def generate_combinations(arrays, current_combination=[], index=0, current_confidence=1.0):
            if len(sequences) >= 10: # ‡∏à‡∏≥‡∏Å‡∏±‡∏î‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ä‡∏∏‡∏î‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÑ‡∏°‡πà‡πÉ‡∏´‡πâ‡∏•‡πâ‡∏ô‡πÄ‡∏Å‡∏¥‡∏ô‡πÑ‡∏õ
                return

            if index == len(arrays):
                sequence_str = "".join(map(str, current_combination))

                real_breakdown = []
                for i, digit in enumerate(current_combination):
                    candidate_info = next(
                        (c for c in arrays[i]['candidates'] if c['digit'] == digit),
                        {'probability': 0.1} # Default confidence if not found
                    )
                    real_breakdown.append({
                        'position': i,
                        'digit': digit,
                        'confidence': candidate_info['probability']
                    })

                sequences.append({
                    'sequence': sequence_str,
                    'confidence': current_confidence,
                    'breakdown': real_breakdown
                })
                return

            # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å candidates ‡∏à‡∏≤‡∏Å‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á ‡πÇ‡∏î‡∏¢‡∏û‡∏¥‡∏à‡∏≤‡∏£‡∏ì‡∏≤ top 2 ‡∏´‡∏£‡∏∑‡∏≠ 3 ‡∏ï‡∏±‡∏ß‡πÅ‡∏£‡∏Å
            candidates_for_this_pos = arrays[index]['candidates'][:2] # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å 2 ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
            if len(candidates_for_this_pos) < 2 and len(arrays[index]['candidates']) > 2:
                candidates_for_this_pos.append(arrays[index]['candidates'][2]) # ‡∏ñ‡πâ‡∏≤‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏ó‡∏µ‡πà 3 ‡∏Å‡πá‡πÄ‡∏≠‡∏≤‡∏°‡∏≤‡∏û‡∏¥‡∏à‡∏≤‡∏£‡∏ì‡∏≤

            for candidate in candidates_for_this_pos:
                generate_combinations(
                    arrays,
                    current_combination + [candidate['digit']],
                    index + 1,
                    current_confidence * candidate['probability']
                )

        # ‡∏Å‡∏£‡∏≠‡∏á position_predictions ‡∏ó‡∏µ‡πà‡∏°‡∏µ candidates ‡∏à‡∏£‡∏¥‡∏á‡πÜ ‡∏Å‡πà‡∏≠‡∏ô‡∏™‡πà‡∏á‡πÉ‡∏´‡πâ generate_combinations
        valid_pos_preds = [p for p in position_predictions if p['candidates']]
        if len(valid_pos_preds) == 6: # ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ 6 ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏Ñ‡∏£‡∏ö‡∏ñ‡πâ‡∏ß‡∏ô
            generate_combinations(valid_pos_preds)
        else:
            print("‚ö†Ô∏è ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ä‡∏∏‡∏î‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢ 6 ‡∏´‡∏•‡∏±‡∏Å‡πÑ‡∏î‡πâ‡∏Ñ‡∏£‡∏ö‡∏ñ‡πâ‡∏ß‡∏ô ‡πÄ‡∏ô‡∏∑‡πà‡∏≠‡∏á‡∏à‡∏≤‡∏Å‡∏ö‡∏≤‡∏á‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ú‡∏π‡πâ‡∏™‡∏°‡∏±‡∏Ñ‡∏£‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°")

        # ‡∏à‡∏±‡∏î‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏ä‡∏∏‡∏î‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏ï‡∏≤‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡∏°‡∏±‡πà‡∏ô‡πÉ‡∏à
        sequences.sort(key=lambda x: x['confidence'], reverse=True)

        return {
            'predictions': sequences[:5], # ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å 5 ‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà‡∏°‡∏±‡πà‡∏ô‡πÉ‡∏à‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
            'positionAnalysis': position_predictions
        }

    # ‡∏™‡∏£‡πâ‡∏≤‡∏á features ‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÅ‡∏•‡πâ‡∏ß
    features = extract_sequential_features(processed_data)

    # ‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏ú‡∏•
    sequential_6_digit_predictions = predict_sequential_6_digits(features)
    specialized_2_digit_predictions = predict_last2_digits(features)

    # ‡πÄ‡∏Å‡πá‡∏ö‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡πÑ‡∏ß‡πâ‡πÉ‡∏ô‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£ global ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏™‡πà‡∏á‡∏ï‡πà‡∏≠
    _predictions_cell2 = {
        'sixDigit': sequential_6_digit_predictions,
        'twoDigit': specialized_2_digit_predictions
    }
    _analysis_results_cell2 = {
        'totalRecords': len(processed_data),
        'originalRecords': len(processed_data), # ‡πÉ‡∏ä‡πâ len(processed_data) ‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡πÅ‡∏•‡πâ‡∏ß
        'dataRange': {
            'from': processed_data[0]['dateStr'],
            'to': processed_data[-1]['dateStr']
        },
        'qualityScore': (len(processed_data) / len(processed_data)) * 100, # ‡∏Ñ‡∏ß‡∏£‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô 100% ‡∏ñ‡πâ‡∏≤ processed_data ‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
        'improvements': [
            "‚úÖ ‡∏Å‡∏£‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô",
            "‚úÖ ‡πÄ‡∏ï‡∏¥‡∏° 0 ‡πÅ‡∏•‡∏∞‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• (6 ‡∏´‡∏•‡∏±‡∏Å, 2 ‡∏´‡∏•‡∏±‡∏Å)",
            "‚úÖ Standardize Date Format",
            "üîÑ ‡∏™‡∏£‡πâ‡∏≤‡∏á Feature ‡∏ó‡∏≤‡∏á‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏ï‡∏¥‡∏° (‡∏Ñ‡πà‡∏≤‡∏™‡πà‡∏ß‡∏ô‡πÄ‡∏ö‡∏µ‡πà‡∏¢‡∏á‡πÄ‡∏ö‡∏ô‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô)",
            "üîÑ ‡∏õ‡∏£‡∏±‡∏ö‡∏õ‡∏£‡∏∏‡∏á Lookback Window ‡πÉ‡∏´‡πâ‡∏¢‡∏∑‡∏î‡∏´‡∏¢‡∏∏‡πà‡∏ô‡∏Ç‡∏∂‡πâ‡∏ô",
            "üîÑ ‡πÉ‡∏ä‡πâ‡∏Å‡∏≤‡∏£‡∏ñ‡πà‡∏ß‡∏á‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å‡∏ï‡∏≤‡∏°‡∏Ñ‡∏ß‡∏≤‡∏°‡πÉ‡∏´‡∏°‡πà‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô Transition Probabilities",
            "üîÑ ‡∏£‡∏∞‡∏ö‡∏ö Ensemble Scoring ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏•‡∏Ç 2 ‡∏ï‡∏±‡∏ß‡∏•‡πà‡∏≤‡∏á",
            "üîÑ ‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç 6 ‡∏´‡∏•‡∏±‡∏Å‡∏û‡∏¢‡∏≤‡∏¢‡∏≤‡∏°‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà‡∏´‡∏•‡∏≤‡∏Å‡∏´‡∏•‡∏≤‡∏¢‡∏°‡∏≤‡∏Å‡∏Ç‡∏∂‡πâ‡∏ô"
        ]
    }
    print("‚úÖ ‡∏£‡∏∞‡∏ö‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡πÉ‡∏ô‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ñ‡∏±‡∏î‡πÑ‡∏õ")

```
