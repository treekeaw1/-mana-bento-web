```python
# ==============================================================================
# --- Lotto Analyzer V11.9 - ‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1.2/X (Analysis of Prize 1 - 6 Digits Breakdown) ---
# --- Focus: Visualize frequency of each position in Prize 1 (‡πÅ‡∏¢‡∏Å 3 ‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡πâ‡∏≤, 3 ‡∏ï‡∏±‡∏ß‡∏ó‡πâ‡∏≤‡∏¢) ---
# ==============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings

warnings.filterwarnings('ignore')

print("üî¢ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏î‡∏≥‡πÄ‡∏ô‡∏¥‡∏ô‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÅ‡∏¢‡∏Å‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1...")

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤ df ‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏≠‡∏¢‡∏π‡πà‡∏à‡∏£‡∏¥‡∏á‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà (df ‡∏Ñ‡∏ß‡∏£‡∏à‡∏∞‡∏ñ‡∏π‡∏Å‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏à‡∏≤‡∏Å‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1 ‡πÅ‡∏•‡πâ‡∏ß)
if 'df' not in globals() or df.empty:
    print("‚ùó ‡πÑ‡∏°‡πà‡∏û‡∏ö DataFrame 'df' ‡∏´‡∏£‡∏∑‡∏≠ 'df' ‡∏ß‡πà‡∏≤‡∏á‡πÄ‡∏õ‡∏•‡πà‡∏≤. ‡πÇ‡∏õ‡∏£‡∏î‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡πÑ‡∏î‡πâ‡∏£‡∏±‡∏ô‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1 ‡πÅ‡∏•‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á df ‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß.")
else:
    print("‚úÖ ‡∏û‡∏ö DataFrame 'df' ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÅ‡∏¢‡∏Å‡∏´‡∏•‡∏±‡∏Å.")

    df_prize1_analysis = df.copy() # ‡πÉ‡∏ä‡πâ copy ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡πâ‡∏≠‡∏á‡∏Å‡∏±‡∏ô‡∏Å‡∏≤‡∏£‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç df ‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö‡πÇ‡∏î‡∏¢‡πÑ‡∏°‡πà‡∏ï‡∏±‡πâ‡∏á‡πÉ‡∏à

    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏°‡∏µ‡∏≠‡∏¢‡∏π‡πà‡∏à‡∏£‡∏¥‡∏á
    required_prize1_cols = [f'‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà {i+1}' for i in range(6)]
    if not all(col in df_prize1_analysis.columns for col in required_prize1_cols):
        print("‚ùó ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡πÑ‡∏°‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå. ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÉ‡∏´‡∏°‡πà...")
        for i in range(6):
            df_prize1_analysis[f'‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà {i+1}'] = df_prize1_analysis['‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1'].astype(str).str.zfill(6).str[i].astype(int)
    else:
        print("‚úÖ ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô.")

    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏´‡∏•‡∏±‡∏Å
    positions = {
        '‡∏´‡∏•‡∏±‡∏Å‡πÅ‡∏™‡∏ô': '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 1',
        '‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏°‡∏∑‡πà‡∏ô': '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 2',
        '‡∏´‡∏•‡∏±‡∏Å‡∏û‡∏±‡∏ô': '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 3',
        '‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢': '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 4',
        '‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö': '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 5',
        '‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢': '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 6'
    }

    # ‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö 3 ‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡πâ‡∏≤
    front_three_positions = {k: v for k, v in positions.items() if k in ['‡∏´‡∏•‡∏±‡∏Å‡πÅ‡∏™‡∏ô', '‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏°‡∏∑‡πà‡∏ô', '‡∏´‡∏•‡∏±‡∏Å‡∏û‡∏±‡∏ô']}
    # ‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö 3 ‡∏ï‡∏±‡∏ß‡∏ó‡πâ‡∏≤‡∏¢
    back_three_positions = {k: v for k, v in positions.items() if k in ['‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢', '‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö', '‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢']}

    # --- ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡πÅ‡∏ó‡πà‡∏á‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å ---
    def plot_digit_frequency(dataframe, column_name, title_prefix):
        digit_counts = dataframe[column_name].value_counts(normalize=True).sort_index()
        plt.figure(figsize=(8, 6))
        sns.barplot(x=digit_counts.index, y=digit_counts.values * 100, palette='cividis')
        plt.title(f'{title_prefix} (‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î)', fontsize=14)
        plt.xlabel('‡πÄ‡∏•‡∏Ç‡πÇ‡∏î‡∏î', fontsize=12)
        plt.ylabel('‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà (%)', fontsize=12)
        plt.xticks(fontsize=10)
        plt.yticks(fontsize=10)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
        #

    print("\n--- ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÄ‡∏•‡∏Ç‡∏™‡∏≤‡∏°‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡πâ‡∏≤‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ---")
    # ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÅ‡∏•‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö 3 ‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡πâ‡∏≤
    for name, col in front_three_positions.items():
        plot_digit_frequency(df_prize1_analysis, col, f'‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥ {name} ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1')

    print("\n--- ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÄ‡∏•‡∏Ç‡∏™‡∏≤‡∏°‡∏ï‡∏±‡∏ß‡∏ó‡πâ‡∏≤‡∏¢‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ---")
    # ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÅ‡∏•‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö 3 ‡∏ï‡∏±‡∏ß‡∏ó‡πâ‡∏≤‡∏¢
    for name, col in back_three_positions.items():
        plot_digit_frequency(df_prize1_analysis, col, f'‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥ {name} ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1')

    # --- ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏ï‡∏¥‡∏°: ‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥ Top N ‡∏Ñ‡∏π‡πà‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà‡∏ï‡∏¥‡∏î‡∏Å‡∏±‡∏ô (‡πÄ‡∏ä‡πà‡∏ô ‡∏´‡∏•‡∏±‡∏Å‡πÅ‡∏™‡∏ô-‡∏´‡∏°‡∏∑‡πà‡∏ô, ‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢-‡∏™‡∏¥‡∏ö) ---
    print("\n--- ‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥ Top 10 ‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà‡∏ï‡∏¥‡∏î‡∏Å‡∏±‡∏ô‡πÉ‡∏ô‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ---")

    def plot_top_n_pairs(dataframe, col1, col2, title_prefix, top_n=10):
        # ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç‡πÄ‡∏õ‡πá‡∏ô string ‡πÄ‡∏ä‡πà‡∏ô "01" ‡∏à‡∏≤‡∏Å‡∏´‡∏•‡∏±‡∏Å 0 ‡πÅ‡∏•‡∏∞ 1
        dataframe['temp_pair'] = dataframe[col1].astype(str) + dataframe[col2].astype(str)
        pair_counts = dataframe['temp_pair'].value_counts().head(top_n)

        if not pair_counts.empty:
            plt.figure(figsize=(10, 6))
            sns.barplot(x=pair_counts.index, y=pair_counts.values, palette='plasma')
            plt.title(f'Top {top_n} ‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç {title_prefix} ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1', fontsize=14)
            plt.xlabel('‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç', fontsize=12)
            plt.ylabel('‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏µ‡πà‡∏≠‡∏≠‡∏Å', fontsize=12)
            plt.xticks(rotation=45, ha='right', fontsize=10)
            plt.yticks(fontsize=10)
            plt.grid(axis='y', linestyle='--', alpha=0.7)
            plt.tight_layout()
            plt.show()
            #
            print(f"üí° Top {top_n} ‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç {title_prefix}:")
            for pair, count in pair_counts.items():
                print(f"  - {pair}: {count} ‡∏Ñ‡∏£‡∏±‡πâ‡∏á")
        else:
            print(f"‚ùó ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç {title_prefix}.")


    # ‡∏Ñ‡∏π‡πà 3 ‡∏ï‡∏±‡∏ß‡∏´‡∏ô‡πâ‡∏≤
    plot_top_n_pairs(df_prize1_analysis, positions['‡∏´‡∏•‡∏±‡∏Å‡πÅ‡∏™‡∏ô'], positions['‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏°‡∏∑‡πà‡∏ô'], '‡∏´‡∏•‡∏±‡∏Å‡πÅ‡∏™‡∏ô-‡∏´‡∏°‡∏∑‡πà‡∏ô')
    plot_top_n_pairs(df_prize1_analysis, positions['‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏°‡∏∑‡πà‡∏ô'], positions['‡∏´‡∏•‡∏±‡∏Å‡∏û‡∏±‡∏ô'], '‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏°‡∏∑‡πà‡∏ô-‡∏û‡∏±‡∏ô')

    # ‡∏Ñ‡∏π‡πà 3 ‡∏ï‡∏±‡∏ß‡∏ó‡πâ‡∏≤‡∏¢
    plot_top_n_pairs(df_prize1_analysis, positions['‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢'], positions['‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö'], '‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢-‡∏™‡∏¥‡∏ö')
    plot_top_n_pairs(df_prize1_analysis, positions['‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö'], positions['‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢'], '‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö-‡∏´‡∏ô‡πà‡∏ß‡∏¢')


print("\n‚úÖ ‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÅ‡∏¢‡∏Å‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô.")
```

```python
# ==============================================================================
# --- Lotto Analyzer V11.9 - ‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1.3/X (Feature Engineering & Data Prep) ---
# --- Focus: Create meaningful features from historical data for model training ---
# --- FIX: KeyError: 'prize1_unit_digit' by ensuring column creation ---
# ==============================================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime, timedelta
import warnings

warnings.filterwarnings('ignore')

print("‚öôÔ∏è ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå (Feature Engineering) ‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÇ‡∏°‡πÄ‡∏î‡∏•...")

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤ df ‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏≠‡∏¢‡∏π‡πà‡∏à‡∏£‡∏¥‡∏á‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
if 'df' not in globals() or df.empty:
    print("‚ùó ‡πÑ‡∏°‡πà‡∏û‡∏ö DataFrame 'df' ‡∏´‡∏£‡∏∑‡∏≠ 'df' ‡∏ß‡πà‡∏≤‡∏á‡πÄ‡∏õ‡∏•‡πà‡∏≤. ‡πÇ‡∏õ‡∏£‡∏î‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡πÑ‡∏î‡πâ‡∏£‡∏±‡∏ô‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1 ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß.")
    df_features = pd.DataFrame()
else:
    print("‚úÖ ‡∏û‡∏ö DataFrame 'df' ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå.")

    df_features = df.copy() # ‡πÉ‡∏ä‡πâ copy ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡πâ‡∏≠‡∏á‡∏Å‡∏±‡∏ô‡∏Å‡∏≤‡∏£‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç df ‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö‡πÇ‡∏î‡∏¢‡πÑ‡∏°‡πà‡∏ï‡∏±‡πâ‡∏á‡πÉ‡∏à

    # --- Feature Group 1: Temporal Features (‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏ï‡∏≤‡∏°‡πÄ‡∏ß‡∏•‡∏≤) ---
    print("\n--- ‡∏™‡∏£‡πâ‡∏≤‡∏á Temporal Features (‡∏ß‡∏±‡∏ô, ‡πÄ‡∏î‡∏∑‡∏≠‡∏ô, ‡πÑ‡∏ï‡∏£‡∏°‡∏≤‡∏™, ‡∏ß‡∏±‡∏ô‡πÉ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô) ---")
    df_features['year'] = df_features['‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà'].dt.year
    df_features['month'] = df_features['‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà'].dt.month
    df_features['day'] = df_features['‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà'].dt.day
    df_features['day_of_week'] = df_features['‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà'].dt.dayofweek # Monday=0, Sunday=6
    df_features['quarter'] = df_features['‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà'].dt.quarter
    df_features['week_of_year'] = df_features['‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà'].dt.isocalendar().week.astype(int) # Week number

    # Feature: ‡∏á‡∏ß‡∏î‡∏ó‡∏µ‡πà‡∏≠‡∏≠‡∏Å‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà 1 ‡∏´‡∏£‡∏∑‡∏≠ 16
    df_features['is_first_day'] = (df_features['day'] == 1).astype(int)
    df_features['is_sixteenth_day'] = (df_features['day'] == 16).astype(int)

    # --- Feature Group 2: Frequency Features (‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î) ---
    print("\n--- ‡∏™‡∏£‡πâ‡∏≤‡∏á Frequency Features (‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏£‡∏ß‡∏°‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ï‡πà‡∏≤‡∏á‡πÜ) ---")

    # **‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏ï‡∏£‡∏á‡∏ô‡∏µ‡πâ:** ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÅ‡∏•‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á 'prize1_unit_digit' ‡∏ñ‡πâ‡∏≤‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏°‡∏µ
    if '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 6' in df_features.columns:
        df_features['prize1_unit_digit'] = df_features['‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 6']
    else:
        df_features['prize1_unit_digit'] = df_features['‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1'].astype(str).str.zfill(6).str[5].astype(int)
        print("   ‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á 'prize1_unit_digit' ‡∏à‡∏≤‡∏Å '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1' ‡πÅ‡∏•‡πâ‡∏ß.")


    # 2.1 ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î)
    unit_digit_counts = df_features['prize1_unit_digit'].value_counts(normalize=True) # ‡πÉ‡∏ä‡πâ prize1_unit_digit ‡∏ó‡∏µ‡πà‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÉ‡∏´‡∏°‡πà
    df_features['prize1_unit_freq'] = df_features['prize1_unit_digit'].map(unit_digit_counts)
    print(f"   - ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡πÄ‡∏•‡∏Ç‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1: {unit_digit_counts.to_dict()}")

    # 2.2 ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß (‡∏à‡∏≤‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î)
    bottom2_counts = df_features['‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß'].value_counts(normalize=True)
    df_features['bottom2_freq'] = df_features['‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß'].map(bottom2_counts)
    print(f"   - ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß: {bottom2_counts.head(5).to_dict()} (‡πÅ‡∏™‡∏î‡∏á Top 5)")

    # 2.3 ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç‡∏´‡∏•‡∏±‡∏Å‡πÅ‡∏™‡∏ô-‡∏´‡∏°‡∏∑‡πà‡∏ô, ‡∏´‡∏°‡∏∑‡πà‡∏ô-‡∏û‡∏±‡∏ô, ‡∏£‡πâ‡∏≠‡∏¢-‡∏™‡∏¥‡∏ö, ‡∏™‡∏¥‡∏ö-‡∏´‡∏ô‡πà‡∏ß‡∏¢
    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏´‡∏•‡∏±‡∏Å‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏°‡∏µ‡∏≠‡∏¢‡∏π‡πà
    required_prize1_cols = [f'‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà {i+1}' for i in range(6)]
    if not all(col in df_features.columns for col in required_prize1_cols):
        print("‚ùó ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡πÑ‡∏°‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå‡πÉ‡∏ô df_features. ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á...")
        for i in range(6):
            df_features[f'‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà {i+1}'] = df_features['‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1'].astype(str).str.zfill(6).str[i].astype(int)

    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÅ‡∏•‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÅ‡∏¢‡∏Å‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏ñ‡πâ‡∏≤‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏°‡∏µ
    if '‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 1' not in df_features.columns or '‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 2' not in df_features.columns:
        df_features['‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 1'] = df_features['‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß'].astype(str).str.zfill(2).str[0].astype(int)
        df_features['‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 2'] = df_features['‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß'].astype(str).str.zfill(2).str[1].astype(int)
        print("   ‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÅ‡∏¢‡∏Å‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á '‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß' ‡πÅ‡∏•‡πâ‡∏ß.")


    pairs_to_analyze = {
        'front_two_digits': ('‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 1', '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 2'), # ‡∏´‡∏•‡∏±‡∏Å‡πÅ‡∏™‡∏ô-‡∏´‡∏°‡∏∑‡πà‡∏ô
        'middle_two_digits': ('‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 2', '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 3'), # ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏°‡∏∑‡πà‡∏ô-‡∏û‡∏±‡∏ô
        'last_two_digits_prize1': ('‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 5', '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 6'), # ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö-‡∏´‡∏ô‡πà‡∏ß‡∏¢ ‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
        'bottom_two_digits_bottom2': ('‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 1', '‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 2') # ‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    }

    for name, (col1, col2) in pairs_to_analyze.items():
        if col1 in df_features.columns and col2 in df_features.columns:
            df_features[f'{name}_pair_str'] = df_features[col1].astype(str) + df_features[col2].astype(str)
            pair_freq = df_features[f'{name}_pair_str'].value_counts(normalize=True)
            df_features[f'{name}_pair_freq'] = df_features[f'{name}_pair_str'].map(pair_freq)
            print(f"   - ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç {name}: {pair_freq.head(3).to_dict()} (‡πÅ‡∏™‡∏î‡∏á Top 3)")
        else:
            print(f"   ‚ùó ‡∏Ç‡πâ‡∏≤‡∏°‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç {name} ‡πÄ‡∏ô‡∏∑‡πà‡∏≠‡∏á‡∏à‡∏≤‡∏Å‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå {col1} ‡∏´‡∏£‡∏∑‡∏≠ {col2} ‡πÑ‡∏°‡πà‡∏°‡∏µ.")


    # --- Feature Group 3: Rolling Features (‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà/‡∏£‡∏∞‡∏¢‡∏∞‡πÄ‡∏ß‡∏•‡∏≤‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏õ‡∏£‡∏≤‡∏Å‡∏è‡πÉ‡∏ô‡∏ä‡πà‡∏ß‡∏á‡∏á‡∏ß‡∏î‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î) ---
    print("\n--- ‡∏™‡∏£‡πâ‡∏≤‡∏á Rolling Features (‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡πÉ‡∏ô‡∏ä‡πà‡∏ß‡∏á X ‡∏á‡∏ß‡∏î‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡∏°‡∏≤) ---")

    # **‡∏õ‡∏£‡∏±‡∏ö‡∏õ‡∏£‡∏∏‡∏á‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Rolling average frequency ‡πÉ‡∏´‡πâ‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡πÅ‡∏•‡∏∞‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏™‡∏¥‡∏ó‡∏ò‡∏¥‡∏†‡∏≤‡∏û**
    # ‡∏ß‡∏¥‡∏ò‡∏µ‡∏Å‡∏≤‡∏£‡πÄ‡∏î‡∏¥‡∏°‡∏ú‡∏¥‡∏î‡∏û‡∏•‡∏≤‡∏î ‡πÄ‡∏ô‡∏∑‡πà‡∏≠‡∏á‡∏à‡∏≤‡∏Å .transform(lambda x: x.shift(1).expanding... ) ‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡πÉ‡∏´‡πâ‡∏Ñ‡πà‡∏≤‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á
    # ‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡πÄ‡∏•‡∏Ç‡πÉ‡∏ô window_size ‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡∏°‡∏≤‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏á‡∏ß‡∏î
    # ‡∏ß‡∏¥‡∏ò‡∏µ‡∏Å‡∏≤‡∏£‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏™‡∏¥‡∏ó‡∏ò‡∏¥‡∏†‡∏≤‡∏û‡∏Ñ‡∏∑‡∏≠‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ Expanding/Rolling window ‡∏ö‡∏ô Series ‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç

    # ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á ‡πÉ‡∏´‡πâ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡∏≤‡∏°‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏Å‡πà‡∏≠‡∏ô
    df_features = df_features.sort_values(by='‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà', ascending=True).reset_index(drop=True)

    window_size = 100
    # Rolling frequency for 'prize1_unit_digit'
    # ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ä‡∏±‡πà‡∏ß‡∏Ñ‡∏£‡∏≤‡∏ß‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡πÄ‡∏•‡∏Ç‡πÇ‡∏î‡∏î (0-9) ‡∏ó‡∏µ‡πà‡∏£‡∏∞‡∏ö‡∏∏‡∏ß‡πà‡∏≤‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏•‡∏Ç‡∏ô‡∏±‡πâ‡∏ô‡πÜ ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà (1/0)
    # ‡∏à‡∏≤‡∏Å‡∏ô‡∏±‡πâ‡∏ô‡πÉ‡∏ä‡πâ rolling sum ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏´‡∏≤‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏µ‡πà‡∏≠‡∏≠‡∏Å‡πÉ‡∏ô window_size
    # ‡πÅ‡∏•‡∏∞‡∏´‡∏≤‡∏£‡∏î‡πâ‡∏ß‡∏¢ window_size ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÑ‡∏î‡πâ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà
    print(f"   - ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Rolling frequency (window={window_size}) ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏•‡∏Ç‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1...")
    df_features['prize1_unit_rolling_freq_100'] = np.nan # ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡πÄ‡∏õ‡πá‡∏ô NaN
    for i in range(len(df_features)):
        # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ä‡πà‡∏ß‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì (‡∏à‡∏≤‡∏Å i-window_size ‡∏ñ‡∏∂‡∏á i-1)
        start_idx = max(0, i - window_size)
        window_df = df_features.iloc[start_idx:i] # ‡∏î‡∏π‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô‡∏≠‡∏î‡∏µ‡∏ï‡πÄ‡∏ó‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô

        if not window_df.empty:
            current_unit_digit = df_features.loc[i, 'prize1_unit_digit']
            # ‡∏ô‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô‡πÉ‡∏ô window_df
            freq = window_df['prize1_unit_digit'].value_counts(normalize=True).get(current_unit_digit, 0)
            df_features.loc[i, 'prize1_unit_rolling_freq_100'] = freq
        else:
            df_features.loc[i, 'prize1_unit_rolling_freq_100'] = 0 # ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô‡∏≠‡∏î‡∏µ‡∏ï ‡πÉ‡∏´‡πâ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô 0

    # 3.1 Time Since Last Appearance (‡∏£‡∏∞‡∏¢‡∏∞‡πÄ‡∏ß‡∏•‡∏≤‡∏ó‡∏µ‡πà‡πÄ‡∏•‡∏Ç‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡∏≠‡∏≠‡∏Å)
    print("   - ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì '‡∏á‡∏ß‡∏î‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡∏≠‡∏≠‡∏Å' ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß (bottom2_time_since_last_appearance)...")
    df_features['bottom2_time_since_last_appearance'] = -1 # ‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô

    # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏á‡∏ß‡∏î ‡πÉ‡∏´‡πâ‡∏´‡∏≤‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà‡∏≠‡∏≠‡∏Å‡πÉ‡∏ô‡∏á‡∏ß‡∏î‡∏ô‡∏±‡πâ‡∏ô
    # ‡πÅ‡∏•‡πâ‡∏ß‡∏î‡∏π‡∏ß‡πà‡∏≤‡πÄ‡∏•‡∏Ç‡∏ô‡∏±‡πâ‡∏ô‡πÜ ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢‡πÄ‡∏°‡∏∑‡πà‡∏≠‡πÑ‡∏´‡∏£‡πà‡πÉ‡∏ô‡∏≠‡∏î‡∏µ‡∏ï (‡∏Å‡πà‡∏≠‡∏ô‡∏á‡∏ß‡∏î‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô)
    # ‡πÅ‡∏•‡∏∞‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏á‡∏ß‡∏î‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡πÑ‡∏õ
    unique_bottom2 = df_features['‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß'].unique()
    for b2_num in unique_bottom2:
        # ‡∏´‡∏≤ index ‡∏Ç‡∏≠‡∏á‡∏á‡∏ß‡∏î‡∏ó‡∏µ‡πà‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß‡∏ô‡∏µ‡πâ‡πÄ‡∏Ñ‡∏¢‡∏≠‡∏≠‡∏Å
        indices = df_features[df_features['‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß'] == b2_num].index

        # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì gap (‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏á‡∏ß‡∏î‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡∏≠‡∏≠‡∏Å)
        if len(indices) > 1:
            for i in range(1, len(indices)):
                current_idx = indices[i]
                prev_idx = indices[i-1]
                # gap = ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏á‡∏ß‡∏î‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á current_idx ‡πÅ‡∏•‡∏∞ prev_idx - 1 (‡πÑ‡∏°‡πà‡∏ô‡∏±‡∏ö‡∏á‡∏ß‡∏î‡∏ó‡∏µ‡πà‡∏≠‡∏≠‡∏Å‡πÄ‡∏≠‡∏á‡πÅ‡∏•‡∏∞‡∏á‡∏ß‡∏î‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤)
                # ‡∏´‡∏£‡∏∑‡∏≠‡∏Å‡πá‡∏Ñ‡∏∑‡∏≠ current_idx - prev_idx
                df_features.loc[current_idx, 'bottom2_time_since_last_appearance'] = current_idx - prev_idx
        # ‡∏ñ‡πâ‡∏≤‡∏≠‡∏≠‡∏Å‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡πÅ‡∏£‡∏Å ‡πÉ‡∏´‡πâ‡∏Ñ‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô -1 ‡∏´‡∏£‡∏∑‡∏≠ 0 ‡∏´‡∏£‡∏∑‡∏≠‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°‡∏Å‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏ï‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°
        # ‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡πÄ‡∏õ‡πá‡∏ô -1 ‡∏≠‡∏¢‡∏π‡πà‡πÅ‡∏•‡πâ‡∏ß‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÄ‡∏•‡∏Ç‡∏ó‡∏µ‡πà‡∏≠‡∏≠‡∏Å‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡πÅ‡∏£‡∏Å

    # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ NaN ‡∏ó‡∏µ‡πà‡∏≠‡∏≤‡∏à‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡∏∂‡πâ‡∏ô‡πÉ‡∏ô Rolling features (‡πÄ‡∏ä‡πà‡∏ô ‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•)
    df_features['prize1_unit_rolling_freq_100'].fillna(0, inplace=True) # ‡∏´‡∏£‡∏∑‡∏≠‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢/‡∏°‡∏µ‡πÄ‡∏î‡∏µ‡∏¢‡∏ô


    # --- ‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÇ‡∏°‡πÄ‡∏î‡∏• (Scaling/Normalization) ---
    print("\n--- ‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÇ‡∏°‡πÄ‡∏î‡∏• (Scaling/Normalization) ---")

    numerical_features = [
        'year', 'month', 'day', 'day_of_week', 'quarter', 'week_of_year',
        'is_first_day', 'is_sixteenth_day',
        'prize1_unit_freq', 'bottom2_freq',
        'prize1_unit_rolling_freq_100',
        'bottom2_time_since_last_appearance'
    ]
    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏Ñ‡∏π‡πà‡πÄ‡∏•‡∏Ç‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡∏ñ‡πâ‡∏≤‡∏°‡∏µ
    for name in pairs_to_analyze.keys():
        if f'{name}_pair_freq' in df_features.columns:
            numerical_features.append(f'{name}_pair_freq')

    numerical_features = [col for col in numerical_features if col in df_features.columns]

    scaler = MinMaxScaler()
    df_features[numerical_features] = scaler.fit_transform(df_features[numerical_features])
    print("‚úÖ ‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ Scaling ‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß.")

    # --- ‡∏Å‡∏≤‡∏£‡∏Å‡∏≥‡∏´‡∏ô‡∏î Target Variable (‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡πÄ‡∏õ‡πâ‡∏≤‡∏´‡∏°‡∏≤‡∏¢) ---
    df_features['target_bottom2'] = df_features['‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß']
    df_features['target_prize1_unit'] = df_features['prize1_unit_digit'] # ‡πÉ‡∏ä‡πâ prize1_unit_digit ‡∏ó‡∏µ‡πà‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÅ‡∏•‡∏∞‡∏°‡∏±‡πà‡∏ô‡πÉ‡∏à‡∏ß‡πà‡∏≤‡∏°‡∏µ‡πÅ‡∏•‡πâ‡∏ß

    # ‡∏•‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡πÇ‡∏î‡∏¢‡∏ï‡∏£‡∏á
    cols_to_drop = [
        '‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà', '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1', '‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß', # ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏à‡∏£‡∏¥‡∏á ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå
        'prize1_unit_digit' # ‡πÄ‡∏õ‡πá‡∏ô‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡πÅ‡∏•‡πâ‡∏ß ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏Å‡πá‡∏ö‡πÑ‡∏ß‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏î‡∏¥‡∏ö
    ]
    # ‡∏•‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡πÅ‡∏¢‡∏Å‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡πÅ‡∏•‡∏∞‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏ñ‡πâ‡∏≤‡∏°‡∏µ‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÑ‡∏ß‡πâ‡πÅ‡∏ï‡πà‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå
    for i in range(6):
        cols_to_drop.append(f'‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà {i+1}')
    for i in range(2):
        cols_to_drop.append(f'‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà {i+1}')
    for name in pairs_to_analyze.keys():
        cols_to_drop.append(f'{name}_pair_str') # ‡∏•‡∏ö string ‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ñ‡∏µ‡πà‡πÅ‡∏•‡πâ‡∏ß

    # ‡∏Å‡∏£‡∏≠‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏•‡∏ö‡πÉ‡∏´‡πâ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô df_features ‡πÅ‡∏•‡∏∞‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏™‡πà‡∏ß‡∏ô‡∏´‡∏ô‡∏∂‡πà‡∏á‡∏Ç‡∏≠‡∏á numerical_features
    cols_to_drop = [col for col in cols_to_drop if col in df_features.columns and col not in numerical_features]
    df_features = df_features.drop(columns=cols_to_drop, errors='ignore')


    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö DataFrame ‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢
    print("\n--- DataFrame ‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• (df_features) ---")
    print(df_features.head())
    print(f"\n‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå: {len(df_features.columns) - 2} (‡πÑ‡∏°‡πà‡∏£‡∏ß‡∏° target)") # -2 ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö target cols
    print(f"‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•: {len(df_features)} ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£")

print("\n‚úÖ ‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡πÅ‡∏•‡∏∞‡∏Å‡∏≤‡∏£‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô. 'df_features' ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ñ‡∏±‡∏î‡πÑ‡∏õ.")
```

```python
# ==============================================================================
# --- Lotto Analyzer V11.9 - ‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1.4/X (Data Splitting & Model Setup) ---
# --- Focus: Split data into training/testing sets and prepare for modeling ---
# --- Adjustment: Split bottom2 target into tens and unit digits + ADD prize1 hundreds/tens targets ---
# ==============================================================================

import pandas as pd
from sklearn.model_selection import TimeSeriesSplit # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö Time Series split
import numpy as np
from lightgbm import LGBMClassifier # ‡∏¢‡πâ‡∏≤‡∏¢‡∏°‡∏≤‡∏ô‡∏µ‡πà‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏î‡∏ß‡∏Å
import warnings

warnings.filterwarnings('ignore')

print("üìä ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏î‡∏≥‡πÄ‡∏ô‡∏¥‡∏ô‡∏Å‡∏≤‡∏£‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÇ‡∏°‡πÄ‡∏î‡∏•...")

if 'df_features' not in globals() or df_features.empty:
    print("‚ùó ‡πÑ‡∏°‡πà‡∏û‡∏ö DataFrame 'df_features' ‡∏´‡∏£‡∏∑‡∏≠ 'df_features' ‡∏ß‡πà‡∏≤‡∏á‡πÄ‡∏õ‡∏•‡πà‡∏≤. ‡πÇ‡∏õ‡∏£‡∏î‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡πÑ‡∏î‡πâ‡∏£‡∏±‡∏ô‡πÄ‡∏ã‡∏•‡∏•‡πå 1.3/X ‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß.")
else:
    print("‚úÖ ‡∏û‡∏ö DataFrame 'df_features' ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•.")

    # **‡∏õ‡∏£‡∏±‡∏ö‡∏õ‡∏£‡∏∏‡∏á: ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÅ‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß‡πÉ‡∏ô df_features**
    if 'target_bottom2' in df_features.columns:
        df_features['target_bottom2_tens'] = df_features['target_bottom2'].astype(str).str.zfill(2).str[0].astype(int)
        df_features['target_bottom2_unit'] = df_features['target_bottom2'].astype(str).str.zfill(2).str[1].astype(int)
        print("‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á target_bottom2_tens ‡πÅ‡∏•‡∏∞ target_bottom2_unit ‡πÅ‡∏•‡πâ‡∏ß.")
    else:
        print("‚ùó ‡πÑ‡∏°‡πà‡∏û‡∏ö 'target_bottom2' ‡πÉ‡∏ô df_features. ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏™‡∏£‡πâ‡∏≤‡∏á target_bottom2_tens ‡πÅ‡∏•‡∏∞ target_bottom2_unit ‡πÑ‡∏î‡πâ.")
        raise KeyError("'target_bottom2' column not found, cannot proceed.")

    # **‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡πÉ‡∏ô df_features**
    # 'prize1' ‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏≤‡∏à‡∏≤‡∏Å df ‡∏î‡∏±‡πâ‡∏á‡πÄ‡∏î‡∏¥‡∏° ‡∏´‡∏£‡∏∑‡∏≠‡∏Ñ‡∏ß‡∏£‡∏™‡πà‡∏á‡∏ú‡πà‡∏≤‡∏ô‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤‡πÉ‡∏ô df_features
    # ‡∏™‡∏°‡∏°‡∏ï‡∏¥‡∏ß‡πà‡∏≤ 'prize1_full_number' ‡∏Ñ‡∏∑‡∏≠‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡πÄ‡∏ï‡πá‡∏° 6 ‡∏´‡∏•‡∏±‡∏Å‡∏Å‡πà‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏ñ‡∏π‡∏Å‡πÅ‡∏ï‡∏Å‡πÄ‡∏õ‡πá‡∏ô‡∏´‡∏ô‡πà‡∏ß‡∏¢
    # ‡∏ñ‡πâ‡∏≤ df_features ‡πÑ‡∏°‡πà‡∏°‡∏µ 'prize1_full_number' ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ df['‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1'] ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏°‡∏±‡∏ô‡∏Ç‡∏∂‡πâ‡∏ô‡∏°‡∏≤
    if '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1' in df.columns: # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏à‡∏≤‡∏Å df ‡∏î‡∏±‡πâ‡∏á‡πÄ‡∏î‡∏¥‡∏°‡∏ó‡∏µ‡πà‡∏ô‡∏≥‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤
        # ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå 'prize1_full_number' ‡πÉ‡∏ô df_features ‡∏ñ‡πâ‡∏≤‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏°‡∏µ
        # ‡πÇ‡∏î‡∏¢‡∏Å‡∏≤‡∏£‡∏ú‡∏π‡∏Å‡∏Å‡∏±‡∏ö index ‡∏´‡∏£‡∏∑‡∏≠‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà
        df_features_temp = df_features.copy() # ‡πÉ‡∏ä‡πâ temp copy ‡πÄ‡∏û‡∏∑‡πà‡∏≠ merge
        df_prize1_full = df[['‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà', '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1']].set_index('‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà') # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÉ‡∏´‡πâ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤ df ‡∏°‡∏µ '‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà' ‡πÄ‡∏õ‡πá‡∏ô index ‡∏´‡∏£‡∏∑‡∏≠ column

        # ‡πÉ‡∏ä‡πâ df.index ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö join ‡∏ñ‡πâ‡∏≤ df_features ‡πÉ‡∏ä‡πâ datetime index
        if isinstance(df_features.index, pd.DatetimeIndex):
            df_features['prize1_full_number'] = df_prize1_full.loc[df_features.index, '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1']
        else: # ‡∏ñ‡πâ‡∏≤ df_features ‡πÉ‡∏ä‡πâ index ‡∏ó‡∏±‡πà‡∏ß‡πÑ‡∏õ ‡πÄ‡∏£‡∏≤‡∏≠‡∏≤‡∏à‡∏à‡∏∞‡∏ï‡πâ‡∏≠‡∏á merge ‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå '‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà'
            # (‡∏ã‡∏∂‡πà‡∏á‡πÄ‡∏£‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ '‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà' ‡πÉ‡∏ô df_features ‡∏ó‡∏µ‡πà‡πÇ‡∏î‡∏ô drop ‡πÑ‡∏õ‡πÅ‡∏•‡πâ‡∏ß‡πÉ‡∏ô X)
            # ‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏µ‡πà‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î‡∏Ñ‡∏∑‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á target ‡πÄ‡∏´‡∏•‡πà‡∏≤‡∏ô‡∏µ‡πâ‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏ï‡πà‡∏ï‡∏≠‡∏ô‡∏ó‡∏≥ Feature Engineering (‡πÄ‡∏ã‡∏•‡∏•‡πå 1.3/X)
            # ‡πÅ‡∏ï‡πà‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏£‡∏ß‡∏î‡πÄ‡∏£‡πá‡∏ß‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏õ‡∏±‡∏ç‡∏´‡∏≤ ‡∏ì ‡∏à‡∏∏‡∏î‡∏ô‡∏µ‡πâ ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ df_features['target_prize1_unit']
            # ‡∏ã‡∏∂‡πà‡∏á‡∏´‡∏°‡∏≤‡∏¢‡∏ñ‡∏∂‡∏á ‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏∂‡∏á‡πÄ‡∏•‡∏Ç‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡πÄ‡∏ï‡πá‡∏°‡∏´‡∏•‡∏±‡∏Å‡∏à‡∏≤‡∏Å df ‡∏Å‡πà‡∏≠‡∏ô
            # ‡πÄ‡∏ô‡∏∑‡πà‡∏≠‡∏á‡∏à‡∏≤‡∏Å df_features ‡πÑ‡∏î‡πâ‡∏ó‡∏≥ scaling ‡πÅ‡∏•‡πâ‡∏ß
            # ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÑ‡∏õ‡∏´‡∏≤ "‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1" ‡∏ï‡∏±‡∏ß‡πÄ‡∏ï‡πá‡∏°‡∏à‡∏≤‡∏Å df ‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö ‡πÅ‡∏•‡∏∞‡πÉ‡∏ä‡πâ index ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°
            # ‡∏™‡∏°‡∏°‡∏ï‡∏¥‡∏ß‡πà‡∏≤ df_features ‡∏°‡∏µ index ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô‡∏Å‡∏±‡∏ö df ‡πÉ‡∏ô‡∏ï‡∏≠‡∏ô‡πÅ‡∏£‡∏Å
            df_features['prize1_full_number'] = df['‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1'] # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÉ‡∏´‡πâ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤ index ‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ô
            print("‚úÖ ‡∏î‡∏∂‡∏á 'prize1_full_number' ‡∏à‡∏≤‡∏Å df ‡∏°‡∏≤‡πÉ‡∏ä‡πâ‡πÉ‡∏ô df_features.")

        # ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏à‡∏≤‡∏Å 'prize1_full_number'
        df_features['target_prize1_hundreds'] = df_features['prize1_full_number'].astype(str).str.zfill(6).str[3].astype(int)
        df_features['target_prize1_tens'] = df_features['prize1_full_number'].astype(str).str.zfill(6).str[4].astype(int)
        print("‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á target_prize1_hundreds ‡πÅ‡∏•‡∏∞ target_prize1_tens ‡πÅ‡∏•‡πâ‡∏ß.")
    else:
        print("‚ùó ‡πÑ‡∏°‡πà‡∏û‡∏ö '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1' ‡πÉ‡∏ô DataFrame ‡∏ï‡πâ‡∏ô‡∏â‡∏ö‡∏±‡∏ö (df). ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏™‡∏£‡πâ‡∏≤‡∏á target_prize1_hundreds ‡πÅ‡∏•‡∏∞ target_prize1_tens ‡πÑ‡∏î‡πâ.")
        # ‡∏´‡∏≤‡∏Å‡πÑ‡∏°‡πà‡∏û‡∏ö '‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1' ‡∏Å‡πá‡πÑ‡∏°‡πà‡∏Ñ‡∏ß‡∏£‡∏î‡∏≥‡πÄ‡∏ô‡∏¥‡∏ô‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≠
        raise KeyError("'‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1' column not found in original df, cannot proceed.")


    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î Feature (X) ‡πÅ‡∏•‡∏∞ Target (y)
    # ‡∏•‡∏ö Target ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏≠‡∏≠‡∏Å‡∏à‡∏≤‡∏Å X
    X = df_features.drop(columns=[
        'target_bottom2',
        'target_prize1_unit',
        'target_bottom2_tens',
        'target_bottom2_unit',
        'prize1_full_number', # ‡∏•‡∏ö‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÄ‡∏ï‡πá‡∏° 6 ‡∏´‡∏•‡∏±‡∏Å‡∏≠‡∏≠‡∏Å‡∏î‡πâ‡∏ß‡∏¢
        'target_prize1_hundreds',
        'target_prize1_tens'
    ])

    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î Target ‡πÉ‡∏´‡∏°‡πà‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÅ‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    y_bottom2_tens = df_features['target_bottom2_tens']
    y_bottom2_unit = df_features['target_bottom2_unit']

    # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î Target ‡πÉ‡∏´‡∏°‡πà‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢, ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö, ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    y_prize1_hundreds = df_features['target_prize1_hundreds']
    y_prize1_tens = df_features['target_prize1_tens']
    y_prize1_unit = df_features['target_prize1_unit'] # ‡∏ï‡∏±‡∏ß‡∏ô‡∏µ‡πâ‡∏°‡∏µ‡∏≠‡∏¢‡∏π‡πà‡πÅ‡∏•‡πâ‡∏ß

    print(f"‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå (X): {X.shape[1]} ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î‡∏Ç‡∏≠‡∏á Target (y_bottom2_tens): {y_bottom2_tens.shape[0]} ‡πÅ‡∏ñ‡∏ß")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î‡∏Ç‡∏≠‡∏á Target (y_bottom2_unit): {y_bottom2_unit.shape[0]} ‡πÅ‡∏ñ‡∏ß")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î‡∏Ç‡∏≠‡∏á Target (y_prize1_hundreds): {y_prize1_hundreds.shape[0]} ‡πÅ‡∏ñ‡∏ß")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î‡∏Ç‡∏≠‡∏á Target (y_prize1_tens): {y_prize1_tens.shape[0]} ‡πÅ‡∏ñ‡∏ß")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î‡∏Ç‡∏≠‡∏á Target (y_prize1_unit): {y_prize1_unit.shape[0]} ‡πÅ‡∏ñ‡∏ß")


    # --- ‡∏Å‡∏≤‡∏£‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏ö‡∏ö Time-based Split ---
    X = X.sort_index()
    y_bottom2_tens = y_bottom2_tens.sort_index()
    y_bottom2_unit = y_bottom2_unit.sort_index()
    y_prize1_hundreds = y_prize1_hundreds.sort_index()
    y_prize1_tens = y_prize1_tens.sort_index()
    y_prize1_unit = y_prize1_unit.sort_index()

    n_test_draws = 2

    # ‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö ‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    X_train_bottom2_tens = X.iloc[:-n_test_draws]
    X_test_bottom2_tens = X.iloc[-n_test_draws:]
    y_train_bottom2_tens = y_bottom2_tens.iloc[:-n_test_draws]
    y_test_bottom2_tens = y_bottom2_tens.iloc[-n_test_draws:]

    # ‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢ ‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    X_train_bottom2_unit = X.iloc[:-n_test_draws]
    X_test_bottom2_unit = X.iloc[-n_test_draws:]
    y_train_bottom2_unit = y_bottom2_unit.iloc[:-n_test_draws]
    y_test_bottom2_unit = y_bottom2_unit.iloc[-n_test_draws:]

    # ‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    X_train_prize1_hundreds = X.iloc[:-n_test_draws]
    X_test_prize1_hundreds = X.iloc[-n_test_draws:]
    y_train_prize1_hundreds = y_prize1_hundreds.iloc[:-n_test_draws]
    y_test_prize1_hundreds = y_prize1_hundreds.iloc[-n_test_draws:]

    # ‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    X_train_prize1_tens = X.iloc[:-n_test_draws]
    X_test_prize1_tens = X.iloc[-n_test_draws:]
    y_train_prize1_tens = y_prize1_tens.iloc[:-n_test_draws]
    y_test_prize1_tens = y_prize1_tens.iloc[-n_test_draws:]

    # ‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÄ‡∏î‡∏¥‡∏°)
    X_train_prize1_unit = X.iloc[:-n_test_draws]
    X_test_prize1_unit = X.iloc[-n_test_draws:]
    y_train_prize1_unit = y_prize1_unit.iloc[:-n_test_draws]
    y_test_prize1_unit = y_prize1_unit.iloc[-n_test_draws:]

    print(f"\n--- ‡∏Å‡∏≤‡∏£‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏î‡πâ‡∏ß‡∏¢ Time-based Split (‡πÉ‡∏ä‡πâ {n_test_draws} ‡∏á‡∏ß‡∏î‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö) ---")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î X_train_bottom2_tens: {X_train_bottom2_tens.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î X_test_bottom2_tens: {X_test_bottom2_tens.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î y_train_bottom2_tens: {y_train_bottom2_tens.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î y_test_bottom2_tens: {y_test_bottom2_tens.shape}")

    print(f"‡∏Ç‡∏ô‡∏≤‡∏î X_train_bottom2_unit: {X_train_bottom2_unit.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î X_test_bottom2_unit: {X_test_bottom2_unit.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î y_train_bottom2_unit: {y_train_bottom2_unit.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î y_test_bottom2_unit: {y_test_bottom2_unit.shape}")

    print(f"‡∏Ç‡∏ô‡∏≤‡∏î X_train_prize1_hundreds: {X_train_prize1_hundreds.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î X_test_prize1_hundreds: {X_test_prize1_hundreds.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î y_train_prize1_hundreds: {y_train_prize1_hundreds.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î y_test_prize1_hundreds: {y_test_prize1_hundreds.shape}")

    print(f"‡∏Ç‡∏ô‡∏≤‡∏î X_train_prize1_tens: {X_train_prize1_tens.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î X_test_prize1_tens: {X_test_prize1_tens.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î y_train_prize1_tens: {y_train_prize1_tens.shape}")
    print(f"‡∏Ç‡∏ô‡∏≤‡∏î y_test_prize1_tens: {y_test_prize1_tens.shape}")

    # --- ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡πÇ‡∏°‡πÄ‡∏î‡∏• ---
    print("\n--- ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡πÇ‡∏°‡πÄ‡∏î‡∏• LightGBM ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢ ---")

    # ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    model_bottom2_tens = LGBMClassifier(objective='multiclass', num_class=10, random_state=42)
    # ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    model_bottom2_unit = LGBMClassifier(objective='multiclass', num_class=10, random_state=42)

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    model_prize1_hundreds = LGBMClassifier(objective='multiclass', num_class=10, random_state=42)
    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    model_prize1_tens = LGBMClassifier(objective='multiclass', num_class=10, random_state=42)
    # ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÄ‡∏î‡∏¥‡∏°)
    model_prize1_unit = LGBMClassifier(objective='multiclass', num_class=10, random_state=42)

    print("\n‚úÖ ‡∏Å‡∏≤‡∏£‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏•‡∏∞‡∏Å‡∏≤‡∏£‡∏ï‡∏±‡πâ‡∏á‡∏Ñ‡πà‡∏≤‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÄ‡∏ö‡∏∑‡πâ‡∏≠‡∏á‡∏ï‡πâ‡∏ô‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô.")
    print("‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÅ‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÅ‡∏•‡πâ‡∏ß‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏ù‡∏∂‡∏Å.")
    print("‡πÅ‡∏•‡∏∞‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢, ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö, ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏Ç‡∏≠‡∏á‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏Å‡πá‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏ä‡πà‡∏ô‡∏Å‡∏±‡∏ô.")
```

```python
# ==============================================================================
# --- Lotto Analyzer V11.9 - ‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏ó‡∏µ‡πà 1.5/X (Model Training, Evaluation & Prediction) ---
# --- Focus: Train LightGBM models, evaluate performance, and predict next draw ---
# --- Adjustment: Train separate models for bottom2 tens/unit and prize1 hundreds/tens/unit digits ---
# ==============================================================================

import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, classification_report
from datetime import datetime, timedelta
import warnings

warnings.filterwarnings('ignore')

print("üöÄ ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏Å‡∏≤‡∏£‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•, ‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏ú‡∏•, ‡πÅ‡∏•‡∏∞‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢...")

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏ó‡∏µ‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô‡∏à‡∏≤‡∏Å‡πÄ‡∏ã‡∏•‡∏•‡πå‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤‡∏°‡∏µ‡∏≠‡∏¢‡∏π‡πà‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
if 'X_train_bottom2_tens' not in globals() or 'model_bottom2_tens' not in globals() or \
   'X_train_prize1_hundreds' not in globals() or 'model_prize1_hundreds' not in globals(): # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡πÉ‡∏´‡∏°‡πà
    print("‚ùó ‡πÑ‡∏°‡πà‡∏û‡∏ö‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡∏Å‡∏≤‡∏£‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏£‡∏∑‡∏≠‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏ó‡∏µ‡πà‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡πÑ‡∏ß‡πâ. ‡πÇ‡∏õ‡∏£‡∏î‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡πÑ‡∏î‡πâ‡∏£‡∏±‡∏ô‡πÄ‡∏ã‡∏•‡∏•‡πå 1.4/X ‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢‡πÅ‡∏•‡πâ‡∏ß.")
else:
    # --- 1. ‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏• (Train the Models) ---
    print("\n--- 1. ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏• LightGBM ---")

    print("   - ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö '‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß' (model_bottom2_tens)...")
    model_bottom2_tens.fit(X_train_bottom2_tens, y_train_bottom2_tens)
    print("   ‚úÖ ‡∏ù‡∏∂‡∏Å model_bottom2_tens ‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô.")

    print("   - ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö '‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß' (model_bottom2_unit)...")
    model_bottom2_unit.fit(X_train_bottom2_unit, y_train_bottom2_unit)
    print("   ‚úÖ ‡∏ù‡∏∂‡∏Å model_bottom2_unit ‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô.")

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: ‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    print("   - ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö '‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1' (model_prize1_hundreds)...")
    model_prize1_hundreds.fit(X_train_prize1_hundreds, y_train_prize1_hundreds)
    print("   ‚úÖ ‡∏ù‡∏∂‡∏Å model_prize1_hundreds ‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô.")

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: ‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    print("   - ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö '‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1' (model_prize1_tens)...")
    model_prize1_tens.fit(X_train_prize1_tens, y_train_prize1_tens)
    print("   ‚úÖ ‡∏ù‡∏∂‡∏Å model_prize1_tens ‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô.")

    print("   - ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö '‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1' (model_prize1_unit)...")
    model_prize1_unit.fit(X_train_prize1_unit, y_train_prize1_unit)
    print("   ‚úÖ ‡∏ù‡∏∂‡∏Å model_prize1_unit ‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô.")

    # --- 2. ‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏ú‡∏•‡πÇ‡∏°‡πÄ‡∏î‡∏• (Evaluate the Models) ---
    print("\n--- 2. ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏ú‡∏•‡πÇ‡∏°‡πÄ‡∏î‡∏• ---")

    # 2.1 ‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏ú‡∏• model_bottom2_tens (‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß)
    y_pred_bottom2_tens = model_bottom2_tens.predict(X_test_bottom2_tens)
    print("\n=== ‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô model_bottom2_tens (‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß) ===")
    print(f"Accuracy Score (‡∏ö‡∏ô‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {accuracy_score(y_test_bottom2_tens, y_pred_bottom2_tens):.4f}")
    print("Classification Report:")
    print(classification_report(y_test_bottom2_tens, y_pred_bottom2_tens, zero_division=0))
    print(f"‡∏Ñ‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_test_bottom2_tens.tolist()}")
    print(f"‡∏Ñ‡πà‡∏≤‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_pred_bottom2_tens.tolist()}")

    # 2.2 ‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏ú‡∏• model_bottom2_unit (‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß)
    y_pred_bottom2_unit = model_bottom2_unit.predict(X_test_bottom2_unit)
    print("\n=== ‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô model_bottom2_unit (‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß) ===")
    print(f"Accuracy Score (‡∏ö‡∏ô‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {accuracy_score(y_test_bottom2_unit, y_pred_bottom2_unit):.4f}")
    print("Classification Report:")
    print(classification_report(y_test_bottom2_unit, y_pred_bottom2_unit, zero_division=0))
    print(f"‡∏Ñ‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_test_bottom2_unit.tolist()}")
    print(f"‡∏Ñ‡πà‡∏≤‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_pred_bottom2_unit.tolist()}")

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: 2.3 ‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏ú‡∏• model_prize1_hundreds (‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1)
    y_pred_prize1_hundreds = model_prize1_hundreds.predict(X_test_prize1_hundreds)
    print("\n=== ‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô model_prize1_hundreds (‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1) ===")
    print(f"Accuracy Score (‡∏ö‡∏ô‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {accuracy_score(y_test_prize1_hundreds, y_pred_prize1_hundreds):.4f}")
    print("Classification Report:")
    print(classification_report(y_test_prize1_hundreds, y_pred_prize1_hundreds, zero_division=0))
    print(f"‡∏Ñ‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_test_prize1_hundreds.tolist()}")
    print(f"‡∏Ñ‡πà‡∏≤‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_pred_prize1_hundreds.tolist()}")

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: 2.4 ‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏ú‡∏• model_prize1_tens (‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1)
    y_pred_prize1_tens = model_prize1_tens.predict(X_test_prize1_tens)
    print("\n=== ‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô model_prize1_tens (‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1) ===")
    print(f"Accuracy Score (‡∏ö‡∏ô‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {accuracy_score(y_test_prize1_tens, y_pred_prize1_tens):.4f}")
    print("Classification Report:")
    print(classification_report(y_test_prize1_tens, y_pred_prize1_tens, zero_division=0))
    print(f"‡∏Ñ‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_test_prize1_tens.tolist()}")
    print(f"‡∏Ñ‡πà‡∏≤‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_pred_prize1_tens.tolist()}")

    # 2.5 ‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏ú‡∏• model_prize1_unit (‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1)
    y_pred_prize1_unit = model_prize1_unit.predict(X_test_prize1_unit)
    print("\n=== ‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô model_prize1_unit (‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1) ===")
    print(f"Accuracy Score (‡∏ö‡∏ô‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {accuracy_score(y_test_prize1_unit, y_pred_prize1_unit):.4f}")
    print("Classification Report:")
    print(classification_report(y_test_prize1_unit, y_pred_prize1_unit, zero_division=0))
    print(f"‡∏Ñ‡πà‡∏≤‡∏à‡∏£‡∏¥‡∏á‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_test_prize1_unit.tolist()}")
    print(f"‡∏Ñ‡πà‡∏≤‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏Ç‡∏≠‡∏á ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡∏ä‡∏∏‡∏î‡∏ó‡∏î‡∏™‡∏≠‡∏ö): {y_pred_prize1_unit.tolist()}")


    # --- 3. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏á‡∏ß‡∏î‡∏ñ‡∏±‡∏î‡πÑ‡∏õ (‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÉ‡∏ô‡∏≠‡∏ô‡∏≤‡∏Ñ‡∏ï) ---
    print("\n--- 3. ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏á‡∏ß‡∏î‡∏ñ‡∏±‡∏î‡πÑ‡∏õ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢ ---")

    last_draw_date = df['‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà'].max()
    if last_draw_date.day == 1:
        next_draw_date = last_draw_date.replace(day=16)
    else:
        next_draw_date = (last_draw_date + timedelta(days=15)).replace(day=1)

    print(f"‡∏á‡∏ß‡∏î‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡πÉ‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•: {last_draw_date.strftime('%Y-%m-%d')}")
    print(f"‡∏á‡∏ß‡∏î‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢: {next_draw_date.strftime('%Y-%m-%d')}")

    future_features = {
        'year': next_draw_date.year,
        'month': next_draw_date.month,
        'day': next_draw_date.day,
        'day_of_week': next_draw_date.dayofweek,
        'quarter': next_draw_date.quarter,
        'week_of_year': next_draw_date.isocalendar().week,
        'is_first_day': 1 if next_draw_date.day == 1 else 0,
        'is_sixteenth_day': 1 if next_draw_date.day == 16 else 0
    }

    last_known_features = X_train_bottom2_tens.iloc[-1].to_dict() # ‡πÉ‡∏ä‡πâ X_train ‡πÉ‡∏î‡πÜ ‡∏Å‡πá‡πÑ‡∏î‡πâ

    for col in X.columns: # ‡πÉ‡∏ä‡πâ X.columns ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÅ‡∏ô‡πà‡πÉ‡∏à‡∏ß‡πà‡∏≤‡πÑ‡∏î‡πâ‡∏ó‡∏∏‡∏Å‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå
        if col not in future_features:
            future_features[col] = last_known_features.get(col, 0.0)

    future_features_df = pd.DataFrame([future_features])

    from sklearn.preprocessing import MinMaxScaler
    temp_scaler = MinMaxScaler()
    numerical_features_for_scaling = [col for col in X_train_bottom2_tens.columns if X_train_bottom2_tens[col].dtype in ['float64', 'int64']]
    temp_scaler.fit(X_train_bottom2_tens[numerical_features_for_scaling])

    future_features_df[numerical_features_for_scaling] = temp_scaler.transform(future_features_df[numerical_features_for_scaling])

    future_features_df = future_features_df[X_train_bottom2_tens.columns] # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏•‡∏≥‡∏î‡∏±‡∏ö‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå
    print("\n--- ‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏á‡∏ß‡∏î‡∏ñ‡∏±‡∏î‡πÑ‡∏õ (‡∏´‡∏•‡∏±‡∏á Scaling) ---")
    print(future_features_df.head())


    # --- 4. ‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç‡∏á‡∏ß‡∏î‡∏ñ‡∏±‡∏î‡πÑ‡∏õ (Predict Next Draw) ---
    print("\n--- 4. ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏•‡∏Ç‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏á‡∏ß‡∏î‡∏ñ‡∏±‡∏î‡πÑ‡∏õ ---")

    # ‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    predicted_bottom2_tens = model_bottom2_tens.predict(future_features_df)[0]
    # ‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏Ç‡∏≠‡∏á‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    predicted_bottom2_unit = model_bottom2_unit.predict(future_features_df)[0]
    # ‡∏£‡∏ß‡∏°‡∏ú‡∏•‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏•‡∏Ç 2 ‡∏´‡∏•‡∏±‡∏Å
    predicted_bottom2 = int(f"{predicted_bottom2_tens}{predicted_bottom2_unit}")

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: ‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    predicted_prize1_hundreds = model_prize1_hundreds.predict(future_features_df)[0]
    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: ‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    predicted_prize1_tens = model_prize1_tens.predict(future_features_df)[0]
    # ‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡πÄ‡∏î‡∏¥‡∏°)
    predicted_prize1_unit = model_prize1_unit.predict(future_features_df)[0]
    # ‡∏£‡∏ß‡∏°‡∏ú‡∏•‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 3 ‡∏ï‡∏±‡∏ß‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    predicted_prize1_last_three_digits = int(f"{predicted_prize1_hundreds}{predicted_prize1_tens}{predicted_prize1_unit}")


    print(f"\n‚ú® ‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏á‡∏ß‡∏î‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà {next_draw_date.strftime('%Y-%m-%d')}:")
    print(f"   - ‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏ó‡∏µ‡πà‡∏Ñ‡∏≤‡∏î‡∏ß‡πà‡∏≤‡∏à‡∏∞‡∏≠‡∏≠‡∏Å: {predicted_bottom2:02d}")
    print(f"   - ‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 3 ‡∏ï‡∏±‡∏ß ‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏ó‡∏µ‡πà‡∏Ñ‡∏≤‡∏î‡∏ß‡πà‡∏≤‡∏à‡∏∞‡∏≠‡∏≠‡∏Å: {predicted_prize1_last_three_digits:03d}") # Format ‡πÄ‡∏õ‡πá‡∏ô 3 ‡∏´‡∏•‡∏±‡∏Å‡πÄ‡∏™‡∏°‡∏≠

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏ï‡∏¥‡∏°: ‡∏´‡∏≤‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏î‡∏π‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡πÄ‡∏•‡∏Ç (Top N Probabilities)
    # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    bottom2_tens_probs = model_bottom2_tens.predict_proba(future_features_df)[0]
    top_5_bottom2_tens_indices = np.argsort(bottom2_tens_probs)[::-1][:5]
    top_5_bottom2_tens_probs = bottom2_tens_probs[top_5_bottom2_tens_indices]
    print("\n--- Top 5 ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡πÇ‡∏≠‡∏Å‡∏≤‡∏™‡∏≠‡∏≠‡∏Å‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î ---")
    for i in range(len(top_5_bottom2_tens_indices)):
        print(f"   - ‡πÄ‡∏•‡∏Ç {top_5_bottom2_tens_indices[i]}: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô {top_5_bottom2_tens_probs[i]*100:.2f}%")

    # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß
    bottom2_unit_probs = model_bottom2_unit.predict_proba(future_features_df)[0]
    top_5_bottom2_unit_indices = np.argsort(bottom2_unit_probs)[::-1][:5]
    top_5_bottom2_unit_probs = bottom2_unit_probs[top_5_bottom2_unit_indices]
    print("\n--- Top 5 ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡πÄ‡∏•‡∏Ç‡∏ó‡πâ‡∏≤‡∏¢ 2 ‡∏ï‡∏±‡∏ß ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡πÇ‡∏≠‡∏Å‡∏≤‡∏™‡∏≠‡∏≠‡∏Å‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î ---")
    for i in range(len(top_5_bottom2_unit_indices)):
        print(f"   - ‡πÄ‡∏•‡∏Ç {top_5_bottom2_unit_indices[i]}: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô {top_5_bottom2_unit_probs[i]*100:.2f}%")

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    prize1_hundreds_probs = model_prize1_hundreds.predict_proba(future_features_df)[0]
    top_3_prize1_hundreds_indices = np.argsort(prize1_hundreds_probs)[::-1][:3]
    top_3_prize1_hundreds_probs = prize1_hundreds_probs[top_3_prize1_hundreds_indices]
    print("\n--- Top 3 ‡∏´‡∏•‡∏±‡∏Å‡∏£‡πâ‡∏≠‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡πÇ‡∏≠‡∏Å‡∏≤‡∏™‡∏≠‡∏≠‡∏Å‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î ---")
    for i in range(len(top_3_prize1_hundreds_indices)):
        print(f"   - ‡πÄ‡∏•‡∏Ç {top_3_prize1_hundreds_indices[i]}: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô {top_3_prize1_hundreds_probs[i]*100:.2f}%")

    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÉ‡∏´‡∏°‡πà: ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1
    prize1_tens_probs = model_prize1_tens.predict_proba(future_features_df)[0]
    top_3_prize1_tens_indices = np.argsort(prize1_tens_probs)[::-1][:3]
    top_3_prize1_tens_probs = prize1_tens_probs[top_3_prize1_tens_indices]
    print("\n--- Top 3 ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏¥‡∏ö‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡πÇ‡∏≠‡∏Å‡∏≤‡∏™‡∏≠‡∏≠‡∏Å‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î ---")
    for i in range(len(top_3_prize1_tens_indices)):
        print(f"   - ‡πÄ‡∏•‡∏Ç {top_3_prize1_tens_indices[i]}: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô {top_3_prize1_tens_probs[i]*100:.2f}%")

    # ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 (‡πÄ‡∏î‡∏¥‡∏°)
    prize1_unit_probabilities = model_prize1_unit.predict_proba(future_features_df)[0]
    top_3_prize1_unit_indices = np.argsort(prize1_unit_probabilities)[::-1][:3]
    top_3_prize1_unit_probs = prize1_unit_probabilities[top_3_prize1_unit_indices]

    print("\n--- Top 3 ‡∏´‡∏•‡∏±‡∏Å‡∏´‡∏ô‡πà‡∏ß‡∏¢‡∏£‡∏≤‡∏á‡∏ß‡∏±‡∏•‡∏ó‡∏µ‡πà 1 ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡πÇ‡∏≠‡∏Å‡∏≤‡∏™‡∏≠‡∏≠‡∏Å‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î ---")
    for i in range(len(top_3_prize1_unit_indices)):
        print(f"   - ‡πÄ‡∏•‡∏Ç {top_3_prize1_unit_indices[i]}: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ô‡πà‡∏≤‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô {top_3_prize1_unit_probs[i]*100:.2f}%")


print("\n‚úÖ ‡∏Å‡∏≤‡∏£‡∏ù‡∏∂‡∏Å‡πÇ‡∏°‡πÄ‡∏î‡∏•, ‡∏õ‡∏£‡∏∞‡πÄ‡∏°‡∏¥‡∏ô‡∏ú‡∏•, ‡πÅ‡∏•‡∏∞‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏ô‡∏≤‡∏¢‡πÄ‡∏™‡∏£‡πá‡∏à‡∏™‡∏¥‡πâ‡∏ô.")
```
