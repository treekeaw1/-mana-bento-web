```python
import pandas as pd
from io import StringIO
from collections import defaultdict, Counter
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from datetime import datetime, timedelta

# Data provided by the user (corrected the problematic date entry)
data = """2025/07/01,949246,91
2025/06/16,507392,06
2025/06/01,559352,20
2025/05/16,251309,87
2025/05/02,213388,06
2025/04/16,266227,85
2025/04/01,669687,36
2025/03/16,757563,32
2025/03/01,818894,54
2025/02/16,847377,50
2025/02/01,558700,51
2025/01/17,807779,23
2025/01/02,730209,51
2024/12/16,097863,21
2024/12/01,669843,61
2024/11/16,187221,38
2024/11/01,536044,32
2024/10/16,482962,00
2024/10/01,718665,59
2024/09/16,608662,37
2024/09/01,199606,94
2024/08/16,095867,28
2024/08/01,407041,46
2024/07/16,367336,21
2024/07/01,434503,89
2024/06/16,518504,31
2024/06/01,530593,42
2024/05/16,205690,60
2024/05/02,980116,17
2024/04/16,943598,79
2024/04/01,803481,90
2024/03/16,997626,78
2024/03/01,253603,79
2024/02/16,941395,43
2024/02/01,607063,09
2024/01/17,105979,61
2023/12/30,625544,89
2023/12/16,356757,85
2023/12/01,251097,91
2023/11/16,557990,14
2023/11/01,743951,63
2023/10/16,931446,44
2023/10/01,727202,66
2023/09/16,320812,46
2023/09/01,915478,91
2023/08/16,471782,67
2023/07/31,260453,11
2023/07/16,169530,62
2023/07/01,922605,16
2023/06/16,264872,30
2023/06/01,125272,09
2023/05/16,132903,99
2023/05/02,843019,65
2023/04/16,984906,71
2023/04/01,087907,99
2023/03/16,025873,73
2023/03/01,417652,55
2023/02/16,590417,80
2023/02/01,297411,92
2023/01/17,812519,47
2022/12/30,157196,58
2022/12/16,845093,14
2022/12/01,375805,08
2022/11/16,121789,64
2022/11/01,913106,70
2022/10/16,613106,15
2022/10/01,484669,50
2022/09/16,943703,75
2022/09/01,929332,83
2022/08/16,331583,42
2022/08/01,436594,14
2022/07/16,620405,53
2022/07/01,981417,61
2022/06/16,361807,92
2022/06/01,319196,02
2022/05/16,155012,06
2022/05/02,658642,09
2022/04/16,395919,58
2022/04/01,970618,10
2022/03/16,737867,03
2022/03/01,061905,07
2022/02/17,098597,57
2022/02/01,944308,30
2022/01/17,880159,92
2021/12/30,819068,36
2021/12/16,639235,83
2021/12/01,077258,82
2021/11/16,032761,57
2021/11/01,045037,95
2021/10/16,386372,38
2021/10/01,578171,83
2021/09/16,070935,90
2021/09/01,114475,79
2021/08/16,046750,23
2021/08/01,910261,69
2021/07/16,556725,70
2021/07/01,713517,29
2021/06/16,691861,17
2021/06/01,292972,45
2021/05/16,684579,14
2021/05/02,501272,18
2021/04/16,100787,56
2021/04/01,472270,05
2021/03/16,890422,19
2021/03/01,835538,73
2021/02/16,424603,39
2021/02/01,912307,97
2021/01/17,384395,15
2020/12/30,803628,19
2020/12/16,201303,70
2020/12/01,100994,84
2020/11/16,972661,46
2020/11/01,506404,40
2020/10/16,286051,38
2020/10/01,837893,59
2020/09/16,244083,57
2020/09/01,999997,98
2020/08/16,945811,88
2020/08/01,569391,92
2020/07/16,873286,53
2020/07/01,347258,83
2020/06/16,516967,64
2020/06/01,831567,24
2020/04/01,051095,22
2020/03/16,503446,77
2020/03/01,875938,98
2020/02/16,781403,94
2020/02/01,589227,06
2020/01/17,491774,68
2019/12/30,510541,81
2019/12/16,529924,97
2019/12/01,453522,81
2019/11/16,017223,32
2019/11/01,967375,79
2019/10/16,812564,15
2019/10/01,691197,59
2019/09/16,340388,85
2019/09/01,798787,20
2019/08/16,775476,89
2019/08/01,387006,58
2019/07/15,369765,88
2019/07/01,943647,86
2019/06/16,174055,29
2019/06/01,516461,46
2019/05/16,962526,71
2019/05/02,061324,25
2019/04/16,570331,23
2019/04/01,109767,52
2019/03/16,724628,64
2019/03/01,345650,65
2019/02/16,074824,56
2019/02/01,967134,04
2019/01/17,197079,65
2018/12/30,735867,02
2018/12/16,356564,62
2018/12/01,021840,67
2018/11/16,989903,16
2018/11/01,149840,58
2018/10/16,200515,93
2018/10/01,452643,99
2018/09/16,149760,79
2018/09/01,734510,26
2018/08/16,586117,10
2018/08/01,386602,78
2018/07/16,596324,27
2018/07/01,963623,83
2018/06/16,223131,46
2018/06/01,988117,95
2018/05/16,075629,20
2018/05/02,248038,85
2018/04/16,739229,60
2018/04/01,412073,85
2018/03/16,218559,82
2018/03/02,759415,29
2018/02/16,309915,39
2018/02/01,026853,31
2018/01/17,203823,50
2017/12/30,911234,98
2017/12/16,955596,89
2017/12/01,451005,33
2017/11/16,292391,98
2017/11/01,533726,85
2017/10/16,413494,86
2017/10/01,880714,52
2017/09/16,170143,71
2017/09/01,143224,65
2017/08/16,715431,37
2017/08/01,756519,36
2017/07/16,820327,87
2017/07/01,112360,26
2017/06/16,943142,47
2017/06/01,053630,61
2017/05/16,454891,53
2017/05/02,008656,35
2017/04/16,816729,40
2017/04/01,392785,80
2017/03/16,273863,92
2017/03/01,978453,78
2017/02/16,229116,14
2017/02/01,054672,42
2017/01/17,145157,25
2016/12/30,377712,46
2016/12/16,435286,35
2016/12/01,086069,77
2016/11/16,858383,44
2016/11/01,785438,86
2016/10/16,571947,98
2016/10/01,887102,33
2016/09/16,240650,42
2016/09/01,638684,62
2016/08/16,254004,33
2016/08/01,272932,57
2016/07/16,449764,55
2016/07/01,082460,53
2016/06/16,073816,79
2016/06/01,511825,14
2016/05/16,141737,98
2016/05/02,399459,02
2016/04/16,221609,87
2014/04/01,066720,92
2014/03/16,134918,32
2014/03/01,439686,06
2014/02/16,356364,98
2014/02/01,927800,09
2014/01/17,304371,50
2013/12/30,008217,02
2013/12/17,930255,08
2013/12/01,915350,78
2013/11/16,795283,03
2013/11/01,361211,45
2013/10/16,968630,62
2013/10/01,594825,07
2013/09/16,743148,06
2013/09/01,021094,89
2013/08/16,033363,40
2013/08/01,518677,53
2013/07/16,121507,49
2013/07/01,759049,26
2013/06/16,644742,05
2013/06/02,388881,65
2013/05/16,011421,38
2013/05/02,543466,30
2013/04/16,506260,38
2013/04/01,605704,70
2013/03/16,048151,92
2013/03/01,240237,34
2013/02/16,001864,90
2013/02/01,155537,79
2013/01/16,244351,74
2012/12/30,461704,57
2012/12/16,948354,90
2012/12/01,480449,11
2012/11/16,479804,25
2012/11/01,206608,44
2012/10/16,656409,94
2012/10/01,375615,44
2012/09/16,772269,35
2012/09/01,856763,22
2012/08/16,662842,91
2012/08/01,766391,82
2012/07/16,468728,45
2012/07/01,378477,39
2012/06/16,673920,95
2012/06/01,781198,18
2012/05/16,087523,20
2012/05/02,103297,52
2012/04/16,153406,26
2012/04/01,028866,95
2012/03/16,531404,79
2012/03/01,906318,35
2012/02/16,384245,01
2012/02/01,180149,95
2012/01/16,306902,52
2011/12/30,561072,48
2011/12/16,341767,79
2011/12/01,168795,27
2011/11/16,806925,28
2011/11/01,739804,47
2011/10/16,963289,60
2011/10/01,647882,14
2011/09/16,562684,63
2011/09/01,548123,05
2011/08/16,321327,20
2011/08/01,356435,82
2011/07/16,566996,86
2011/07/01,646905,51
2011/06/16,289673,69
2011/06/01,935489,90
2011/05/16,687125,56
2011/05/02,603458,07
2011/04/16,843846,86
2011/04/01,571688,53
2011/03/16,968433,52
2011/03/01,976241,37
2011/02/16,368257,09
2011/02/01,565566,66
2011/01/16,820981,08
2010/12/30,302358,00
2010/12/16,529524,72
2010/12/01,110443,43
2010/11/16,639500,15
2010/11/01,524694,63
2010/10/16,281343,28
2010/10/01,124025,58
2010/09/16,540143,79
2010/09/01,329997,07
2010/08/16,683877,28
2010/08/01,895590,50
2010/07/16,904050,11
2010/07/01,915900,60
2010/06/16,159373,51
2010/06/01,882727,38
2010/05/16,814418,31
2010/05/02,889501,29
2010/04/16,583470,62
2010/04/01,257562,69
2010/03/16,607064,08
2010/03/01,222518,79
2010/02/16,648684,18
2010/02/01,320605,32
2010/01/16,451445,81
2009/12/30,526402,65
2009/12/16,884178,21
2009/12/01,408147,02
2009/11/16,997777,57
2009/11/01,805540,54
2009/10/16,955756,83
2009/10/01,511052,15
2009/09/16,731198,28
2009/09/01,724533,85
2009/08/16,536960,62
2009/08/01,218756,12
2009/07/16,116556,12
2009/07/01,622953,51
2009/06/16,351276,88
2009/06/01,562370,46
2009/05/16,406417,05
2009/05/02,054136,85
2009/04/16,825988,44
2009/04/01,814931,01
2009/03/16,593331,96
2009/03/01,656037,97
2009/02/16,481746,27
2009/02/01,610089,55
2009/01/16,281062,23
2008/12/30,884112,49
2008/12/16,334380,24
2008/12/01,181752,09
2008/11/16,813993,43
2008/11/01,191100,59
2008/10/16,621377,42
2008/10/01,488372,02
2008/09/16,017422,66
2008/09/01,354656,11
2008/08/16,911097,64
2008/08/01,210008,10
2008/07/16,180387,34
2008/07/01,480239,68
2008/06/16,500104,73
2008/06/01,444874,81
2008/05/16,480012,12
2008/05/02,360371,06
2008/04/16,211743,96
2008/04/01,959517,22
2008/03/16,364222,97
2008/03/01,215227,97
2008/02/16,133707,03
2008/02/01,186312,14
2008/01/16,073577,67
2007/12/30,994304,87
2007/12/16,685141,05
2007/12/01,776980,59
2007/11/16,055986,58
2007/11/01,689140,85
2007/10/16,258487,00
2007/10/01,169387,06"""

# Read data into a DataFrame
df = pd.read_csv(StringIO(data), header=None, names=['Date', 'Main_Prize', 'Last_Two_Digits'])

# Convert 'Date' to datetime and sort (ensure newest data is at the top for easy slicing)
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')
df = df.sort_values(by='Date', ascending=False).reset_index(drop=True)

# --- Feature Engineering ---
def create_features(df_input):
    df = df_input.copy() # Work on a copy

    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day
    df['DayOfWeek'] = df['Date'].dt.dayofweek
    df['DayOfMonth'] = df['Date'].dt.day
    df['DayOfYear'] = df['Date'].dt.dayofyear
    df['WeekOfYear'] = df['Date'].dt.isocalendar().week.astype(int)

    # Convert to string and pad with zeros for consistent digit extraction
    # Handle potential non-numeric data before conversion to string
    df['Main_Prize_str'] = df['Main_Prize'].fillna(0).astype(int).astype(str).str.zfill(6) # Ensure 6 digits
    df['Last_Two_Digits_str'] = df['Last_Two_Digits'].fillna(0).astype(int).astype(str).str.zfill(2) # Ensure 2 digits

    # Extract individual digits
    # Use .str.strip() to remove any leading/trailing spaces
    df['Main_Prize_D1'] = df['Main_Prize_str'].str[0].astype(int)
    df['Main_Prize_D2'] = df['Main_Prize_str'].str[1].astype(int)
    df['Main_Prize_D3'] = df['Main_Prize_str'].str[2].astype(int)
    df['Main_Prize_D4'] = df['Main_Prize_str'].str[3].astype(int)
    df['Main_Prize_D5'] = df['Main_Prize_str'].str[4].astype(int)
    df['Main_Prize_D6'] = df['Main_Prize_str'].str[5].astype(int)

    df['Last_Two_D1'] = df['Last_Two_Digits_str'].str[0].astype(int)
    df['Last_Two_D2'] = df['Last_Two_Digits_str'].str[1].astype(int)

    # Additional features: Sum of digits
    # Ensure only digits are summed, handling potential spaces/non-digits
    df['Main_Prize_Sum'] = df['Main_Prize_str'].apply(lambda x: sum(int(d) for d in x if d.isdigit()))
    df['Last_Two_Sum'] = df['Last_Two_Digits_str'].apply(lambda x: sum(int(d) for d in x if d.isdigit()))

    # Features based on previous draws (Lag features)
    for i in range(1, 4): # Lag by 1, 2, and 3 draws
        df[f'Prev_Main_Prize_Sum_{i}'] = df['Main_Prize_Sum'].shift(i)
        df[f'Prev_Last_Two_Sum_{i}'] = df['Last_Two_Sum'].shift(i)
        df[f'Prev_Last_Two_D1_{i}'] = df['Last_Two_D1'].shift(i)
        df[f'Prev_Last_Two_D2_{i}'] = df['Last_Two_D2'].shift(i)

    # Rolling window features
    window_sizes = [3, 5, 7]
    for window in window_sizes:
        # Shift by 1 to ensure we are using historical data *before* the current row
        df[f'Rolling_Main_Prize_Sum_Mean_{window}'] = df['Main_Prize_Sum'].rolling(window=window).mean().shift(1)
        df[f'Rolling_Last_Two_Sum_Mean_{window}'] = df['Last_Two_Sum'].rolling(window=window).mean().shift(1)

    # Drop temporary string columns
    df = df.drop(columns=['Main_Prize_str', 'Last_Two_Digits_str'])

    # Drop rows with NaN values created by shifting/rolling (do this at the very end of feature creation)
    return df.dropna()

# Create features for the entire dataset
df_features = create_features(df.copy()) # Use a copy to avoid SettingWithCopyWarning

# Separate features (X) and targets (y)
X_main = df_features.drop(columns=['Date', 'Main_Prize', 'Last_Two_Digits',
                                   'Main_Prize_D1', 'Main_Prize_D2', 'Main_Prize_D3',
                                   'Main_Prize_D4', 'Main_Prize_D5', 'Main_Prize_D6',
                                   'Last_Two_D1', 'Last_Two_D2', 'Main_Prize_Sum', 'Last_Two_Sum'])

y_last_two_d1 = df_features['Last_Two_D1']
y_last_two_d2 = df_features['Last_Two_D2']

y_main_d4 = df_features['Main_Prize_D4']
y_main_d5 = df_features['Main_Prize_D5']
y_main_d6 = df_features['Main_Prize_D6']


# --- Model Training (RandomForestClassifier for each digit) ---

def train_and_predict_digit_model(X, y, test_data_row):
    """Trains a RandomForestClassifier and predicts a single digit."""
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X, y) # Train on all available data for the best prediction

    # Ensure test_data_row has all the features the model was trained on
    test_data_for_prediction = test_data_row[X.columns] # Use X.columns for the correct order

    probabilities = model.predict_proba(test_data_for_prediction)

    # Get feature importances
    feature_importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)

    # Get predicted digit (most probable)
    predicted_digit = model.predict(test_data_for_prediction)[0]

    # Get top N most probable digits based on probabilities
    top_n = 3 # For example, top 3 most probable digits

    # The classes_ attribute holds the unique classes seen by the model (e.g., 0-9)
    # Sort probabilities for the single prediction row
    sorted_probs_idx = np.argsort(probabilities[0])[::-1] # Sort descending
    top_digits_probs = []
    for i in sorted_probs_idx:
        digit = model.classes_[i]
        prob = probabilities[0][i]
        top_digits_probs.append((digit, prob))

    return predicted_digit, top_digits_probs[:top_n], feature_importances # Return only top N

# --- Prediction for the next draw ---
print("--- เริ่มต้นการทำนายสำหรับงวดถัดไป ---")

# Define the date for the next draw (User input)
next_draw_date_str = "2025-07-01"
next_draw_date = pd.to_datetime(next_draw_date_str)
print(f"จะทำนายสำหรับงวดวันที่: {next_draw_date.strftime('%Y-%m-%d')}")
print("-" * 75)

# Prepare features for the next draw
# We need the last available row from df_features for prediction
last_known_data = df_features.iloc[0].copy() # Most recent data for lag features

# Create a DataFrame for the next draw's features with the correct structure
next_draw_features_df = pd.DataFrame(index=[0], columns=X_main.columns)

# Update date-related features for the next draw's date
next_draw_features_df.loc[0, 'Year'] = next_draw_date.year
next_draw_features_df.loc[0, 'Month'] = next_draw_date.month
next_draw_features_df.loc[0, 'Day'] = next_draw_date.day
next_draw_features_df.loc[0, 'DayOfWeek'] = next_draw_date.weekday()
next_draw_features_df.loc[0, 'DayOfMonth'] = next_draw_date.day
next_draw_features_df.loc[0, 'DayOfYear'] = next_draw_date.dayofyear
next_draw_features_df.loc[0, 'WeekOfYear'] = next_draw_date.isocalendar().week

# Shift the 'previous' features for the prediction row.
next_draw_features_df.loc[0, 'Prev_Main_Prize_Sum_1'] = last_known_data['Main_Prize_Sum']
next_draw_features_df.loc[0, 'Prev_Last_Two_Sum_1'] = last_known_data['Last_Two_Sum']
next_draw_features_df.loc[0, 'Prev_Last_Two_D1_1'] = last_known_data['Last_Two_D1']
next_draw_features_df.loc[0, 'Prev_Last_Two_D2_1'] = last_known_data['Last_Two_D2']

next_draw_features_df.loc[0, 'Prev_Main_Prize_Sum_2'] = last_known_data['Prev_Main_Prize_Sum_1']
next_draw_features_df.loc[0, 'Prev_Last_Two_Sum_2'] = last_known_data['Prev_Last_Two_Sum_1']
next_draw_features_df.loc[0, 'Prev_Last_Two_D1_2'] = last_known_data['Prev_Last_Two_D1_1']
next_draw_features_df.loc[0, 'Prev_Last_Two_D2_2'] = last_known_data['Prev_Last_Two_D2_1']

next_draw_features_df.loc[0, 'Prev_Main_Prize_Sum_3'] = last_known_data['Prev_Main_Prize_Sum_2']
next_draw_features_df.loc[0, 'Prev_Last_Two_Sum_3'] = last_known_data['Prev_Last_Two_Sum_2']
next_draw_features_df.loc[0, 'Prev_Last_Two_D1_3'] = last_known_data['Prev_Last_Two_D1_2']
next_draw_features_df.loc[0, 'Prev_Last_Two_D2_3'] = last_known_data['Prev_Last_Two_D2_2']

# Recalculate rolling means for the prediction row.
for window in [3, 5, 7]:
    next_draw_features_df.loc[0, f'Rolling_Main_Prize_Sum_Mean_{window}'] = df_features[f'Rolling_Main_Prize_Sum_Mean_{window}'].iloc[0]
    next_draw_features_df.loc[0, f'Rolling_Last_Two_Sum_Mean_{window}'] = df_features[f'Rolling_Last_Two_Sum_Mean_{window}'].iloc[0]

# Ensure prediction input has the same columns as training data and correct dtypes
prediction_input = next_draw_features_df.astype(X_main.dtypes)

# --- Generate Predictions ---

print("\n--- ผลการทำนายเลข 2 ตัวล่าง ---")
pred_d1_l2, top_probs_d1_l2, imp_d1_l2 = train_and_predict_digit_model(X_main, y_last_two_d1, prediction_input)
pred_d2_l2, top_probs_d2_l2, imp_d2_l2 = train_and_predict_digit_model(X_main, y_last_two_d2, prediction_input)

print(f"  **หลักสิบ (เลข 2 ตัวล่าง):** ทำนายคือ **{pred_d1_l2}**")
print("  Top 3 ตัวเลขที่มีโอกาสสูงสุด:")
for digit, prob in top_probs_d1_l2:
    print(f"    - เลข {digit} (ความน่าจะเป็น: {prob:.2%})")

print(f"\n  **หลักหน่วย (เลข 2 ตัวล่าง):** ทำนายคือ **{pred_d2_l2}**")
print("  Top 3 ตัวเลขที่มีโอกาสสูงสุด:")
for digit, prob in top_probs_d2_l2:
    print(f"    - เลข {digit} (ความน่าจะเป็น: {prob:.2%})")

print(f"\n  **เลข 2 ตัวล่างที่ทำนาย:** **{pred_d1_l2}{pred_d2_l2}**")
print("-" * 75)

print("\n--- ผลการทำนายเลขท้าย 3 ตัวของรางวัลที่ 1 ---")
pred_d4_m, top_probs_d4_m, imp_d4_m = train_and_predict_digit_model(X_main, y_main_d4, prediction_input)
pred_d5_m, top_probs_d5_m, imp_d5_m = train_and_predict_digit_model(X_main, y_main_d5, prediction_input)
pred_d6_m, top_probs_d6_m, imp_d6_m = train_and_predict_digit_model(X_main, y_main_d6, prediction_input)

print(f"  **หลักร้อย (เลขท้าย 3 ตัว):** ทำนายคือ **{pred_d4_m}**")
print("  Top 3 ตัวเลขที่มีโอกาสสูงสุด:")
for digit, prob in top_probs_d4_m:
    print(f"    - เลข {digit} (ความน่าจะเป็น: {prob:.2%})")

print(f"\n  **หลักสิบ (เลขท้าย 3 ตัว):** ทำนายคือ **{pred_d5_m}**")
print("  Top 3 ตัวเลขที่มีโอกาสสูงสุด:")
for digit, prob in top_probs_d5_m:
    print(f"    - เลข {digit} (ความน่าจะเป็น: {prob:.2%})")

print(f"\n  **หลักหน่วย (เลขท้าย 3 ตัว):** ทำนายคือ **{pred_d6_m}**")
print("  Top 3 ตัวเลขที่มีโอกาสสูงสุด:")
for digit, prob in top_probs_d6_m:
    print(f"    - เลข {digit} (ความน่าจะเป็น: {prob:.2%})")

print(f"\n  **เลขท้าย 3 ตัวของรางวัลที่ 1 ที่ทำนาย:** **{pred_d4_m}{pred_d5_m}{pred_d6_m}**")
print("-" * 75)


print("\n--- Top 5 Features ที่มีผลต่อการทำนาย (ตัวอย่าง) ---")
print("\n  **สำหรับการทำนายหลักสิบเลข 2 ตัวล่าง:**")
print(imp_d1_l2.head())
print("\n  **สำหรับการทำนายหลักหน่วยเลข 2 ตัวล่าง:**")
print(imp_d2_l2.head())
print("\n  **สำหรับการทำนายหลักร้อยเลขท้าย 3 ตัวรางวัลที่ 1:**")
print(imp_d4_m.head())
print("\n  **สำหรับการทำนายหลักสิบเลขท้าย 3 ตัวรางวัลที่ 1:**")
print(imp_d5_m.head())
print("\n  **สำหรับการทำนายหลักหน่วยเลขท้าย 3 ตัวรางวัลที่ 1:**")
print(imp_d6_m.head())
print("-" * 75)

print("\n## ข้อควรทราบ:")
print("- การทำนายนี้ใช้โมเดล Machine Learning (Random Forest) ซึ่งเรียนรู้จากรูปแบบในข้อมูลในอดีตที่คุณให้มาเท่านั้น")
print("- ผลลัพธ์ที่ได้เป็นการบอก 'แนวโน้ม' หรือ 'ความเป็นไปได้สูงสุด' ตามที่โมเดลเรียนรู้ ไม่ใช่การรับประกันผลลัพธ์")
print("- ลอตเตอรี่เป็นการสุ่ม และไม่มีวิธีการทำนายที่แม่นยำ 100%")
print("- การลงทุนมีความเสี่ยง โปรดใช้วิจารณญาณในการตัดสินใจ")
```
