```python
# ---- 0. เตรียมไลบรารี ----
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter, defaultdict
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.utils import to_categorical
import warnings

# ปิด warning สำหรับ to_categorical (เป็นข้อควรระวังเรื่องค่าที่หายไป)
warnings.filterwarnings('ignore', category=FutureWarning, module='tensorflow')

# --- ข้อมูล CSV ที่เราใช้มาตลอด (แก้ไข '22003' เป็น '2003' และตัดบรรทัดว่างแรกออก) ---
csv_content = """วันที่,รางวัลที่ 1 (6 หลัก),เลข 2 ตัวล่าง
2025/06/01,559352,20
2025/05/16,251309,87
2025/05/02,213388,06
2025/04/16,266227,85
2025/04/01,669687,36
2025/03/16,757563,32
2025/03/01,818894,54
2025/02/16,847377,50
2025/02/01,558700,51
2025/01/17,807779,23
2025/01/02,730209,51
2024/12/16,097863,21
2024/12/01,669843,61
2024/11/16,187221,38
2024/11/01,536044,32
2024/10/16,482962,00
2024/10/01,718665,59
2024/09/16,608662,37
2024/09/01,199606,94
2024/08/16,095867,28
2024/08/01,407041,46
2024/07/16,367336,21
2003/11/16,238511,68
2003/11/01,941438,47
2003/10/16,305500,03
2003/10/01,912040,43
2003/09/16,600589,53
2003/09/01,187813,92
2003/08/16,354771,00
2003/08/01,766098,91
2003/07/16,679545,75
"""

# ---- 1. โหลดข้อมูล ----
df = pd.read_csv(io.StringIO(csv_content.strip()), header=0)
df.columns = ['date','prize1','bottom2']

df['prize1'] = df['prize1'].astype(str).str.zfill(6)
df['bottom2'] = df['bottom2'].astype(str).str.zfill(2)

df['date'] = pd.to_datetime(df['date'], format='%Y/%m/%d')
df = df.sort_values(by='date', ascending=True).reset_index(drop=True)

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['prize1_3d'] = df['prize1'].apply(lambda x: str(x)[-3:])

# ---- 2. Feature Engineering ----
df_numeric_for_feat = df.copy()
df_numeric_for_feat['bottom2_int'] = df_numeric_for_feat['bottom2'].astype(int)
df_numeric_for_feat['prize1_3d_int'] = df_numeric_for_feat['prize1_3d'].astype(int)

for lag in range(1, 6):
    df_numeric_for_feat[f'lag2_{lag}'] = df_numeric_for_feat['bottom2_int'].shift(lag)
    df_numeric_for_feat[f'lag3_{lag}'] = df_numeric_for_feat['prize1_3d_int'].shift(lag)

# แก้ไข: เปลี่ยน x.iloc[-1] เป็น x[-1] สำหรับ raw=True
df_numeric_for_feat['freq2_last_10'] = df_numeric_for_feat['bottom2_int'].rolling(window=10).apply(lambda x: (x==x[-1]).sum(), raw=True)
df_numeric_for_feat['freq3_last_10'] = df_numeric_for_feat['prize1_3d_int'].rolling(window=10).apply(lambda x: (x==x[-1]).sum(), raw=True)

# แก้ไข: เปลี่ยน x.iloc[-1] เป็น x[-1] สำหรับ raw=True
df_numeric_for_feat['gap2'] = df_numeric_for_feat['bottom2_int'].rolling(window=10).apply(
    lambda x: (np.arange(len(x))[::-1][x==x[-1]][0] if np.any(x==x[-1]) else 10), raw=True
)
df_numeric_for_feat['gap3'] = df_numeric_for_feat['prize1_3d_int'].rolling(window=10).apply(
    lambda x: (np.arange(len(x))[::-1][x==x[-1]][0] if np.any(x==x[-1]) else 10), raw=True
)

df_numeric_for_feat['weekday'] = df_numeric_for_feat['date'].dt.weekday
df_numeric_for_feat = df_numeric_for_feat.dropna().reset_index(drop=True)

df = df_numeric_for_feat.copy()

print("DataFrame หลังจาก Feature Engineering (5 แถวล่าสุด):")
print(df.tail())

# ---- 3. Visualization ----
plt.figure(figsize=(10,4))
sns.histplot(df['bottom2_int'], bins=100, color='skyblue', kde=True)
plt.title('Frequency of 2-digit bottom numbers')
plt.xlabel('2-digit')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10,4))
sns.histplot(df['prize1_3d_int'], bins=100, color='salmon', kde=True)
plt.title('Frequency of 3-digit last numbers')
plt.xlabel('3-digit')
plt.ylabel('Count')
plt.show()

# ---- 4. RandomForest (2 ตัวล่าง & 3 ตัวท้าย) ----
feature2_cols = [c for c in df.columns if c.startswith('lag2_')] + ['year','month','day','freq2_last_10','gap2','weekday']
X2 = df[feature2_cols]
y2 = df['bottom2_int']
X2_train = X2.iloc[:-1]
y2_train = y2.iloc[:-1]
X2_predict_next = X2.iloc[-1:].copy()

rf2 = RandomForestClassifier(n_estimators=300, random_state=42, class_weight='balanced')
rf2.fit(X2_train, y2_train)
proba2 = rf2.predict_proba(X2_predict_next)[0]
rf2_top = np.argsort(proba2)[-4:][::-1]
rf2_pred_num = [rf2.classes_[i] for i in rf2_top]
rf2_pred = [str(x).zfill(2) for x in rf2_pred_num]

feature3_cols = [c for c in df.columns if c.startswith('lag3_')] + ['year','month','day','freq3_last_10','gap3','weekday']
X3 = df[feature3_cols]
y3 = df['prize1_3d_int']
X3_train = X3.iloc[:-1]
y3_train = y3.iloc[:-1]
X3_predict_next = X3.iloc[-1:].copy()

rf3 = RandomForestClassifier(n_estimators=300, random_state=42, class_weight='balanced')
rf3.fit(X3_train, y3_train)
proba3 = rf3.predict_proba(X3_predict_next)[0]
rf3_top = np.argsort(proba3)[-4:][::-1]
rf3_pred_num = [rf3.classes_[i] for i in rf3_top]
rf3_pred = [str(x).zfill(3) for x in rf3_pred_num]


# ---- 5. LSTM (2 ตัวล่าง & 3 ตัวท้าย) ----
def create_sequences(data, seq_length=5):
    X, y = [], []
    for i in range(len(data)-seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 5
le2 = LabelEncoder()
data2d_int = df['bottom2_int'].values
le2.fit(data2d_int)
data2d = le2.transform(data2d_int)

X2d, y2d = create_sequences(data2d, seq_length)
num_classes_2d = len(le2.classes_)
y2d_cat = to_categorical(y2d, num_classes=num_classes_2d)

model2d = Sequential([
    Embedding(input_dim=num_classes_2d, output_dim=16, input_length=seq_length),
    LSTM(64, return_sequences=False),
    Dense(num_classes_2d, activation='softmax')
])
model2d.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

if len(X2d) > 0:
    model2d.fit(X2d, y2d_cat, epochs=50, batch_size=16, verbose=0)
    last_seq2 = data2d[-seq_length:].reshape(1,-1)
    pred_prob2 = model2d.predict(last_seq2, verbose=0)[0]
    lstm2_top = np.argsort(pred_prob2)[-4:][::-1]
    lstm2_pred_num = le2.inverse_transform(lstm2_top)
    lstm2_pred = [str(x).zfill(2) for x in lstm2_pred_num]
else:
    lstm2_pred = []
    pred_prob2 = np.array([])
    print("Warning: Not enough data to train LSTM for 2-digit numbers.")

le3 = LabelEncoder()
data3d_int = df['prize1_3d_int'].values
le3.fit(data3d_int)
num_classes_3d = len(le3.classes_)
data3d = le3.transform(data3d_int)

X3d, y3d = create_sequences(data3d, seq_length)
y3d_cat = to_categorical(y3d, num_classes=num_classes_3d)

model3d = Sequential([
    Embedding(input_dim=num_classes_3d, output_dim=32, input_length=seq_length),
    LSTM(64, return_sequences=False),
    Dense(num_classes_3d, activation='softmax')
])
model3d.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

if len(X3d) > 0:
    model3d.fit(X3d, y3d_cat, epochs=50, batch_size=16, verbose=0)
    last_seq3 = data3d[-seq_length:].reshape(1,-1)
    pred_prob3 = model3d.predict(last_seq3, verbose=0)[0]
    lstm3_top = np.argsort(pred_prob3)[-4:][::-1]
    lstm3_pred_num = le3.inverse_transform(lstm3_top)
    lstm3_pred = [str(x).zfill(3) for x in lstm3_pred_num]
else:
    lstm3_pred = []
    pred_prob3 = np.array([])
    print("Warning: Not enough data to train LSTM for 3-digit numbers.")


# ---- 6. สรุปผลการทำนาย "งวดถัดไป" ----
print("\n--- การทำนายสำหรับ 'งวดถัดไป' (หลังจาก 2025-06-01) ---")

print("\n===== ผลทำนายเลข 2 ตัวล่าง (4 ชุด) =====")
print("RandomForest:", rf2_pred)
print("LSTM:", lstm2_pred)

print("\n===== ผลทำนายเลข 3 ตัวท้ายรางวัลที่ 1 (4 ชุด) =====")
print("RandomForest:", rf3_pred)
print("LSTM:", lstm3_pred)

print("\n--- ความมั่นใจ (Probability) ของการทำนาย ---")
if len(rf2_top) > 0:
    print(f"ความมั่นใจ (prob) RF 2 ตัวล่างสำหรับ {rf2_pred}: {[f'{proba2[i]:.4f}' for i in rf2_top]}")
if len(lstm2_top) > 0:
    print(f"ความมั่นใจ (prob) LSTM 2 ตัวล่างสำหรับ {lstm2_pred}: {[f'{pred_prob2[i]:.4f}' for i in lstm2_top]}")
if len(rf3_top) > 0:
    print(f"ความมั่นใจ (prob) RF 3 ตัวท้ายสำหรับ {rf3_pred}: {[f'{proba3[i]:.4f}' for i in rf3_top]}")
if len(lstm3_top) > 0:
    print(f"ความมั่นใจ (prob) LSTM 3 ตัวท้ายสำหรับ {lstm3_pred}: {[f'{pred_prob3[i]:.4f}' for i in lstm3_top]}")

print("\n--- สรุปผลทำนายทั้งหมดสำหรับงวดถัดไป ---")
print("เลข 2 ตัวล่างที่คาดการณ์ (รวมจาก RF และ LSTM):")
all_pred_2d = list(set(rf2_pred + lstm2_pred))
print(f"  > {all_pred_2d}")

print("เลข 3 ตัวท้ายรางวัลที่ 1 ที่คาดการณ์ (รวมจาก RF และ LSTM):")
all_pred_3d = list(set(rf3_pred + lstm3_pred))
print(f"  > {all_pred_3d}")
```
