```python
# ==============================================================================
# --- Lotto Analyzer V11.5 - เซลล์ที่ 2/2 ---
# --- Fix: KeyError by moving temp column creation into _create_all_features ---
# ==============================================================================
import pandas as pd
import numpy as np
from math import sin, cos, pi
from collections import Counter
import warnings
import matplotlib.pyplot as plt

# --- AI & ML Tools ---
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
from sklearn.base import BaseEstimator, ClassifierMixin
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical

warnings.filterwarnings('ignore')

# --- Custom XGBoost Wrapper to enforce all classes (retained from V10) ---
class XGBoostFixWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, num_class, estimator=None, **kwargs):
        self.num_class = num_class
        if estimator is None:
            self.estimator = xgb.XGBClassifier(
                n_estimators=100,
                objective='multi:softprob',
                eval_metric='mlogloss',
                num_class=self.num_class,
                random_state=42,
                use_label_encoder=False,
                enable_categorical=True,
                **kwargs
            )
        else:
            self.estimator = estimator
            self.estimator.num_class = self.num_class
            self.estimator.objective = 'multi:softprob'
            self.estimator.eval_metric = 'mlogloss'
            self.estimator.use_label_encoder = False
            self.estimator.enable_categorical = True

        self.classes_ = np.arange(self.num_class)

    def fit(self, X, y, **kwargs):
        unique_y_classes = np.unique(y)
        if not np.array_equal(np.sort(unique_y_classes), self.classes_):
            missing_classes = np.setdiff1d(self.classes_, unique_y_classes)
            if len(missing_classes) > 0:
                dummy_X_missing = pd.DataFrame(np.zeros((len(missing_classes), X.shape[1])), columns=X.columns)
                dummy_y_missing = missing_classes
                X = pd.concat([X, dummy_X_missing], ignore_index=True)
                y = np.concatenate([y, dummy_y_missing])

        self.estimator.fit(X, y, **kwargs)
        if not hasattr(self.estimator, 'classes_') or not np.array_equal(self.estimator.classes_, self.classes_):
            self.estimator.classes_ = self.classes_
        return self

    def predict(self, X):
        return self.estimator.predict(X)

    def predict_proba(self, X):
        proba = self.estimator.predict_proba(X)
        if proba.shape[1] != self.num_class:
            full_proba = np.zeros((proba.shape[0], self.num_class))
            if hasattr(self.estimator, 'classes_') and np.array_equal(self.estimator.classes_, self.classes_):
                full_proba = proba
            else:
                warnings.warn(f"XGBoostFixWrapper: predict_proba received unexpected shape ({proba.shape[1]}) for num_class ({self.num_class}). Attempting to pad with zeros. This indicates potential underlying issue.")
                for i, c in enumerate(self.estimator.classes_):
                    if c < self.num_class:
                        full_proba[:, c] = proba[:, i]
                proba = full_proba
        return proba

# --- Main System Class ---
class UltimateLotterySystem:
    def __init__(self, cleaned_df):
        print("🚀 กำลังเริ่มต้น Ultimate Lottery Analysis & Prediction System V11.5...")
        if cleaned_df is None or cleaned_df.empty:
            raise ValueError("❗ ไม่พบข้อมูล 'df_cleaned' ที่ถูกต้อง โปรดรันเซลล์ที่ 1 ก่อน")

        # --- Moved target column cleaning here, before feature creation ---
        # This ensures 'เลขท้าย 2 ตัว_temp' and 'รางวัลที่ 1_temp' exist before _create_all_features uses them.
        self.df = cleaned_df.copy() # Use self.df for the raw df
        self.df['เลขท้าย 2 ตัว_temp'] = self.df['เลขท้าย 2 ตัว'].apply(
            lambda x: int(x) if pd.notna(x) and str(x).isdigit() and 0 <= int(x) <= 99 else np.nan
        )
        def extract_3d_safe(value):
            s_val = str(value)
            if pd.notna(value) and len(s_val) >= 3 and s_val[-3:].isdigit():
                num = int(s_val[-3:])
                if 0 <= num <= 999:
                    return num
            return np.nan
        self.df['รางวัลที่ 1_temp'] = self.df['รางวัลที่ 1'].apply(extract_3d_safe)

        initial_rows = len(self.df)
        self.df.dropna(subset=['เลขท้าย 2 ตัว_temp', 'รางวัลที่ 1_temp'], inplace=True)
        self.df.reset_index(drop=True, inplace=True)

        if initial_rows > len(self.df):
            print(f"🗑️ ลบไป {initial_rows - len(self.df)} แถวเนื่องจากข้อมูลเป้าหมายไม่สมบูรณ์/ไม่ถูกต้อง")

        if self.df.empty:
            raise ValueError("❗ หลังจากทำความสะอาดข้อมูลเป้าหมายแล้ว ไม่มีข้อมูลที่เหลืออยู่สำหรับฝึกโมเดล โปรดตรวจสอบข้อมูลดิบ.")

        # Now create features using the df with temp columns
        self.df_features = self._create_all_features(self.df)

        self.y_2d_raw = self.df_features['เลขท้าย 2 ตัว_temp'].astype(int)
        self.y_3d_raw = self.df_features['รางวัลที่ 1_temp'].astype(int)

        self.sequences_2d, self.targets_2d = self._create_sequences(self.y_2d_raw, 100)
        self.sequences_3d, self.targets_3d = self._create_sequences(self.y_3d_raw, 1000)

        self.final_model_2d = None
        self.final_model_3d = None

    def _create_rolling_digit_pool(self, series_of_lists, window_size):
        historic_pools = []
        for i in range(len(series_of_lists)):
            start_index = max(0, i - window_size)
            window = series_of_lists.iloc[start_index:i]
            current_pool = set()
            for item in window:
                if isinstance(item, (list, set)):
                    current_pool.update(item)
            historic_pools.append(current_pool)
        return historic_pools

    # ➕ Phase 2: Feature Engineering
    def _create_all_features(self, df):
        print("\n--- ระยะที่ 2: การสร้างฟีเจอร์และวิเคราะห์เชิงคณิตศาสตร์ ---")
        df['day_of_year_sin'] = np.sin(2 * pi * df['งวดวันที่'].dt.dayofyear / 365.25)
        df['day_of_year_cos'] = np.cos(2 * pi * df['งวดวันที่'].dt.dayofyear / 365.25)
        df['days_since_prev_draw'] = df['งวดวันที่'].diff().dt.days.fillna(0)

        df['all_digits_prize1'] = df['รางวัลที่ 1'].apply(
            lambda x: [int(d) for d in str(x) if d.isdigit()] if pd.notna(x) else []
        )
        df['all_digits_bottom2'] = df['เลขท้าย 2 ตัว'].apply(
            lambda x: [int(d) for d in str(x).zfill(2) if d.isdigit()] if pd.notna(x) else []
        )

        def calculate_digit_presence(current_digits, historic_pool):
            if not current_digits or not historic_pool: return 0.0
            return len(set(current_digits).intersection(set(historic_pool))) / len(set(current_digits))

        N_lags = 7
        historic_pool_prize_7d = self._create_rolling_digit_pool(df['all_digits_prize1'].shift(1), N_lags)
        historic_pool_bottom2_7d = self._create_rolling_digit_pool(df['all_digits_bottom2'].shift(1), N_lags)

        df['digit_presence_prize1'] = [calculate_digit_presence(cd, hp) for cd, hp in zip(df['all_digits_prize1'], historic_pool_prize_7d)]
        df['digit_presence_bottom2'] = [calculate_digit_presence(cd, hp) for cd, hp in zip(df['all_digits_bottom2'], historic_pool_bottom2_7d)]

        df['sum_prize1'] = df['all_digits_prize1'].apply(lambda x: sum(x) if x else 0)
        df['sum_bottom2'] = df['all_digits_bottom2'].apply(lambda x: sum(x) if x else 0)

        df['digital_root_prize1'] = df['sum_prize1'].apply(lambda n: (n - 1) % 9 + 1 if n > 0 else 0)
        df['mod9_prize1'] = df['sum_prize1'] % 9

        # --- NEW FEATURE: Last seen related features ---
        # Initialize columns for last seen
        df['last_2d_was_prev_draw'] = 0 # Default to 0 (False)
        df['time_since_last_seen_2d'] = -1 # Default to -1
        df['last_3d_was_prev_draw'] = 0 # Default to 0 (False)
        df['time_since_last_seen_3d'] = -1 # Default to -1

        # Calculate 'last_2d_was_prev_draw' and 'time_since_last_seen_2d'
        last_seen_2d_idx = {} # Stores the index of the last appearance for each 2-digit number
        last_seen_3d_idx = {} # Stores the index of the last appearance for each 3-digit number

        for i in range(len(df)): # Iterate through all rows
            current_2d = df.loc[i, 'เลขท้าย 2 ตัว_temp']
            current_3d = df.loc[i, 'รางวัลที่ 1_temp']

            # Calculate 'last_X_was_prev_draw'
            if i > 0: # Check from the second row onwards
                if pd.notna(current_2d) and pd.notna(df.loc[i-1, 'เลขท้าย 2 ตัว_temp']):
                    if int(df.loc[i-1, 'เลขท้าย 2 ตัว_temp']) == int(current_2d):
                        df.loc[i, 'last_2d_was_prev_draw'] = 1

                if pd.notna(current_3d) and pd.notna(df.loc[i-1, 'รางวัลที่ 1_temp']):
                    if int(df.loc[i-1, 'รางวัลที่ 1_temp']) == int(current_3d):
                        df.loc[i, 'last_3d_was_prev_draw'] = 1

            # Calculate 'time_since_last_seen_X'
            if pd.notna(current_2d):
                current_2d_int = int(current_2d)
                if current_2d_int in last_seen_2d_idx:
                    df.loc[i, 'time_since_last_seen_2d'] = i - last_seen_2d_idx[current_2d_int]
                else:
                    df.loc[i, 'time_since_last_seen_2d'] = i + 1 # First occurrence, set a unique large value based on index
                last_seen_2d_idx[current_2d_int] = i # Update last seen index

            if pd.notna(current_3d):
                current_3d_int = int(current_3d)
                if current_3d_int in last_seen_3d_idx:
                    df.loc[i, 'time_since_last_seen_3d'] = i - last_seen_3d_idx[current_3d_int]
                else:
                    df.loc[i, 'time_since_last_seen_3d'] = i + 1 # First occurrence, set a unique large value
                last_seen_3d_idx[current_3d_int] = i # Update last seen index

        # Shift these features by 1 to make them available *before* the draw (for prediction)
        # Use .copy() to avoid SettingWithCopyWarning if original df might be chained
        df['last_2d_was_prev_draw'] = df['last_2d_was_prev_draw'].shift(1).fillna(0).astype(int)
        df['time_since_last_seen_2d'] = df['time_since_last_seen_2d'].shift(1).fillna(df['time_since_last_seen_2d'].max() + 1).astype(int)
        df['last_3d_was_prev_draw'] = df['last_3d_was_prev_draw'].shift(1).fillna(0).astype(int)
        df['time_since_last_seen_3d'] = df['time_since_last_seen_3d'].shift(1).fillna(df['time_since_last_seen_3d'].max() + 1).astype(int)

        print("✅ สร้างฟีเจอร์ทั้งหมดเรียบร้อยแล้ว")
        return df

    # Helper for LSTM data
    def _create_sequences(self, data, num_classes, seq_length=10):
        X, y = [], []
        if data.empty or len(data) <= seq_length: return np.empty((0, seq_length, 1)), np.empty((0, num_classes))

        data_values = data.values
        filtered_data_values = data_values[(data_values >= 0) & (data_values < num_classes)]

        if len(filtered_data_values) < seq_length + 1:
            return np.empty((0, seq_length, 1)), np.empty((0, num_classes))

        for i in range(len(filtered_data_values) - seq_length):
            X.append(filtered_data_values[i:i + seq_length])
            y.append(filtered_data_values[i + seq_length])

        if not X: return np.empty((0, seq_length, 1)), np.empty((0, num_classes))

        X_final = np.array(X).reshape(-1, seq_length, 1)
        y_final = to_categorical(np.array(y), num_classes=num_classes)
        return X_final, y_final

    # 🤖 Phase 3: AI Predictive Modeling
    def train_final_models(self):
        print("\n--- ระยะที่ 3: การพัฒนาระบบปัญญาประดิษฐ์ ---")

        feature_cols = ['day_of_year_sin', 'day_of_year_cos', 'days_since_prev_draw',
                        'digit_presence_prize1', 'digit_presence_bottom2', 'sum_prize1', 'sum_bottom2',
                        'digital_root_prize1', 'mod9_prize1',
                        'last_2d_was_prev_draw', 'time_since_last_seen_2d',
                        'last_3d_was_prev_draw', 'time_since_last_seen_3d']

        X_features_full = self.df_features[feature_cols].fillna(0)

        # --- Train for 2-digit ---
        print("\n--- กำลังฝึกโมเดลสำหรับเลขท้าย 2 ตัว ---")
        self.final_model_2d = {}
        num_classes_2d = 100

        self.le_2d = LabelEncoder()
        self.le_2d.fit(np.arange(num_classes_2d))

        xgb_2d_estimator = XGBoostFixWrapper(num_class=num_classes_2d)

        estimators_2d = [
            ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
            ('xgb', xgb_2d_estimator)
        ]

        full_dummy_X_2d = pd.DataFrame(np.zeros((num_classes_2d, len(feature_cols))), columns=feature_cols)
        full_dummy_y_2d = pd.Series(np.arange(num_classes_2d))

        X_train_2d_combined = pd.concat([X_features_full, full_dummy_X_2d], ignore_index=True)
        y_train_2d_combined = pd.concat([self.y_2d_raw, full_dummy_y_2d], ignore_index=True)

        y_train_2d_encoded = self.le_2d.transform(y_train_2d_combined)

        print(f"📊 คลาสจริงที่มีอยู่ใน y_2d_raw: {np.sort(self.y_2d_raw.unique())}")
        print(f"🔄 คลาสทั้งหมดที่ใช้ฝึกโมเดล (Stacking) หลังจากรวม Dummy และ Encode: {np.sort(np.unique(y_train_2d_encoded))}")

        if not np.array_equal(np.sort(np.unique(y_train_2d_encoded)), np.arange(num_classes_2d)):
            missing = np.setdiff1d(np.arange(num_classes_2d), np.sort(np.unique(y_train_2d_encoded)))
            print(f"❗❗ ข้อผิดพลาดร้ายแรง: แม้จะเข้ารหัสแล้ว คลาสสำหรับเลข 2 ตัวยังไม่ครบถ้วน! คาดหวัง: {np.arange(num_classes_2d)}, ได้รับ: {np.sort(np.unique(y_train_2d_encoded))}. คลาสที่ขาด: {missing}")
            raise ValueError("Invalid target classes after all attempts to include dummy samples and encoding. Data integrity or encoding issue suspected.")

        stack_2d = StackingClassifier(estimators=estimators_2d, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42))

        print("💡 กำลังฝึก StackingClassifier (XGBoost + RandomForest) สำหรับเลขท้าย 2 ตัว...")
        stack_2d.fit(X_train_2d_combined, y_train_2d_encoded)
        self.final_model_2d['stack'] = stack_2d

        lstm_2d = Sequential([LSTM(64, input_shape=(10, 1)), Dense(num_classes_2d, activation='softmax')])
        lstm_2d.compile(optimizer='adam', loss='categorical_crossentropy')
        if self.sequences_2d.shape[0] > 0:
            print("💡 กำลังฝึก LSTM Model สำหรับเลขท้าย 2 ตัว...")
            lstm_2d.fit(self.sequences_2d, self.targets_2d, epochs=20, batch_size=32, verbose=0)
            self.final_model_2d['lstm'] = lstm_2d

        # --- Train for 3-digit ---
        print("\n--- กำลังฝึกโมเดลสำหรับเลขท้าย 3 ตัว ---")
        self.final_model_3d = {}
        num_classes_3d = 1000

        self.le_3d = LabelEncoder()
        self.le_3d.fit(np.arange(num_classes_3d))

        xgb_3d_estimator = XGBoostFixWrapper(num_class=num_classes_3d)

        estimators_3d = [
            ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
            ('xgb', xgb_3d_estimator)
        ]

        full_dummy_X_3d = pd.DataFrame(np.zeros((num_classes_3d, len(feature_cols))), columns=feature_cols)
        full_dummy_y_3d = pd.Series(np.arange(num_classes_3d))

        X_train_3d_combined = pd.concat([X_features_full, full_dummy_X_3d], ignore_index=True)
        y_train_3d_combined = pd.concat([self.y_3d_raw, full_dummy_y_3d], ignore_index=True)

        y_train_3d_encoded = self.le_3d.transform(y_train_3d_combined)

        print(f"📊 คลาสจริงที่มีอยู่ใน y_3d_raw: {np.sort(self.y_3d_raw.unique())}")
        print(f"🔄 คลาสทั้งหมดที่ใช้ฝึกโมเดล (Stacking) หลังจากรวม Dummy และ Encode: {np.sort(np.unique(y_train_3d_encoded))}")

        if not np.array_equal(np.sort(np.unique(y_train_3d_encoded)), np.arange(num_classes_3d)):
            missing = np.setdiff1d(np.arange(num_classes_3d), np.sort(np.unique(y_train_3d_encoded)))
            print(f"❗❗ ข้อผิดพลาดร้ายแรง: แม้จะเข้ารหัสแล้ว คลาสสำหรับเลข 3 ตัวยังไม่ครบถ้วน! คาดหวัง: {np.arange(num_classes_3d)}, ได้รับ: {np.sort(np.unique(y_train_3d_encoded))}. คลาสที่ขาด: {missing}")
            raise ValueError("Invalid target classes after all attempts to include dummy samples and encoding. Data integrity or encoding issue suspected.")

        stack_3d = StackingClassifier(estimators=estimators_3d, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42))

        print("💡 กำลังฝึก StackingClassifier (XGBoost + RandomForest) สำหรับเลขท้าย 3 ตัว...")
        stack_3d.fit(X_train_3d_combined, y_train_3d_encoded)
        self.final_model_3d['stack'] = stack_3d

        lstm_3d = Sequential([LSTM(64, input_shape=(10, 1)), Dense(num_classes_3d, activation='softmax')])
        lstm_3d.compile(optimizer='adam', loss='categorical_crossentropy')
        if self.sequences_3d.shape[0] > 0:
            print("💡 กำลังฝึก LSTM Model สำหรับเลขท้าย 3 ตัว...")
            lstm_3d.fit(self.sequences_3d, self.targets_3d, epochs=20, batch_size=32, verbose=0)
            self.final_model_3d['lstm'] = lstm_3d

        print("✅ ฝึกฝน Final Models ทั้งหมดเรียบร้อย")

    # 🎯 Phase 4: Prediction Engine
    def predict_next_draw(self, top_n=5):
        print("\n" + "="*80); print("ระยะที่ 4: 🏆 ผลการทำนายสำหรับงวดถัดไป 🏆"); print("="*80)

        feature_cols = ['day_of_year_sin', 'day_of_year_cos', 'days_since_prev_draw',
                        'digit_presence_prize1', 'digit_presence_bottom2', 'sum_prize1', 'sum_bottom2',
                        'digital_root_prize1', 'mod9_prize1',
                        'last_2d_was_prev_draw', 'time_since_last_seen_2d',
                        'last_3d_was_prev_draw', 'time_since_last_seen_3d']

        if self.df_features.empty:
            print("❗ ไม่มีข้อมูลฟีเจอร์สำหรับทำนาย โปรดตรวจสอบข้อมูลเริ่มต้น")
            return

        last_features = self.df_features[feature_cols].iloc[[-1]]

        last_actual_2d_draw = self.y_2d_raw.iloc[-1]
        last_actual_3d_draw = self.y_3d_raw.iloc[-1]

        # --- Predict 2-digit ---
        print("\n--- ผลการทำนายเลขท้าย 2 ตัว ---")
        if 'stack' in self.final_model_2d:
            stack_probs_2d_encoded = self.final_model_2d['stack'].predict_proba(last_features)[0]

            final_probs_2d = stack_probs_2d_encoded

            if 'lstm' in self.final_model_2d and self.y_2d_raw.shape[0] >= 10:
                if self.y_2d_raw.tail(10).shape[0] == 10:
                    last_seq_2d = self.y_2d_raw.tail(10).values.reshape(1, 10, 1)
                    lstm_probs_2d = self.final_model_2d['lstm'].predict(last_seq_2d, verbose=0)[0]
                    if lstm_probs_2d.shape[0] == stack_probs_2d_encoded.shape[0]:
                        final_probs_2d = (stack_probs_2d_encoded * 0.5) + (lstm_probs_2d * 0.5)
                    else:
                        print("⚠️ คำเตือน: ขนาดของ LSTM probability ไม่ตรงกับ Stack probability สำหรับ 2-digit. ใช้ Stack probability เพียงอย่างเดียว.")
                else:
                    print("⚠️ คำเตือน: ข้อมูลไม่เพียงพอสำหรับสร้าง LSTM sequence สำหรับเลข 2 ตัว (ต้องมีอย่างน้อย 10 งวด). ใช้ Stack probability เพียงอย่างเดียว.")

            # Filter out the last drawn number
            filtered_probs_2d = np.copy(final_probs_2d)
            if last_actual_2d_draw is not None and 0 <= last_actual_2d_draw <= 99:
                filtered_probs_2d[last_actual_2d_draw] = -1.0 # Set probability of last drawn to very low to exclude it

            top_encoded_indices_2d = np.argsort(filtered_probs_2d)[::-1][:top_n]
            top_actual_classes_2d = self.le_2d.inverse_transform(top_encoded_indices_2d)

            for i, actual_class in enumerate(top_actual_classes_2d):
                prob = final_probs_2d[actual_class] # Use original probability for display
                print(f"  อันดับที่ {i+1}: {actual_class:02d} (Confidence: {prob:.2%})")
        else:
            print("❗ โมเดล Stacking สำหรับเลข 2 ตัวยังไม่ถูกฝึก")


        # --- Predict 3-digit ---
        print("\n--- ผลการทำนายเลขท้าย 3 ตัว ---")
        if 'stack' in self.final_model_3d:
            stack_probs_3d_encoded = self.final_model_3d['stack'].predict_proba(last_features)[0]

            final_probs_3d = stack_probs_3d_encoded

            if 'lstm' in self.final_model_3d and self.y_3d_raw.shape[0] >= 10:
                if self.y_3d_raw.tail(10).shape[0] == 10:
                    last_seq_3d = self.y_3d_raw.tail(10).values.reshape(1, 10, 1)
                    lstm_probs_3d = self.final_model_3d['lstm'].predict(last_seq_3d, verbose=0)[0]
                    if lstm_probs_3d.shape[0] == stack_probs_3d_encoded.shape[0]:
                        final_probs_3d = (stack_probs_3d_encoded * 0.5) + (lstm_probs_3d * 0.5)
                    else:
                        print("⚠️ คำเตือน: ขนาดของ LSTM probability ไม่ตรงกับ Stack probability สำหรับ 3-digit. ใช้ Stack probability เพียงอย่างเดียว.")
                else:
                    print("⚠️ คำเตือน: ข้อมูลไม่เพียงพอสำหรับสร้าง LSTM sequence สำหรับเลข 3 ตัว (ต้องมีอย่างน้อย 10 งวด). ใช้ Stack probability เพียงอย่างเดียว.")

            # Filter out the last drawn number
            filtered_probs_3d = np.copy(final_probs_3d)
            if last_actual_3d_draw is not None and 0 <= last_actual_3d_draw <= 999:
                 filtered_probs_3d[last_actual_3d_draw] = -1.0 # Set probability of last drawn to very low to exclude it

            top_encoded_indices_3d = np.argsort(filtered_probs_3d)[::-1][:top_n]
            top_actual_classes_3d = self.le_3d.inverse_transform(top_encoded_indices_3d)

            for i, actual_class in enumerate(top_actual_classes_3d):
                prob = final_probs_3d[actual_class] # Use original probability for display
                print(f"  อันดับที่ {i+1}: {actual_class:03d} (Confidence: {prob:.2%})")
        else:
            print("❗ โมเดล Stacking สำหรับเลข 3 ตัวยังไม่ถูกฝึก")


# --- Main Execution Block ---
if 'df_cleaned' in globals() and not df_cleaned.empty:
    system = UltimateLotterySystem(df_cleaned)
    system.train_final_models()
    system.predict_next_draw()
else:
    print("\n❗ ไม่สามารถเริ่มการทำงานของเซลล์ที่ 2 ได้ เนื่องจาก 'df_cleaned' จากเซลล์ที่ 1 ไม่พร้อมใช้งาน")
```

```python
# ==============================================================================
# --- Lotto Analyzer V11.6 - เซลล์ที่ 3/3 (Advanced Optimization) ---
# --- Focus: Hyperparameter Tuning for Stacking/XGBoost/RandomForest ---
# --- Using: RandomizedSearchCV with TimeSeriesSplit ---
# ==============================================================================

# Import necessary libraries for tuning
from sklearn.model_selection import RandomizedSearchCV, TimeSeriesSplit
from scipy.stats import randint, uniform
# import lightgbm as lgb # If we decide to try LightGBM later

print("🚀 กำลังเริ่มต้นการปรับปรุงความแม่นยำของโมเดล (Phase 5)...")

# ตรวจสอบว่ามี system object ที่ถูกสร้างจากเซลล์ที่ 2 แล้ว
if 'system' not in globals():
    print("❗ โปรดรันเซลล์ที่ 2 ก่อน เพื่อให้ 'system' object ถูกสร้าง")
else:
    # เตรียมข้อมูลและฟีเจอร์
    feature_cols = ['day_of_year_sin', 'day_of_year_cos', 'days_since_prev_draw',
                    'digit_presence_prize1', 'digit_presence_bottom2', 'sum_prize1', 'sum_bottom2',
                    'digital_root_prize1', 'mod9_prize1',
                    'last_2d_was_prev_draw', 'time_since_last_seen_2d',
                    'last_3d_was_prev_draw', 'time_since_last_seen_3d']

    X_features_full = system.df_features[feature_cols].fillna(0)

    # --- Prepare for 2-digit tuning ---
    y_2d_encoded = system.le_2d.transform(system.y_2d_raw) # Use raw target for tuning
    num_classes_2d = 100

    # Add dummy data for tuning as well (important for consistency with training)
    full_dummy_X_2d = pd.DataFrame(np.zeros((num_classes_2d, len(feature_cols))), columns=feature_cols)
    full_dummy_y_2d_encoded = system.le_2d.transform(pd.Series(np.arange(num_classes_2d)))

    X_tune_2d_combined = pd.concat([X_features_full, full_dummy_X_2d], ignore_index=True)
    y_tune_2d_combined = np.concatenate([y_2d_encoded, full_dummy_y_2d_encoded])


    # --- Define parameter distributions for RandomizedSearchCV ---
    param_dist_rf = {
        'n_estimators': randint(50, 200),
        'max_depth': randint(3, 10),
        'min_samples_split': randint(2, 10),
        'min_samples_leaf': randint(1, 5)
    }

    param_dist_xgb = {
        'estimator__n_estimators': randint(50, 200), # Note: 'estimator__' prefix for nested estimator
        'estimator__learning_rate': uniform(0.01, 0.15),
        'estimator__max_depth': randint(3, 8),
        'estimator__subsample': uniform(0.6, 0.4), # range from 0.6 to 1.0
        'estimator__colsample_bytree': uniform(0.6, 0.4), # range from 0.6 to 1.0
        'estimator__gamma': uniform(0, 0.2)
    }

    # --- Setup TimeSeriesSplit for Cross-Validation ---
    # n_splits determines how many splits (folds)
    # The last fold will always be the latest data, acting as validation.
    # We need to make sure n_splits doesn't exceed available data points after dummy data (if dummy data is at the end)
    # For TimeSeriesSplit, dummy data should probably NOT be part of the split, only for the final fit.
    # Let's use TimeSeriesSplit only on the *real* data, and then train the final model with dummy data.

    # Number of splits should be less than the number of unique time steps.
    n_splits_ts = min(5, len(X_features_full) - 1) # Ensure at least 2 splits

    # Use TimeSeriesSplit for tuning on the *real historical data*
    tscv_2d = TimeSeriesSplit(n_splits=n_splits_ts) # For 2D
    tscv_3d = TimeSeriesSplit(n_splits=n_splits_ts) # For 3D

    # --- Tuning for 2-digit models ---
    print("\n--- กำลังปรับจูน Hyperparameters สำหรับเลขท้าย 2 ตัว ---")

    # Random Forest Tuning
    rf_2d_base = RandomForestClassifier(random_state=42)
    rand_search_rf_2d = RandomizedSearchCV(rf_2d_base, param_distributions=param_dist_rf,
                                           n_iter=10, cv=tscv_2d, verbose=1, random_state=42, n_jobs=-1, scoring='accuracy')
    rand_search_rf_2d.fit(X_features_full, y_2d_encoded) # Tune on real data
    best_rf_2d = rand_search_rf_2d.best_estimator_
    print(f"✅ Best RF 2D Params: {rand_search_rf_2d.best_params_}")

    # XGBoost Tuning
    # For XGBoost, we need to tune the wrapper. RandomizedSearchCV can tune nested estimators.
    xgb_2d_wrapper_base = XGBoostFixWrapper(num_class=num_classes_2d)
    rand_search_xgb_2d = RandomizedSearchCV(xgb_2d_wrapper_base, param_distributions=param_dist_xgb,
                                            n_iter=10, cv=tscv_2d, verbose=1, random_state=42, n_jobs=-1, scoring='accuracy')
    rand_search_xgb_2d.fit(X_features_full, y_2d_encoded) # Tune on real data
    best_xgb_2d_wrapper = rand_search_xgb_2d.best_estimator_
    print(f"✅ Best XGB 2D Params: {rand_search_xgb_2d.best_params_}")

    # --- Re-train Stacking with best estimators (on combined data including dummy) ---
    print("\n--- กำลังฝึก StackingClassifier (2D) ด้วย Hyperparameters ที่ดีที่สุด ---")
    estimators_2d_tuned = [
        ('rf', best_rf_2d),
        ('xgb', best_xgb_2d_wrapper)
    ]

    # We still use StratifiedKFold for the final Stacking Classifier's internal CV
    # because the dummy data needs to be stratified.
    stack_2d_tuned = StackingClassifier(estimators=estimators_2d_tuned, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42))
    stack_2d_tuned.fit(X_tune_2d_combined, y_tune_2d_combined)
    system.final_model_2d['stack'] = stack_2d_tuned
    print("✅ StackingClassifier (2D) ฝึกเสร็จสิ้น")


    # --- Tuning for 3-digit models ---
    print("\n--- กำลังปรับจูน Hyperparameters สำหรับเลขท้าย 3 ตัว ---")

    y_3d_encoded = system.le_3d.transform(system.y_3d_raw) # Use raw target for tuning
    num_classes_3d = 1000

    # Add dummy data for tuning as well (important for consistency with training)
    full_dummy_X_3d = pd.DataFrame(np.zeros((num_classes_3d, len(feature_cols))), columns=feature_cols)
    full_dummy_y_3d_encoded = system.le_3d.transform(pd.Series(np.arange(num_classes_3d)))

    X_tune_3d_combined = pd.concat([X_features_full, full_dummy_X_3d], ignore_index=True)
    y_tune_3d_combined = np.concatenate([y_3d_encoded, full_dummy_y_3d_encoded])

    # Random Forest Tuning (for 3D)
    rf_3d_base = RandomForestClassifier(random_state=42)
    rand_search_rf_3d = RandomizedSearchCV(rf_3d_base, param_distributions=param_dist_rf,
                                           n_iter=10, cv=tscv_3d, verbose=1, random_state=42, n_jobs=-1, scoring='accuracy')
    rand_search_rf_3d.fit(X_features_full, y_3d_encoded) # Tune on real data
    best_rf_3d = rand_search_rf_3d.best_estimator_
    print(f"✅ Best RF 3D Params: {rand_search_rf_3d.best_params_}")

    # XGBoost Tuning (for 3D)
    xgb_3d_wrapper_base = XGBoostFixWrapper(num_class=num_classes_3d)
    rand_search_xgb_3d = RandomizedSearchCV(xgb_3d_wrapper_base, param_distributions=param_dist_xgb,
                                            n_iter=10, cv=tscv_3d, verbose=1, random_state=42, n_jobs=-1, scoring='accuracy')
    rand_search_xgb_3d.fit(X_features_full, y_3d_encoded) # Tune on real data
    best_xgb_3d_wrapper = rand_search_xgb_3d.best_estimator_
    print(f"✅ Best XGB 3D Params: {rand_search_xgb_3d.best_params_}")

    # --- Re-train Stacking with best estimators (on combined data including dummy) ---
    print("\n--- กำลังฝึก StackingClassifier (3D) ด้วย Hyperparameters ที่ดีที่สุด ---")
    estimators_3d_tuned = [
        ('rf', best_rf_3d),
        ('xgb', best_xgb_3d_wrapper)
    ]
    stack_3d_tuned = StackingClassifier(estimators=estimators_3d_tuned, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42))
    stack_3d_tuned.fit(X_tune_3d_combined, y_tune_3d_combined)
    system.final_model_3d['stack'] = stack_3d_tuned
    print("✅ StackingClassifier (3D) ฝึกเสร็จสิ้น")


    # --- LSTM Model Retraining (Optional, if tuning needed) ---
    # For LSTM, tuning often involves optimizing network architecture (layers, units),
    # learning rate, and epochs. This is usually done manually or with specialized tools
    # like Keras Tuner. For simplicity, we'll keep epochs=20 as adjusted in V11.4 for now.
    # If you want to tune LSTM, it would involve a separate tuning loop.

    print("\n✅ การปรับปรุงโมเดลขั้นสูงเสร็จสิ้น")

    # หลังจากปรับปรุงโมเดลเสร็จสิ้น เราก็สามารถเรียกการทำนายอีกครั้งได้เลย
    system.predict_next_draw()
```
